# **Inventory Analysis**

## **Introduction**

**Apa Itu Inventory Analysis?**

Inventory Analysis adalah:

> analisis data stok/barang untuk membantu perusahaan mengelola persediaan dengan lebih efisien.


**Latar Belakang**

Dua ekstrem yang sering terjadi:

* **Overstock**

  * gudang penuh
  * biaya penyimpanan tinggi
  * risiko barang usang / kadaluarsa

* **Stockout**

  * barang habis saat dibutuhkan
  * kehilangan penjualan
  * customer kecewa

Masalah utama:

> Tidak ada keseimbangan antara supply dan demand

**Tujuan**

- Tujuan Utama:

    > Menjaga stok di level optimal:
    > - tidak terlalu banyak,
    > - tidak terlalu sedikit.

- Tujuan Spesifik:

    - Menghindari overstock (biaya tinggi)
    - Menghindari stockout (lost sales)
    - Menjaga aliran barang tetap lancar
    - Mengoptimalkan penggunaan cash

Contoh Pertanyaan Bisnis

* Produk mana paling cepat habis?
* Produk mana jarang terjual?
* Kapan harus restock?
* Supplier mana paling lambat?
* Kenapa stockout sering terjadi?

**Kenapa Perlu Metrik?**

Stok itu dinamis:

* permintaan berubah
* supply tidak selalu stabil
* lead time berbeda-beda

Tanpa metrik:

> Keputusan hanya berdasarkan feeling → sangat berisiko

## **Metrics**

### **Introduction**

**Metrik Utama**


- Inventory Turnover

    **Definisi:**  
    Seberapa cepat stok terjual dalam periode tertentu

    Konsep:

    $$Inventory\ Turnover = \frac{\text{Cost of Goods Sold (COGS)}}{\text{Average Inventory}}$$


    **Intuisi:**  

    > Seberapa “cepat uang berputar”

    * Tinggi → barang cepat laku ✔
    * Rendah → barang menumpuk ❌

    **Makna penting:**  

    * Turnover tinggi = efisien
    * Turnover rendah = overstock

    **Kontribusi ke tujuan:**  

    ✔ Mengukur efisiensi stok

    $$ \ $$

- Days Inventory Outstanding (DIO)

    **Definisi:**    
    Berapa lama rata-rata barang berada di gudang

    Konsep:

    $$DIO = \frac{365}{Inventory\ Turnover}$$


    **Intuisi:**  

    > Berapa hari uang “terkunci” dalam bentuk barang

    * DIO tinggi → barang lama tidak terjual ❌
    * DIO rendah → barang cepat bergerak ✔

    **Kontribusi ke tujuan:**  

    ✔ Mengontrol efisiensi waktu stok

    $$ \ $$

- Stock Level

    **Definisi:**  
    Jumlah barang yang tersedia saat ini
    
    **Intuisi:**  
    
    > Ini “posisi saat ini”
    
    * Terlalu tinggi → overstock
    * Terlalu rendah → risk stockout
    
    Tapi:
    
    * Angka ini harus dilihat bersama demand
    
    **Kontribusi ke tujuan:**  

    ✔ Monitoring kondisi real-time

    $$ \ $$

- Stockout Rate

    **Definisi:**  
    Persentase kejadian kehabisan stok

    Konsep:

    $$Stockout\ Rate = \frac{\text{Number of Stockout Events}}{\text{Total Demand Events}} \times 100\% $$


    **Intuisi:**  

    > Seberapa sering kita “kehilangan kesempatan jual”

    * Tinggi → banyak lost sales ❌
    * Rendah → supply cukup ✔

    **Kontribusi ke tujuan:**  

    ✔ Mengukur kehilangan revenue



### **Addition**

**Hubungan Antar Metrik**


Konflik Utama:

> **Mengurangi stockout vs menghindari overstock**


- Turnover Tinggi + Stockout Tinggi

    **Artinya:**

    * barang cepat habis ✔
    * tapi supply kurang ❌


    Intuisi:

    > Demand tinggi, tapi stok tidak cukup


- Turnover Rendah + DIO Tinggi

    **Artinya:**

    * barang lama di gudang ❌


    Intuisi:

    > Uang “nganggur” dalam bentuk stok


- Stock Level Tinggi + Stockout Rendah

    **Artinya:**

    * aman dari kehabisan ✔
    * tapi mungkin overstock ❌

    Perlu balance, tidak ada angka “sempurna”

**Dari Metrik → Insight Nyata**

- Contoh kamu:  
    > Produk sering habis sebelum restock

- Insight:

    > Demand lebih cepat dari supply

    Kemungkinan:

    * forecast salah
    * lead time lama
    * stok awal terlalu kecil


**Dari Insight → Keputusan**

- Action yang tepat:

    - Tingkatkan jumlah pembelian (restock lebih banyak)
    - Percepat supply chain
    - Gunakan safety stock
    - Perbaiki demand forecasting

- Contoh lain:

    * DIO tinggi
    * Turnover rendah


    Keputusan:

    * kurangi pembelian
    * diskon untuk menghabiskan stok
    * evaluasi produk


- Intuisi Besar yang Harus Kamu Tangkap

    > Inventory itu bukan tentang “punya banyak barang”, tapi “punya barang yang tepat, di waktu yang tepat, dalam jumlah yang tepat”

## **Inventory Turnover Analysis**

Paling penting dalam inventory.

Fokus:

> mengukur seberapa cepat stok terjual dan diganti.

Semakin cepat turnover:

- biasanya barang lebih laku,
- cash flow lebih sehat.


Inventory Turnover

Mengukur:

> berapa kali rata-rata stok “berputar” dalam periode tertentu.

Rumus

$$\text{Inventory Turnover}= \frac{\text{COGS}}{\text{Average Inventory}}$$


Keterangan:

| Variabel          | Arti               |
| ----------------- | ------------------ |
| COGS              | cost of goods sold |
| Average Inventory | rata-rata stok     |



Intuisi  

- Turnover Tinggi

    ✅ barang cepat laku  
    ✅ stok bergerak cepat  
    ✅ gudang lebih efisien  

- Turnover Rendah

    ❌ barang lama tersimpan  
    ❌ risiko dead stock  
    ❌ cash flow tertahan  

Contoh

| COGS     | Avg Inventory |
| -------- | ------------- |
| 100 juta | 20 juta       |

Hitung:
$$\frac{100}{20} = 5$$

Interpretasi

> Stok berputar 5 kali dalam periode tertentu.

## **Stockout Analysis**

Fokus:

> mendeteksi kehabisan stok dan dampaknya terhadap penjualan.

Stockout berbahaya karena:

- penjualan hilang,
- customer kecewa,
- customer bisa pindah ke kompetitor.

Rumus Stockout Rate

$$\text{Stockout Rate} = \frac{\text{Stockout Events}}{\text{Total Demand}}×100\%$$

Contoh

| Demand | Stockout |
| ------ | -------- |
| 1000   | 50       |

Hitung

$$\text{Stockout Rate} = \frac{50}{1000}×100\% = 5\%$$

Insight

> 5% permintaan gagal dipenuhi karena stok habis.

## **Overstock Analysis**

Fokus:

> mendeteksi stok berlebihan yang terlalu lama tersimpan.

Overstock sering terjadi karena:

- forecasting salah,
- demand turun,
- pembelian terlalu banyak.

Risiko Overstock

❌ biaya gudang  
❌ barang expired  
❌ cash flow tertahan  
❌ barang menjadi dead stock  

Contoh Insight

| Produk | Kondisi            |
| ------ | ------------------ |
| A      | slow moving        |
| B      | stok terlalu besar |

Barang yang terlalu lama di gudang:

> seperti uang yang “diam” dan tidak menghasilkan profit.

## **Reorder Analysis**

Menentukan:

> kapan harus restock agar stok tidak habis.

Tujuannya:

- menjaga supply tetap aman,
- mengurangi stockout.


Reorder Point (ROP)

> Titik minimum stok sebelum melakukan pemesanan ulang.

Rumus

$$\text{ROP = Demand During Lead Time + Safety Stock}$$

Keterangan:

| Variabel                | Arti                             |
| ----------------------- | -------------------------------- |
| Demand During Lead Time | kebutuhan saat menunggu supplier |
| Safety Stock            | stok cadangan                    |

Jika stok menyentuh:

> ROP

maka:

> perusahaan harus segera restock, agar stok tidak habis saat menunggu supplier.

Contoh:

| Metric           | Nilai |
| ---------------- | ----- |
| Demand Lead Time | 100   |
| Safety Stock     | 20    |

Hitung
$$\text{100 + 20 = 120}$$

Interpretasi

> Saat stok tersisa 120 unit, maka lakukan pemesanan ulang.

### **Inventory Turnover Analysis**

Fokus:

> melihat seberapa cepat stok berputar.


- Step 1 — Cari COGS

    COGS:

    > total biaya barang yang terjual.

    Rumus:

    $$
    COGS = Unit\ Sold \times Purchase\ Price
    $$


    Product A

    $$
    990 \times 50.000
    =49.500.000
    $$


    Product B

    $$
    1000 \times 40.000
    =40.000.000
    $$


- Step 2 — Cari Average Inventory

    Rumus sederhana:

    $$
    Average\ Inventory=
    \frac{
    Beginning\ Stock+Ending\ Stock
    }{2}
    $$


    Product A

    Awal:
    $$
    1000
    $$

    Akhir:
    $$
    10
    $$

    $$
    \frac{1000+10}{2}=505
    $$


    Product B

    Awal:
    $$
    6000
    $$

    Akhir:
    $$
    5000
    $$

    Maka:
    $$
    \frac{6000+5000}{2}=5500
    $$


- Step 3 — Inventory Turnover

    Rumus:

    $$\text{Inventory\ Turnover}=\frac{COGS}{\text{Average\ Inventory\ Value}}$$


    Cari Inventory Value

    - Product A

        $$
        505 \times 50.000
        =25.250.000
        $$


    - Product B

        $$
        5500 \times 40.000
        =220.000.000
        $$


    Hitung Turnover

    - Product A

        $$
        \frac{49.500.000}{25.250.000}
        \approx1.96
        $$


    - Product B

        $$
        \frac{40.000.000}{220.000.000}
        \approx0.18
        $$


Insight

| Product | Turnover | Insight                      |
| ------- | -------- | ---------------------------- |
| A       | 1.96     | barang cukup cepat bergerak  |
| B       | 0.18     | barang sangat lambat terjual |



### **Stockout Analysis**

    990+50=1040
    $$


    Product B

    $$
    1000+5=1005
    $$

    $$\text{Stockout\ Rate}=\frac{Stockout\ Events}{Total\ Demand}\times100\%$$

    \approx4.8%
    $$
    
    
    Product B
    
    $$

    $$
    
    
    Insight
    
    | Product | Insight           |
    | ------- | ----------------- |
    | A       | stok sering habis |
    | B       | stok relatif aman |

### **Reorder Analysis**



Fokus:

> menentukan kapan restock.


- Step 1 — Cari Daily Demand

    Misal periode:
    30 hari.


    Product A

    $$
    \frac{990}{30}=33
    $$

    unit/hari.


    Product B

    $$
    \frac{1000}{30}=33
    $$

unit/hari.


- Step 2 — Demand During Lead Time

    Rumus:

    $$
    Daily\ Demand \times Lead\ Time
    $$


    Product A

    $$
    33\times3=99
    $$


    Product B

    $$
    33\times10=330
    $$


- Step 3 — Tambah Safety Stock

    Misal:

    * A = 20
    * B = 50


    Rumus ROP

    $$\text{ROP=Demand\ During\ Lead\ Time+Safety\ Stock}$$


    Product A

    $$
    99+20=119
    $$


    Product B

    $$
    330+50=380
    $$


    Insight
    
    Kalau stok menyentuh:
    
    * A → 119,
    * B → 380,
    
    maka:
    
    * lakukan reorder.
