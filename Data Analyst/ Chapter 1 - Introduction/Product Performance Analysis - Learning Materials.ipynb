# **Product Performance Analysis**

## **Introduction**

**Apa Itu Product Performance Analysis?**

Product Performance Analysis adalah:

> analisis performa produk untuk mengetahui produk mana yang berhasil, bermasalah, atau perlu ditingkatkan.

Contoh Product

| Industri   | Produk           |
|------- |------------- |
| E-commerce | barang           |
| SaaS       | feature aplikasi |
| Streaming  | konten           |
| Game       | item/game mode   |


**Latar Belakang**  

Dalam bisnis, sering terjadi:

- Banyak produk dijual
- Banyak SKU (variasi)
- Banyak transaksi

Tapi…

> Tidak semua produk berkontribusi positif ke bisnis

Contoh real:

- Produk laris → tapi margin tipis
- Produk jarang laku → tapi margin tinggi
- Produk sering diretur → bikin rugi

Masalah utama:  
> Tidak tahu produk mana yang benar-benar menguntungkan

**Tujuan**  

- Tujuan Utama:

    > Mengoptimalkan portofolio produk

- Tujuan Spesifik:
    - Mengetahui produk paling profitable
    - Mengidentifikasi produk bermasalah
    - Mengambil keputusan: scale, improve, atau stop
    - Mengoptimalkan strategi harga, stok, dan marketing

Contoh Pertanyaan Bisnis

* Produk mana revenue tertinggi?
* Produk mana sering direturn?
* Feature mana jarang dipakai?
* Produk mana retention terbaik?
* Produk mana conversion rendah?

**Kenapa Perlu Metrik?**

Kalau hanya lihat:  
> “produk ini laku keras”

Itu belum cukup

Karena:

- Laku ≠ untung
- Banyak order ≠ efisien

Maka:  
> Kita butuh metrik untuk melihat performa dari berbagai sisi

**Jenis Product Performance Analysis**

|---------------- |------------- |

| Revenue Performance | pendapatan       |
| Usage Analysis      | penggunaan       |

## **Metrics**

### **Introduction**

**Metrik**  

- Sales per Product

    - **Definisi:**  
        Jumlah penjualan (unit atau revenue) per produk

    - **Intuisi:**

        > Seberapa “populer” produk itu

        Tapi hati-hati:

        * Produk populer belum tentu menguntungkan

    - **Kontribusi ke tujuan:**

        ✔ Identifikasi demand  
        ✔ Menentukan produk favorit

    $$ \ $$

- Profit per Product

    - **Definisi:**  
        Keuntungan yang dihasilkan tiap produk

        Secara konsep:

        $$Profit = Revenue - Cost$$

    
    - **Intuisi:**

        > Ini metrik paling penting

        * Produk bisa laris tapi rugi
        * Produk bisa sepi tapi sangat menguntungkan

        Fokus bisnis sebenarnya:

        > **bukan jual banyak, tapi untung banyak**

    - **Kontribusi ke tujuan:**

        ✔ Menentukan produk yang layak diprioritaskan

    $$ \ $$

- Conversion Rate per Product

    - **Definisi:**  
        Persentase user yang melihat produk lalu membeli

        Konsep:

        $$Conversion\ Rate = \frac{\text{Purchases}}{\text{Product Views}} \times 100%$$


    - **Intuisi:**

        > Seberapa “menarik dan meyakinkan” produk itu

        Kalau conversion rendah:

        * produk kurang menarik
        * harga tidak sesuai
        * deskripsi kurang jelas

    - **Kontribusi ke tujuan:**

        ✔ Evaluasi kualitas produk dari sisi user

    $$ \ $$

- Return Rate

    - **Definisi:**  
        Persentase produk yang dikembalikan

        Konsep:

        $$Return\ Rate = \frac{\text{Returned Items}}{\text{Total Sold}} \times 100\% $$


    - **Intuisi:**

        > Ini “biaya tersembunyi”

        Return tinggi bisa berarti:

        * kualitas buruk
        * ekspektasi tidak sesuai
        * deskripsi misleading

        Dampaknya:

        * rugi biaya logistik
        * rugi reputasi

    - **Kontribusi ke tujuan:**

        ✔ Identifikasi produk bermasalah

### **Addition**

Hubungan Antar Metrik

- Kasus 1: Sales Tinggi, Profit Rendah

    (contoh kamu)

    **Artinya:**

    * produk laris ✔
    * tapi tidak menghasilkan uang ❌

    Penyebab:

    * biaya produksi tinggi
    * harga terlalu murah
    * diskon berlebihan

    Intuisi:

    > “Kita sibuk jualan, tapi tidak kaya”

    $$ \ $$

- Kasus 2: Conversion Tinggi, Sales Rendah

    **Artinya:**

    * produk menarik ✔
    * tapi exposure kecil ❌

    Solusi:

    * tingkatkan marketing
    * naikkan visibility
    
    $$ \ $$

- Kasus 3: Sales Tinggi, Return Rate Tinggi

    **Artinya:**

    * banyak yang beli ✔
    * banyak yang kecewa ❌

    Ini berbahaya:

    * profit akan terkikis
    * brand rusak
    
    $$ \ $$

- Kasus 4: Profit Tinggi, Sales Rendah

    **Artinya:**

    * produk bagus ✔
    * tapi belum dikenal ❌

    Peluang besar:

    * scale up marketing
    

Dari Metrik → Insight Nyata


- Contoh Kasus:

    Produk A:

    * Sales tinggi
    * Profit rendah
    * Return rate rendah


- Insight:

    * Produk ini disukai pasar
    * Tapi pricing / cost tidak optimal


Dari Insight → Keputusan

- Action yang tepat:

    - Evaluasi harga

        > bisa dinaikkan sedikit

    - Optimasi biaya produksi

        > cari supplier lebih murah

    - Kurangi diskon berlebihan


- Contoh lain:

    - Produk B:
    
        * Conversion rendah
        * Return tinggi
    
    - Keputusan:
    
        * perbaiki kualitas produk
        * perbaiki deskripsi & foto
        * atau bahkan stop produk

- Intuisi

    > Produk terbaik bukan yang paling laris, tapi yang paling menguntungkan dan berkelanjutan


**Alur Berpikir**

- **Masalah:**  
   Tidak semua produk menguntungkan

- **Tujuan:**  
   Optimasi performa produk

- **Solusi:**  
   Gunakan metrik multi-dimensi

- **Insight:**  
   Temukan produk:

   * hero (unggulan)
   * problem (bermasalah)
   * hidden gem

- **Action:**  
   Scale / Improve / Kill

**Insight Lanjutan**

Biasanya analisis ini dikembangkan ke:

* **Pareto Analysis (80/20 rule)**  
  → 20% produk menghasilkan 80% revenue

* **Product Portfolio Strategy**

  * Star product
  * Cash cow
  * Dog product

* Integrasi dengan:

  * Marketing (produk mana dipromosikan)
  * Inventory (stok mana diperbanyak)

**Ringkasan**

| Metrik     | Menjawab            |
| ---------- | ------------------- |
| Sales      | Seberapa laku       |
| Profit     | Seberapa untung     |
| Conversion | Seberapa menarik    |
| Return     | Seberapa bermasalah |

Output akhirnya:

> **Menentukan produk mana yang harus didorong, diperbaiki, atau dihentikan**


## **Sales Performance Analysis**

Analisis untuk:

> mengukur performa penjualan produk.

Biasanya fokus pada:

* jumlah unit terjual,
* growth sales,
* produk paling laku.


Melihat:

> apakah produk berhasil di pasar.

Metric Umum

| Metric       | Fungsi                |
| ------------ | --------------------- |
| Units Sold   | jumlah produk terjual |
| Sales Growth | pertumbuhan penjualan |
| Top Product  | produk paling laku    |

Rumus Sales Growth

$$Sales\ Growth=\frac{Current\ Sales-Previous\ Sales}{Previous\ Sales}\times100\%$$

Insight

* growth tinggi → produk berkembang.
* growth rendah → demand menurun.

## **Revenue Performance Analysis**

Analisis untuk:

> mengukur pendapatan yang dihasilkan produk.


Produk dengan sales tinggi:

* belum tentu revenue tinggi.

Karena:

* harga produk berbeda.

Rumus Revenue

$$Revenue=Price\times Quantity\ Sold$$

Insight

* revenue tinggi → produk profitable.
* revenue turun → perlu evaluasi pricing atau demand.

## **Usage Analysis**

Analisis untuk:

> memahami bagaimana user menggunakan produk.


Fokus pada:

* fitur yang sering dipakai,
* durasi penggunaan,
* aktivitas user.

Metric Umum

| Metric           | Fungsi           |
| ---------------- | ---------------- |
| Session Duration | lama penggunaan  |
| Active Users     | user aktif       |
| Feature Usage    | penggunaan fitur |

Rumus DAU Rate

$$DAU\ Rate=\frac{Daily\ Active\ Users}{Total\ Users}\times100\%$$

Insight

* usage tinggi → produk bernilai.
* usage rendah → user kurang tertarik.

## **Engagement Analysis**




Analisis untuk:

> mengukur interaksi user terhadap produk.


Semakin tinggi engagement:

* biasanya user lebih tertarik,
* peluang retention lebih tinggi.

Metric Umum

| Metric   | Fungsi        |
| -------- | ------------- |
| Clicks   | interaksi     |
| Likes    | ketertarikan  |
| Shares   | penyebaran    |
| Comments | feedback user |

Rumus Engagement Rate

$$Engagement\ Rate=\frac{Total\ Engagement}{Total\ Users}\times100\%$$

Insight

* engagement tinggi → feature menarik.
* engagement rendah → feature kurang relevan.

## **Retention Analysis**




Analisis untuk:

> mengukur kemampuan produk mempertahankan user.


Retention penting karena:

* mendapatkan user baru mahal,
* mempertahankan user lebih menguntungkan.

Rumus Retention Rate

$$Retention\ Rate=\frac{Remaining\ Users}{Initial\ Users}\times100\%$$

Insight

* retention tinggi → user puas.
* retention rendah → risiko churn tinggi.

## **Product Funnel Analysis**

Analisis untuk:

> melihat perjalanan user dalam product.


User biasanya melewati beberapa tahap:

> ```text id="2t5a7r"
> Visit → Signup → Active → Purchase
> ```

Fokus Funnel

Mencari:

> di tahap mana user banyak keluar.

Rumus Funnel Conversion

$$Funnel\ Conversion=\frac{Next\ Stage}{Previous\ Stage}\times100\%$$

Contoh

| Tahap    | User |
| -------- | ---- |
| Visit    | 1000 |
| Signup   | 300  |
| Purchase | 30   |

Insight

* banyak drop setelah signup,
* onboarding mungkin kurang baik,
* atau UX terlalu sulit.

## **Addition**

### **Statistics Analysis**

Statistik dalam Product Analysis

Product analysis biasanya digunakan untuk:

* memahami behavior user,
* mengevaluasi feature,
* meningkatkan engagement,
* dan meningkatkan retention product.


Statistik yang Sering Digunakan

| Statistik   | Fungsi                 |
| ----------- | ---------------------- |
| Mean        | rata-rata usage        |
| Correlation | hubungan metric        |
| T-Test      | compare feature        |
| Chi-Square  | conversion analysis    |
| Regression  | faktor retention       |
| Effect Size | besar pengaruh feature |




Contoh Statistical Analysis

| Pertanyaan                                                 | Statistik / Metode                                     |
| ---------------------------------------------------------- | ------------------------------------------------------ |
| Apakah feature baru meningkatkan engagement?               | A/B testing, p-value, confidence interval, effect size |
| Apakah redesign UI meningkatkan conversion?                | A/B testing, chi-square                                |
| Apakah user aktif memiliki retention lebih tinggi?         | correlation, regression                                |
| Apakah onboarding baru meningkatkan activation rate?       | t-test, A/B testing                                    |
| Faktor apa yang paling memengaruhi retention user?         | regression                                             |
| Apakah feature premium meningkatkan usage time?            | t-test                                                 |
| Apakah terdapat perbedaan engagement antar segment user?   | ANOVA                                                  |
| Apakah bug frequency memengaruhi churn?                    | correlation, regression                                |
| Apakah notifikasi meningkatkan daily active user?          | A/B testing                                            |
| Apakah user mobile dan desktop memiliki behavior berbeda?  | t-test                                                 |
| Apakah feature tertentu meningkatkan conversion rate?      | chi-square test                                        |
| Apakah session duration berhubungan dengan retention?      | correlation                                            |
| Seberapa besar pengaruh feature baru terhadap engagement?  | effect size                                            |
| Apakah update aplikasi meningkatkan customer satisfaction? | t-test, confidence interval                            |
| Apakah waktu loading memengaruhi conversion?               | regression, correlation                                |





Statistik yang Sering Dipakai Junior Product Analyst

| Statistik           | Fungsi                          |
| ------------------- | ------------------------------- |
| Mean                | rata-rata usage/engagement      |
| Correlation         | hubungan antar metric           |
| T-Test              | membandingkan feature           |
| Chi-Square          | conversion analysis             |
| Regression          | faktor retention/conversion     |

Intuisi

Product analysis membantu bisnis memahami:

* feature mana yang benar-benar berguna,
* behavior user terhadap product,
* dan apakah perubahan product meningkatkan engagement, retention, atau conversion.
