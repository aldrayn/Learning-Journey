# **Churn Analysis**

## **Introduction**

**Definisi**  

Churn adalah:

> kondisi ketika customer berhenti menggunakan produk/jasa.  

Contoh Churn

| Industri   | Contoh Churn       |
|------- |--------------- |
| Netflix    | berhenti langganan |
| E-commerce | tidak belanja lagi |
| Game       | berhenti bermain   |
| Bank       | tutup rekening     |
| Telco      | pindah provider    |


Churn ini penting dianalisis karena:
* mencari customer baru mahal,
* mempertahankan customer lebih murah.




**Latar Belakang**  

Dalam banyak bisnis (terutama subscription, e-commerce, aplikasi):

> Pelanggan tidak selalu pergi dengan “teriak” — mereka hilang diam-diam

Contoh:

- Tidak login lagi
- Tidak beli lagi
- Tidak pakai layanan lagi

Masalah utamanya:

- Perusahaan sadar pelanggan hilang setelah terlambat
- Tidak tahu kenapa mereka pergi
- Tidak tahu siapa yang akan pergi berikutnya

Dampaknya:

- Revenue turun
- Biaya akuisisi membengkak (cari pelanggan baru lebih mahal)

**Tujuan**  

Dari masalah tadi:

- Tujuan Utama:

    > Mengurangi kehilangan pelanggan (churn)

- Tujuan Spesifik:
    - Mendeteksi pelanggan yang berisiko churn
    - Memahami penyebab churn
    - Melakukan intervensi sebelum pelanggan pergi



Jenis Churn Analysis

| Jenis                         | Fokus                |
|-------------------------- |----------------- |
| Business/Descriptive Analysis | memahami pola churn  |
| Statistical Analysis          | menguji faktor churn |
| Machine Learning              | memprediksi churn    |

## **Metric**

### **Churn Rate (%)**

**Definisi:**   

> Persentase pelanggan yang berhenti dalam periode tertentu  

Secara konsep:

$$\text{Churn Rate} = \frac{\text{Customers Lost}}{\text{Total Customers at Start}} \times 100\%$$
    
    
Intuisi:
    
 > Seberapa “bocor” bisnis kamu

 Contoh:

 * 5% → normal
 * 20% → bahaya
    
 Makna penting:
    
> Ini adalah indikator utama kesehatan bisnis
    
Kontribusi ke tujuan:
    
✔ Mengukur seberapa besar masalah churn

### **Retention Rate**



**Definisi:**
> Persentase pelanggan yang tetap bertahan

Hubungannya:

$$
Retention = 100\% - Churn
$$

Intuisi:

> Fokus ke yang “bertahan”, bukan yang hilang

Kadang lebih powerful melihat:

> “berapa yang berhasil dipertahankan”

Kontribusi ke tujuan:

✔ Evaluasi keberhasilan strategi retention

### **Customer Tenure**



**Definisi:**  
> Lama waktu pelanggan bertahan

Intuisi:

> Kapan pelanggan biasanya pergi?

Contoh insight:
* Banyak churn di bulan pertama → onboarding jelek
* Banyak churn setelah 3 bulan → value tidak cukup

Kontribusi ke tujuan:

✔ Mengetahui “titik kritis” churn

### **Usage Frequency**



**Definisi:**  
> Seberapa sering pelanggan menggunakan produk/layanan

Intuisi:

> Perilaku adalah sinyal paling jujur
 
Biasanya:
> Penggunaan turun → churn akan datang  
> (Ini leading indicator / indikator sebelum kejadian)

Kontribusi ke tujuan:

✔ Prediksi churn lebih awal

### **Additional**

**Hubungan Antar Metrik**

Churn tidak berdiri sendiri.


- Pola Umum Churn:

    - Usage Frequency turun
    - Customer jadi jarang interaksi
    - Customer akhirnya berhenti

    Jadi:

    > **Churn itu proses, bukan kejadian tiba-tiba**  

    $$ \ $$


- Contoh Analisis Nyata:

    | Metrik     | Kondisi       |
    | ---------- | ------------- |
    | Usage      | turun drastis |
    | Tenure     | baru 1 bulan  |
    | Churn Rate | tinggi        |  

    $$ \ $$

- Insight:

    * Banyak pelanggan tidak menemukan value sejak awal
    * Onboarding atau first experience buruk


Dari Metrik → Penyebab Churn

Beberapa pola umum:

    
- Low Usage → Churn

    **Intuisi:**

    > Kalau tidak dipakai, pasti ditinggal

    **Akar masalah:**

    * Tidak butuh produk
    * Tidak paham cara pakai

    
- Short Tenure → Churn

    **Intuisi:**

    > Pelanggan tidak “terikat”

    **Akar masalah:**

    * First impression buruk
    * Value tidak cepat terasa

    
- High Churn di Segmen Tertentu

    (ini nyambung ke segmentation sebelumnya 👀)

    **Contoh:**

    * pelanggan baru → churn tinggi
    * pelanggan low-value → churn tinggi

**Dari Insight → Keputusan (Actionable)**

- Contoh Kasus:

    Hasil analisis:

    * Usage Frequency turun sebelum churn
    * Banyak churn di minggu pertama

- Insight:

    * Customer tidak menemukan value dengan cepat
    * Engagement rendah di awal

- Keputusan:

    - Kirim **reminder / notifikasi**
    - Buat **onboarding lebih jelas**
    - Beri **promo untuk aktivasi awal**
    - Edukasi penggunaan produk


- Intuisi

    > Churn analysis itu bukan cuma menghitung “siapa yang pergi”, tapi memahami “kenapa mereka mulai menjauh sebelum pergi”


## **Business Descriptive Churn Analysis**

Pertanyaan yang Dicari

* Berapa churn rate?
* User mana paling sering churn?
* Device apa paling tinggi churn?
* Subscription mana paling berisiko?


Metric Penting


- Churn Rate
  $$Churn\ Rate =\frac{Customer\ Churn}{Total\ Customer}\times100\%$$
  
  Contoh
  | Total Customer | Churn |
  |----------- |-- |
  | 1000           | 120   |

  Hitung
  $$
  \frac{120}{1000}\times100\%=12\%
  $$
  
  Interpretasi
  > 12% customer berhenti menggunakan layanan.


- Retention Rate
  
  $$Retention\ Rate =\frac{Customer\ Remaining}{Total\ Customer}\times100%$$
  
  Contoh
  
  $$\frac{880}{1000}=88\%$$


- Revenue Churn

  Mengukur revenue yang hilang.

  $$Revenue\ Churn =\frac{Lost\ Revenue}{Total\ Revenue}$$
  
  Contoh
  
  | Total Revenue | Lost Revenue |
  |---------- |--------- |
  | 100 juta      | 8 juta       |

  Hitung
  $8\%$








Analisis Umum

- Segment Analysis

  Contoh:
  | Segment | Churn Rate |
  |---- |------- |
  | Mobile  | 18%        |
  | Desktop | 8%         |

  Insight
  > User mobile lebih sering churn.


- Cohort Analysis
  
  Melihat:

  > kelompok user berdasarkan waktu join.


  Contoh
  
  | Join Month | Month 1 Retention |
  |------- |-------------- |
  | Januari    | 90%               |
  | Februari   | 75%               |
  
  
  Insight
  
  > User Februari retention lebih buruk.

Dashboard Churn

Biasanya berisi:

✅ churn rate  
✅ retention trend  
✅ churn by segment  
✅ churn by device  
✅ churn by subscription  
✅ revenue lost




Flow

> ```text
> Business Problem
>       ↓
> Define Churn
>       ↓
> Hitung Churn Rate
>       ↓
> Segment Analysis
>       ↓
> Trend Analysis
>       ↓
> Insight
>       ↓
> Recommendation
> ```




Contoh Recommendation

| Temuan                          | Recommendation           |
|---------------------------- |--------------------- |
| Mobile churn tinggi             | optimasi aplikasi mobile |
| User baru cepat churn           | onboarding improvement   |
| Subscription murah churn tinggi | loyalty program          |

## **Statistical Churn Analysis**

Tujuan:

> mencari apakah faktor tertentu benar-benar berhubungan dengan churn.

Karena churn analysis biasanya fokus pada:

- penyebab customer keluar,
- faktor risiko churn,
- dan strategi retention.

Contoh Pertanyaan

| Pertanyaan                                                     | Statistik / Metode          |
| -------------------------------------------------------------- | --------------------------- |
| Apakah premium user churn lebih rendah?                        | chi-square test             |
| Apakah durasi penggunaan memengaruhi churn?                    | regression, correlation     |
| Apakah customer support memengaruhi churn?                     | chi-square test, regression |
| Apakah rata-rata spending customer churn berbeda?              | t-test                      |
| Apakah churn berbeda antar region?                             | ANOVA, chi-square           |
| Apakah user aktif memiliki churn lebih rendah?                 | correlation, regression     |
| Apakah promo retention mengurangi churn?                       | A/B testing, p-value        |
| Faktor apa yang paling memengaruhi churn?                      | logistic regression         |
| Apakah customer dengan complaint lebih sering churn?           | chi-square test             |
| Apakah tenure customer memengaruhi churn?                      | regression                  |
| Apakah terdapat perbedaan churn antar segment customer?        | ANOVA, chi-square           |
| Apakah frekuensi transaksi berhubungan dengan churn?           | correlation                 |
| Apakah onboarding baru menurunkan churn?                       | t-test, effect size         |
| Apakah subscription plan tertentu memiliki churn lebih tinggi? | chi-square test             |
| Seberapa besar pengaruh engagement terhadap churn?             | regression, effect size     |


Statistika yang dipakai

| Kondisi               | Test / Statistik |
| --------------------- | ---------------- |
| Churn yes/no          | Chi-Square       |
| Compare average       | T-Test           |
| Banyak grup           | ANOVA            |
| Faktor pengaruh       | Regression       |
| Hubungan variabel     | Correlation      |
| Efektivitas perubahan | A/B Testing      |
| Besar pengaruh        | Effect Size      |








Contoh

| Plan    | Churn Rate |
|---- |------- |
| Free    | 20%        |
| Premium | 8%         |










Hipotesis


- H0 : Tidak ada perbedaan churn.
- H1 : Ada perbedaan churn.  
  
- Hasil :  
  
    $
    p = 0.01
    $
    
- Interpretasi :    
   
  Karena $p < 0.05$, maka:  
  > premium user memang churn lebih rendah secara signifikan.

Confidence Interval

> Digunakan untuk "melihat range estimasi churn."


Effect Size

> Digunakan untuk "melihat seberapa besar pengaruh faktor terhadap churn."

Flow

> ```text
> Business Problem
>       ↓
> Metric
>       ↓
> Hypothesis
>       ↓
> Statistical Test
>       ↓
> P-value
>       ↓
> Effect Size
>       ↓
> Decision
> ```

## **Machine Learning Churn Prediction**

**Konsep**

- Ini fokusnya:

    > memprediksi customer yang kemungkinan churn.

- Bukan hanya:

    > memahami churn,

- tetapi:

    > memprediksi churn sebelum terjadi.


- Bentuk masalah biasanya:

    > classification problem.

Struktur

- Target

  | Customer | Churn |
  |----- |-- |
  | A        | 1     |
  | B        | 0     |
  
      

- Features yang Sering Dipakai

  | Feature          | Contoh          |
  |------------- |------------ |
  | Usage            | login frequency |
  | Transaction      | total purchase  |
  | Engagement       | watch time      |
  | Customer support | jumlah komplain |
  | Subscription     | premium/free    |
  
      
 
- Model yang Sering Dipakai

  | Model               | Cocok untuk |
  |---------------- |-------- |
  | Logistic Regression | baseline    |
  | Random Forest       | populer     |
  | XGBoost             | powerful    |
  | LightGBM            | cepat       |
  | Neural Network      | advanced    |







Flow

> ```text
> Business Problem
>       ↓
> Data Collection
>       ↓
> Cleaning & Feature Engineering
>       ↓
> Train/Test Split
>       ↓
> Model Training
>       ↓
> Prediction
>       ↓
> Evaluation
>       ↓
> Business Recommendation
> ```

Evaluation Metrics

> Karena churn biasanya imbalance.


Metric Penting

| Metric    | Fungsi                     |
|------ |----------------------- |
| Accuracy  | keseluruhan benar          |
| Precision | prediksi churn yang benar  |
| Recall    | kemampuan menangkap churn  |
| F1-score  | balance precision-recall   |
| ROC-AUC   | kemampuan membedakan kelas |

**Contoh**

Model berhasil menemukan:

> 85% customer yang akan churn.


Output Model

| Customer | Probability Churn |
|----- |-------------- |
| A        | 92%               |
| B        | 15%               |
| C        | 80%               |


Lalu Dipakai Untuk

✅ promo personal  
✅ retention campaign  
✅ diskon  
✅ customer support prioritas  

Dashboard ML Churn

Biasanya berisi:

✅ churn prediction score  
✅ high-risk customer  
✅ feature importance  
✅ churn trend  
✅ model performance
