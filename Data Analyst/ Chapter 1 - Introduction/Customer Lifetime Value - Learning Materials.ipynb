# **Customer Lifetime Value**

## **Introduction**

### **Concept**


**Latar Belakang**

Banyak bisnis melakukan kesalahan besar:

* Fokus ke **penjualan sekali**
* Tidak peduli apakah pelanggan kembali atau tidak

Padahal:

* Mendapatkan pelanggan baru itu mahal
* Pelanggan lama jauh lebih menguntungkan

Masalah utama:

> Semua pelanggan dianggap sama nilainya

Konsep

- Customer Lifetime Value (CLV / LTV) adalah:

    > estimasi total nilai/revenue yang diberikan customer selama menggunakan produk atau layanan.

- CLV mencoba menjawab:

    > “Satu customer rata-rata menghasilkan uang berapa selama menjadi customer?”

- CLV penting karena perusahaan ingin tahu:

    > ```text id="2xnhy9"
    > Customer mana paling bernilai?
    > Apakah biaya akuisisi customer worth it?
    > Bagaimana fokus retention?
    > ```


**Tujuan**
- Tujuan Utama:

    > Mengetahui nilai sebenarnya dari setiap pelanggan

- Tujuan Spesifik:

    - Mengukur total kontribusi pelanggan
    - Menentukan batas biaya akuisisi (marketing)
    - Mengoptimalkan strategi retensi
    - Memprioritaskan pelanggan bernilai tinggi


**Kenapa Perlu Metrik?**

Kalau hanya lihat:

> “customer beli 100 ribu”

Itu misleading

Karena:

* bisa jadi dia beli 1x saja
* atau bisa jadi dia beli terus selama 2 tahun

Maka:

> Kita butuh metrik yang melihat “nilai jangka panjang”

Masalah yang Ingin Dijawab


- Contoh 1

    > “Apakah biaya iklan terlalu mahal dibanding nilai customer?”


- Contoh 2

    > “Customer premium lebih menguntungkan?”


- Contoh 3

    > “Segment mana paling profitable?”

Hubungan dengan Bisnis

- CLV sering dibandingkan dengan:

    > Customer Acquisition Cost (CAC)


    Idealnya

    $$
    CLV > CAC
    $$

    Karena:

    > customer menghasilkan lebih banyak uang dibanding biaya mendapatkannya.

    | Faktor                    | Arti                      |
    |---------------------- |---------------------- |

## **Metric**

### **Average Order Value (AOV)**



Definisi:  
> Rata-rata nilai transaksi

Rumus:

$$
AOV =
\frac{
Total\ Revenue
}{
Total\ Orders
}
$$

Intuisi:
> Seberapa besar setiap pembelian

* tinggi → sekali beli sudah besar
* rendah → butuh volume

Contoh:

| Revenue | Orders |
|---- |--- |
| 10 juta | 100    |

Hasil :
$$100.000$$

### **Purchase Frequency**



Definisi:  
> Seberapa sering pelanggan membeli

Rumus:

$$
Purchase\ Frequency =
\frac{
Total\ Orders
}{
Total\ Customer
}
$$

Intuisi:  
> Seberapa sering mereka kembali

* tinggi → loyal
* rendah → sekali beli lalu hilang

### **Customer Lifetime**



Definisi:  
> Berapa lama pelanggan bertahan

Intuisi:  
> Durasi hubungan dengan bisnis

* panjang → sangat berharga
* pendek → cepat churn

Contoh:
* 12 bulan,
* 3 tahun,
* dll.



### **Addition**

Rumus Inti CLV

- Konsepnya:

    $$CLV = \text{Average Order Value} \times \text{Purchase Frequency} \times \text{Customer Lifespan}$$

    Intuisi Besar:
    > CLV = “Seberapa besar” × “Seberapa sering” × “Seberapa lama”

- Artinya:

    pelanggan ideal adalah yang:
    * belanja besar
    * sering beli
    * bertahan lama

Contoh


- Data

    | Metric             | Nilai        |
    |--------------- |--------- |
    | AOV                | Rp100.000    |
    | Purchase Frequency | 2 kali/bulan |
    | Customer Lifetime  | 12 bulan     |

    $ \ $

- Hitung

    $$
    100.000\times2\times12=2.400.000
    $$


- Interpretasi

    Satu customer diperkirakan menghasilkan:

    $$
    \text{Rp. } 2.400.000
    $$


**Hubungan Antar Komponen**


- APV Tinggi, Tapi Frequency Rendah

    **Artinya:**

    * sekali beli besar
    * tapi tidak kembali

    CLV tetap bisa rendah


- Frequency Tinggi, Tapi APV Rendah

    **Artinya:**

    > sering beli kecil-kecil

    Bisa tetap tinggi kalau konsisten


- Lifespan Pendek

    Ini paling berbahaya
    
    > Karena memotong semua potensi nilai


**Dari Angka → Insight Nyata**


- Contoh kamu:

    * APV = 100 ribu
    * Frequency = 2x/bulan
    * Lifespan = 12 bulan

- Hasil:

    CLV = 2.4 juta

- Insight:

    > Satu customer bernilai 2.4 juta selama hidupnya

**Dari Insight → Keputusan**

Ini bagian paling penting.


- Intuisi utama:

    > Selama biaya mendapatkan customer < CLV → bisnis masih untung


- Keputusan:

    Berani keluar biaya marketing hingga < 2.4 juta


    Tapi hati-hati (ini sering salah!):

    > Tidak semua CLV = profit


- Versi lebih realistis:

    CLV harus dibandingkan dengan:

    > cost (produksi, operasional, dll)


- Idealnya:

    > **Profit CLV > Customer Acquisition Cost (CAC)**

**CLV dalam Strategi Bisnis**


- Marketing Strategy

    * Fokus ke channel dengan CLV tinggi
    * Hindari channel dengan CLV rendah


- Retention Strategy

    Meningkatkan CLV bisa lewat:

    * meningkatkan frequency
    * memperpanjang lifespan
    * meningkatkan value per transaksi



- Customer Segmentation

    * High CLV → prioritas tinggi
    * Low CLV → strategi berbeda

## **Historical CLV**

**Konsep**  

- Jenis paling sederhana.

- Menggunakan:

    > data transaksi yang sudah terjadi, tanpa prediksi masa depan.

- Fokus:

    > total revenue/profit historis dari customer.

- Cocok Untuk

    ✅ Junior analyst  
    ✅ Dashboard  
    ✅ Reporting  
    ✅ Analisis dasar customer  

- Rumus  
  
  $$\text{CLV=Average Order Value × Purchase Frequency × Customer Lifespan}$$

**Contoh**  

| Metric              | Nilai        |
| ------------------- | ------------ |
| Average Order Value | Rp 200.000   |
| Purchase Frequency  | 3 kali/tahun |
| Customer Lifespan   | 5 tahun      |

Perhitungan:

$$CLV =200.000 × 3 × 5 = 3.000.000$$

Artinya:

> satu customer rata-rata bernilai Rp 3 juta selama hidupnya.

**Kelebihan**  

| Kelebihan      | Penjelasan                   |
| -------------- | ---------------------------- |
| Mudah dihitung | cocok pemula                 |
| Cepat          | bisa langsung dari dashboard |
| Tidak kompleks | tanpa machine learning       |


**Kekurangan**  

| Kekurangan      | Penjelasan                               |
| --------------- | ---------------------------------------- |
| Tidak prediktif | hanya melihat masa lalu                  |
| Tidak adaptif   | tidak memperhitungkan perubahan perilaku |
| Kurang akurat   | untuk forecasting jangka panjang         |

#### **BG/NBD Model**

Memprediksi:

> berapa kali customer akan membeli lagi

Customer yang:

- sering membeli,
- baru saja membeli,

biasanya:

> lebih mungkin membeli lagi.

**Asumsi 1 - Transaksi Mengikuti Poisson Process**

$$X(t)∼Poisson(λt)$$

Artinya:

> transaksi terjadi secara acak dalam suatu interval waktu, tetapi memiliki rata-rata kecepatan transaksi tertentu.

Keterangan:  

> $\lambda$ sebagai rate/laju transaksi customer.

Formula Poisson

> Probabilitas terjadi k transaksi dalam waktu t:  
> $$P(X(t)=k)=\frac{(\lambda t)^k e^{-\lambda t}}{k!}$$

Keterangan:
| Simbol    | Arti                                 |
| --------- | ------------------------------------ |
| $X(t)$    | jumlah transaksi dalam waktu (t)     |
| $k$       | jumlah transaksi yang ingin dihitung |
| $\lambda$ | rata-rata transaksi per unit waktu   |
| $t$       | interval waktu                       |
| $e$       | konstanta eksponensial               |
| $k!$      | faktorial                            |

Misal
Jika:

$$λ=2$$

artinya:

> customer rata-rata melakukan 2 transaksi per bulan.

Namun transaksi tidak terjadi secara teratur.

> Contoh:
> 
> - kadang 1 kali,
> - kadang 3 kali,
> - kadang tidak sama sekali.
> 
> Tetapi rata-ratanya tetap sekitar 2.

Oleh karena itu, poisson process digunakan karena perilaku transaksi customer:

- tidak selalu teratur,
- muncul secara acak,
- tetapi tetap memiliki pola rata-rata.

Contoh Sederhana

- Data Customer A

    | Customer | Tanggal | Revenue |
    | -------- | ------- | ------- |
    | A        | 1 Jan   | 200.000 |
    | A        | 10 Jan  | 300.000 |
    | A        | 25 Jan  | 500.000 |
    | A        | 15 Feb  | 400.000 |

    Misalkan observasi sampai 30 April

- Hitung Rate Transaksi (λ)

    Jumlah transaksi customer A:  
    > 4

    Lama observasi:

    > 120 hari

    Maka:

    $$λ=\frac{4}{120}	=0.033$$

    berarti:

    > customer A rata-rata melakukan 0.033 transaksi per hari.

    Atau kira-kira:

    $$0.033×30≈1$$

    ≈ 1 transaksi per bulan.

- Masuk ke Poisson Process

    Poisson digunakan untuk memodelkan:

    > berapa kemungkinan jumlah transaksi yang terjadi dalam suatu interval waktu.

    Formula:

    $$X(t)∼Poisson(λt)$$

    Kalau:

    $$λ=0.033$$

    dan kita ingin melihat:

    $$t=30 \text{hari}$$

    maka:

    $$λt=0.033×30=0.99$$

    Artinya:

    > dalam 30 hari, customer diperkirakan melakukan sekitar 1 transaksi.

    Tetapi Poisson tidak mengatakan “pasti 1”.

    > Poisson memodelkan probabilitas:
    > 
    > - 0 transaksi,
    > - 1 transaksi,
    > - 2 transaksi,
    > - dst.

    Kesimpulan
    
    > Poisson memodelkan:  
    > "jumlah transaksi yang mungkin terjadi dalam interval waktu tertentu berdasarkan rate transaksi customer (λ)

Kenapa Penting di Predictive CLV?

Karena model ingin memperkirakan:

* seberapa aktif customer,
* seberapa sering customer akan membeli lagi.

Dan Poisson membantu memodelkan:

> “kecepatan transaksi customer.”

**Asumsi 2 - Customer Bisa Churn Setelah Transaksi**

Probabilitas churn:


$$p$$

Artinya, setelah membeli, customer bisa:

- tetap aktif,
- atau berhenti selamanya.

Faktor Penting BG/NBD

| Faktor    | Arti                             |
| --------- | -------------------------------- |
| Frequency | seberapa sering membeli          |
| Recency   | seberapa baru transaksi terakhir |
| T         | umur customer dalam observasi    |


Intuisi Besar BG/NBD

$$
P(\text{alive})
\propto
\frac{\text{frequency tinggi}}
{\text{recency lama}}
$$


Semakin sering dan semakin baru customer bertransaksi:

> semakin besar kemungkinan customer masih aktif.


Output BG/NBD

| Customer | Predicted Transactions |
| -------- | ---------------------- |
| A        | 10                     |
| B        | 0.5                    |

Interpretasi:

* Customer A diprediksi membeli lagi 10 kali,
* Customer B hampir churn.

### **Gamma-gamma Model**

Memprediksi:

$$
\text{rata-rata nilai transaksi customer}
$$


Asumsi Matematis  
> Monetary Value Mengikuti Gamma Distribution

$$
M \sim Gamma(p,q)
$$

Artinya:

> nilai transaksi customer mengikuti distribusi Gamma karena transaksi:
>
> * selalu bernilai positif,
> * tidak simetris,
> * banyak transaksi kecil,
> * sedikit transaksi sangat besar.

Keterangan:

| Simbol   | Arti                     |
| -------- | ------------------------ |
| $M$      | nilai transaksi customer |
| $\alpha$ | shape parameter          |
| $\beta$ | rate parameter           |

Formula Gamma Distribution

$$f(x;\alpha,\beta)=\frac{\beta^\alpha x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)}
$$

Keterangan:

| Simbol           | Arti                  |
| ---------------- | --------------------- |
| $x$              | nilai transaksi       |
| $\alpha$         | shape/pola distribusi |
| $\beta$          | rate transaksi        |
| $\Gamma(\alpha)$ | fungsi gamma          |

Intuisi Parameter

- Shape ((\alpha))

    $$\alpha$$

    menggambarkan:

    > bentuk distribusi atau banyaknya pola kejadian/transaksi yang dimodelkan.

    Semakin besar:

    $$\alpha$$

    maka distribusi semakin stabil dan tidak terlalu skewed.



- Rate ((\beta))

    $$\beta$$
    
    menggambarkan:
    
    > seberapa cepat transaksi/revenue berubah.
    
    Semakin besar:
    
    $$\beta$$
    
    maka distribusi semakin terkonsentrasi.

Contoh - Data Customer A

| Transaksi |
| --------- |
| 200k      |
| 300k      |
| 500k      |
| 400k      |

Kita ingin:

> mencari distribusi Gamma yang cocok untuk pola transaksi customer ini.


- Step 1 — Hitung Mean (Rata-Rata)

    $$
    \bar{x}= \frac{200+300+500+400}{4} = 350
    $$

    Artinya:

    > rata-rata transaksi customer sekitar 350 ribu.


- Step 2 — Hitung Variance

    Formula variance:

    $$
    s^2=\frac{\sum (x_i-\bar{x})^2}{n}
    $$


- Hitung Selisih Kuadrat

    | Nilai | Selisih dari Mean | Kuadrat |
    | ----- | ----------------- | ------- |
    | 200   | -150              | 22500   |
    | 300   | -50               | 2500    |
    | 500   | 150               | 22500   |
    | 400   | 50                | 2500    |

    Jumlah:

    $$
    22500+2500+22500+2500
    =
    50000
    $$

    Variance:

    $$
    s^2
    =\frac{50000}{4} = 12500
    $$


- Step 3 — Estimasi Parameter Gamma

    Untuk Gamma Distribution:

    $$
    E[X]
    =
    \frac{\alpha}{\beta}
    $$
    dan:

    $$
    Var(X)=
    \frac{\alpha}{\beta^2}
    $$


    Cari (\beta)

    Gunakan:

    $$
    \beta=\frac{\bar{x}}{s^2}
    $$ 

    $$
    \beta= \frac{350}{12500}= 0.028
    $$


    Cari (\alpha)

    Gunakan:

    $$
    \alpha = \bar{x} \times \beta = 350 \times 0.028 = 9.8
    $$


- Distribusi Gamma yang Didapat

    $$
    X
    \sim
    Gamma(9.8,0.028)
    $$


- Apa Maksudnya?

    Model mengatakan:

    > “transaksi customer ini kemungkinan berasal dari distribusi Gamma dengan bentuk seperti ini.”

    Distribusi tersebut:

    * punya mean sekitar 350,
    * punya variasi sesuai data asli,
    * dan memiliki pola right-skewed.


- Cek Expected Value

    $$E[X]= \frac{9.8}{0.028} =350$$


- Bahasa Manusianya

    Distribusi Gamma yang ditemukan:

    * “belajar” dari histori transaksi customer,
    * lalu menghasilkan distribusi probabilistik,
    * yang rata-rata transaksinya mendekati 350 ribu,
    * dengan pola variasi seperti data asli.


- Penting

    Perhitungan di atas adalah:

    > estimasi sederhana menggunakan method of moments.

    Dalam implementasi asli Gamma-Gamma:

    * biasanya menggunakan Maximum Likelihood Estimation (MLE),
    * yang lebih akurat dan lebih kompleks secara matematis.


Intuisi dari Customer A

Karena histori transaksi A:

| Transaksi |
| --------- |
| 200k      |
| 300k      |
| 500k      |
| 400k      |

model melihat bahwa:

* customer ini cenderung belanja ratusan ribu,
* bukan puluhan ribu,
* dan bukan jutaan besar.

Gamma Distribution mengasumsikan bahwa nilai transaksi bersifat:

- positif,
- kontinu,
- dan skewed ke kanan (right-skewed).

Artinya:

- lebih banyak nilai transaksi berada di sisi kiri (nilai kecil-menengah),
- dan hanya sedikit transaksi yang sangat besar di sisi kanan.

Maka model memprediksi future order value sekitar:

$$
350000
$$

atau mendekati itu.



Kesimpulan

> Gamma Distribution memodelkan:
>
> “berapa besar nilai transaksi customer berdasarkan pola histori revenue mereka.”

Jika:

* Poisson/BG-NBD memodelkan:

  > “berapa sering customer membeli”

  maka:

* Gamma-Gamma memodelkan:

  > “kalau customer membeli, biasanya habis berapa.”


Output Gamma-Gamma

| Customer | Predicted Avg Order Value |
| -------- | ------------------------- |
| A        | Rp 500.000                |
| B        | Rp 80.000                 |

$$CLV=(\text{Expected Transactions})\times(\text{Expected Avg Order Value})$$

Kesimpulan

> “Berapa nilai customer di masa depan?”

- memprediksi apakah customer masih aktif,

## **Segment CLV**

**Definisi**  

Segment CLV membandingkan CLV antar kelompok customer.

Tujuan:

> mengetahui segment mana paling bernilai.

**Contoh**  

| Segment | CLV         |
| ------- | ----------- |
| Premium | Rp 5 juta   |
| Free    | Rp 500 ribu |

Interpretasi:  

Customer premium:  
- jauh lebih profitable,
- lebih layak dipertahankan,
- bisa diberi layanan prioritas.

**Segmentasi yang sering dipakai**  

| Segmentasi          | Contoh                  |
| ------------------- | ----------------------- |
| Subscription        | Free vs Premium         |
| Geography           | Kota A vs Kota B        |
| Demografi           | umur/gender             |
| Acquisition Channel | Facebook Ads vs Organic |
| Device              | Android vs iOS          |


**Insight**  

| Kondisi                       | Insight                     |
| ----------------------------- | --------------------------- |
| CLV premium jauh lebih tinggi | fokus upselling             |
| CLV organic lebih tinggi      | organic traffic berkualitas |
| CLV region tertentu rendah    | market kurang cocok         |
| CLV cohort baru meningkat     | strategi berhasil           |

- regression,

| Pertanyaan                                                    | Statistik / Metode           |
