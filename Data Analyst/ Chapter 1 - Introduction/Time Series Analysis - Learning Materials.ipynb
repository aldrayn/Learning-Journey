# **Time Series Analysis**

## **Introduction**

### **Concept**

**Apa itu Time Series**

Kamu punya gelas, tiap detik kamu isi air.

- Detik ke-1 → 100 ml
- Detik ke-2 → 120 ml
- Detik ke-3 → 140 ml

Pertanyaan:  
> Apakah isi gelas di detik ke-3 acak? 

> Enggak   

Kenapa?

> Karena Isi sekarang = hasil dari yang sebelumnya + tambahan baru

Insight penting

> Data sekarang dipengaruhi oleh masa lalu

Ini inti dari time series.

**Kenapa antar waktu saling berhubungan?**

Karena di dunia nyata:  

- Semua punya “memory” (jejak masa lalu)  
    
    Contoh: Penjualan
    - Hari ini laku 100
    - Besok kemungkinan gak tiba-tiba jadi 0 atau 10.000

    Kenapa?

    - pelanggan masih sama
    - tren masih jalan
    - kebiasaan belum berubah drastis

    Jadi:  
    > Hari ini “nyambung” ke kemarin  

    $$ \ $$

- Ada inertia (gak bisa berubah drastis tiba-tiba)

    Artinya:
    - Sifat kecenderungan alami benda untuk mempertahankan keadaannya
    - Nilai sekarang “kebawa” dari nilai sebelumnya, jadi perubahan gak langsung ekstrem


    Contoh: Suhu
    - Hari ini 30°C
    - Besok biasanya 29–31°C  
    - Jarang banget:  
        30°C → langsung 10°C (kecuali kejadian ekstrem)

    Kenapa?  

    Karena: 
    - panas hari ini masih tersisa
    - butuh waktu untuk berubah drastis


    Jadi:

    > “Panas hari ini masih kebawa ke besok”


    Kesimpulan

    > Nilai sebelumnya masih punya pengaruh kuat ke nilai berikutnya

    Makanya:

    - perubahan → pelan / bertahap
    - bukan → loncat ekstrem

    $$ \ $$

- Ada pola kebiasaan (habit / pattern)  
  
    Contoh: Traffic website
    - Senin ramai
    - Minggu sepi

    Ini berulang

    Jadi:  
    > Masa lalu kasih “petunjuk” masa depan  

    $$ \ $$

- Ada efek sebab-akibat yang berlanjut
    
    Contoh: Harga saham

    Kalau:  
    > lagi naik terus (bullish)

    Biasanya:  
    > besok masih cenderung naik (walau gak selalu)

    Kenapa?  
    > sentimen investor masih sama

**Bandingin sama data biasa**

- Data biasa (non time series)

    Contoh:

    - tinggi badan orang
    - warna mobil

    > Urutan gak penting

    Misal:

    [170, 165, 180]

    dibalik:

    [180, 170, 165]

    > gak berubah makna


- Time Series

    Contoh:

    [100, 120, 140]

    dibalik:

    [140, 120, 100]

    Makna berubah total:

    - yang satu naik
    - yang satu turun
  
    > Urutan = segalanya

**intuisi**  

Time series itu:

> Data yang punya urutan waktu  
> DAN  
> Nilai sekarang dipengaruhi nilai sebelumnya

**Analogi**  

Bayangin kamu jalan:

- posisi sekarang = hasil langkah sebelumnya
- kamu gak teleport ke tempat random

posisi sekarang tergantung:

- posisi tadi
- arah jalan
- kecepatan

Mapping ke data:
- posisi = nilai data
- langkah = waktu

itulah time series**

**Contoh - Penjualan harian**

|Hari	|Sales|
|-------|-----|
|Senin	|100|
|Selasa	|110|
|Rabu	|115|

Kenapa naik pelan?

- pelanggan bertambah sedikit
- marketing mulai jalan

bukan random → ada hubungan

**Kapan TIDAK saling berhubungan?**

Kadang ada data yang:

- benar-benar random
- gak ada pola

Contoh:

> hasil lempar koin

ini bukan time series yang “punya struktur”

**Cara mikir sebagai data analyst**

Setiap lihat data waktu, tanya:

- Apakah nilai sekarang dipengaruhi masa lalu?
- Apakah ada pola?
- Apakah perubahan terjadi bertahap?

Kalau iya:
> itu time series

### **Component**

#### **Trend**

**Definisi**  
  - Trend adalah arah jangka panjang dari data.
  - Seperti arah umum perjalanan hidup — apakah data secara keseluruhan sedang naik, turun, atau stabil.
  - Tidak peduli naik-turunnya kecil (noise), trend tetap menggambarkan “garis besar”.

**Analogi**

Bayangin berat badan:

- hari ini naik 0.5 kg
- besok turun 0.3 kg
- lusa naik lagi

> kelihatannya random

Tapi setelah 3 bulan:

> berat badan kamu naik 5 kg

itu trend

**Contoh - Penjualan Bulanan**  

| Bulan | Sales |
| ----- | ----- |
| Jan   | 100   |
| Feb   | 110   |
| Mar   | 105   |
| Apr   | 120   |
| Mei   | 130   |

Cara analyst lihat:
- Feb turun ke Mar → noise
- tapi Jan → Mei naik
- kesimpulan:  
    > bisnis lagi growing (trend naik)  

- catatan:  
    
    Kalau kamu cuma lihat:

    > Feb → Mar turun

    bisa salah ambil keputusan


**Trend ≠ Fluktuasi**

- Trend: perubahan struktur jangka panjang (arah umum)
- Fluktuasi: perubahan jangka pendek (naik-turun kecil)
- Contoh:  
  📈 Harga saham bisa naik-turun tiap hari, tapi selama 5 tahun bisa tetap naik terus — itulah tren jangka panjang.

**Bentuk Umum Trend**  

Trend biasanya direpresentasikan sebagai fungsi waktu
𝑡 :  
$$ \ $$
$$T_t = f(t)$$
$$ \ $$
Artinya:   
> kita mencoba menggambar “garis halus” yang mewakili arah data  

Catatan:  
Tergantung bentuk datanya, $f(t)$ bisa berupa linear, eksponensial, logaritmik, atau polinomial.  

- **Linear Trend**

    $$ \ $$
    $$ T_t = a + b\cdot t$$
    $$ \ $$
    di mana:

    - 𝑎 = intercept (awal garis tren),
    - 𝑏 = slope (laju perubahan per satuan waktu).  
    
    artinya:  
    > naik (atau turun) dengan kecepatan tetap  

    Contoh nyata:
    - bisnis nambah 100 customer tiap bulan
    - gaji naik Rp500 ribu tiap tahun

    growth stabil  

    $$ \ $$

- Exponential Trend   

  $$ \ $$
  $$ T_t = a \cdot b^t$$
  $$ \ $$
  Keterangan:
  - Jika $b>1 = $ naik secara eksponensial
  - Jika $0<b<1=$ turun secara eksponensial

  Artinya:
  > Perubahan proporsional terhadap nilai sebelumnya.  

  Contoh nyata:
    - startup growth
    - viral content
    - investasi
    - pertumbuhan virus
    - populasi
    
    Bayangin:
    - awal naik sedikit
    - lama-lama “meledak”

    Contoh:  

    > Populasi bertambah 5% per tahun → $1000 \cdot (1.05)^t$  

    $$ \ $$

- Logarithmic Trend  

    $$ \ $$
    $$ T_t = a + b \cdot \ln t$$
    $$ \ $$
    Artinya:  
    > Pertumbuhan cepat di awal lalu melambat seiring waktu.   


    Contoh nyata:
    - user baru di aplikasi
    - belajar skill baru  
     
    Bayangin:
    - awal belajar cepat
    - lama-lama makin susah naik  

    $$ \ $$

- Polynomial Trend (Non-linear)  
    $$ \ $$
    $$ T_t = a + b_1t + b_2t^2 + b_3t^3 + \cdots$$
    $$ \ $$
    Artinya:  
    > data bisa naik → lalu turun → atau sebaliknya  

    Contoh nyata:
    - ekonomi (boom → resesi)
    - produk (naik → jenuh → turun) 
    - suhu bisa naik hingga titik tertentu, lalu menurun. 

**Cara pikir saat lihat trend**

Jangan cuma tanya:

> “naik atau turun?”

Tapi:

- seberapa cepat naik?
- stabil atau berubah?
- linear atau eksponensial?

**Ringkasan**  

> Trend adalah arah utama data dalam jangka panjang.  
> Ia menunjukkan ke mana data bergerak jika fluktuasi kecil diabaikan.

Secara intuitif, trend adalah “cerita besar” dari data.

Secara matematis:  

$$T_t = f(t)$$

Bentuk trend bisa:

- Linear → naik konstan
- Exponential → makin cepat
- Logarithmic → cepat lalu melambat
- Polynomial → naik lalu turun

#### **Seasonality**

**Definisi**  

> **Seasonality (musiman)** adalah **pola berulang yang terjadi secara teratur dalam interval waktu tertentu** (misalnya harian, mingguan, bulanan, tahunan).

Pola ini muncul karena **faktor waktu yang berulang secara alami atau sosial**  

misalnya:

* Penjualan naik tiap akhir tahun (Natal)
* Jumlah pengunjung tempat wisata naik tiap libur panjang
* Konsumsi listrik naik tiap siang (karena AC)
* Trafik internet turun tiap malam

**Intuisi**  

- Kalau kamu lihat data:

    > naik → turun → naik → turun  
    > dan itu kejadian lagi… dan lagi… dan lagi…

    - itu bukan kebetulan  
    - itu kebiasaan yang berulang  

- Seasonality itu “ritme”

    Kalau trend = arah perjalanan,
    maka:

    > seasonality = pola langkahnya

**Contoh - Web Traffic**    

| Hari   | Traffic |
| ------ | ------- |
| Senin  | 1000    |
| Selasa | 1100    |
| ...    | ...     |
| Sabtu  | 1500    |
| Minggu | 1600    |

Insight:
- weekend selalu tinggi
- weekday lebih rendah

Dan ini berulang tiap minggu $\to$ Ada seasonality mingguan

**Seasonality ≠ Trend**

- Versi manusia:

    - Trend = arah (naik/turun)
    - Seasonality = pola berulang

- Analogi:

    Bayangin kamu naik tangga:
    
    - kamu naik ke atas → trend
    - tapi langkah kamu kanan–kiri–kanan → seasonality

**Bentuk Matematis**

- Dalam additive:

	$$Y_t​=T_t​+S_t​+E_t$$
		​

	Intuisi:

	> $S_t$ adalah “tambahan” yang tergantung posisi waktu

	Contoh:
	- Senin = -10
	- Jumat = +20
	- Minggu = +40

	nilai tetap tiap minggu  

	$$ \ $$

- Dalam multiplicative:
	
	$$Y_t	​=T_t	​×S_t	​×E_t$$
	​

	Intuisi:

	> $S_t$ adalah “faktor pengali”

	Contoh:
	- weekday = 0.8
	- weekend = 1.5

	berbentuk rasio, bukan selisih

**Bentuk Seasonality (Shape)**

Ciri khas:
- berulang
- punya periode tetap  

Contoh periode:
- harian (jam kerja vs malam)
- mingguan (weekday vs weekend)
- bulanan (gajian)
- tahunan (Ramadhan, liburan)

**Cara Data Analyst Mengambil Seasonality**

Intuisi:

> “Kelompokkan data berdasarkan posisi dalam periode”

Contoh (mingguan):

Ambil semua:

- Senin → rata-rata
- Selasa → rata-rata

hasilnya:

> pola seasonality

Secara matematis (additive):  

$$S_t=\text{rata-rata nilai pada posisi ke-t}$$

Contoh  
| Hari   | Sales |
| ------ | ----- |
| Senin  | 100   |
| Selasa | 110   |
| ...    | ...   |
| Minggu | 150   |
| Senin  | 105   |
| Selasa | 115   |  

Step:
- semua Senin → (100 + 105)/2 = 102.5
- semua Selasa → (110 + 115)/2 = 112.5

itulah pola seasonality


**Insight**
- Kasus:

    > “Penjualan tinggi di weekend”

- Analyst harus mikir:

    ini karena trend? atau seasonality?

- Jawaban:

    > kalau terjadi tiap minggu → seasonality

- Decision:
    - tambah stok weekend
    - push promo weekday




**Kesalahan Umum**
- Salah:

    > “Penjualan naik → bisnis growing”

- Benar:

    cek dulu:

    - trend naik?
    - atau cuma seasonality?

**Ringkasan**  

Seasonality adalah pola berulang dalam periode waktu tertentu.
Ia merepresentasikan “kebiasaan” data yang muncul secara konsisten.

Secara intuitif, seasonality adalah ritme data.

Dalam model:

- Additive → sebagai selisih tetap
- Multiplicative → sebagai faktor skala

Tambahan:  
> “Seasonality biasanya terlihat jelas setelah trend dihilangkan”

#### **Noise**

**Definisi**  

> Noise adalah bagian dari data yang tidak bisa dijelaskan oleh pola apa pun  

Ketika kita sudah:

- ambil trend (arah besar)
- ambil seasonality (pola berulang)

tapi masih ada sisa kecil yang “acak”

> itulah noise

**Intuisi**  

> Noise itu “ketidaksempurnaan dunia nyata”

Data real tidak pernah:

- rapi
- sempurna
- 100% bisa dijelaskan

**Analogi**  

Bayangin kamu prediksi penjualan:

- trend bilang harusnya 120
- seasonality bilang +30 (weekend)

harusnya 150

Tapi real:

- jadi 147
- atau 155

selisih itu = noise

**Contoh - Penjualan Harian**  

| Hari   | Actual | Prediksi (Trend+Seasonality) |
| ------ | ------ | ---------------------------- |
| Senin  | 100    | 102                          |
| Selasa | 110    | 108                          |
| Rabu   | 105    | 107                          |

Noise:
- Senin → -2
- Selasa → +2
- Rabu → -2
- kecil, random

Insight:

- model sudah cukup bagus
- sisanya memang tidak bisa diprediksi


**Noise ≠ Error (dalam arti salah)**
- Versi manusia:

    > Noise bukan “kesalahan”, tapi “hal yang memang tidak bisa diprediksi”

- Contoh:
    - tiba-tiba hujan
    - server down
    - event mendadak

    tidak ada di pola

**Bentuk Matematis**  

- Additive:
    $$Y_t	​=T_t	​+S_t	​+E_t$$

    Noise:
    $$E_t	​=Y_t	​−T_t	​−S_t$$


    sisa setelah semua pola diambil  

    $$ \ $$

- Multiplicative:
    
    $$Y_t	​=T_t	​×S_t	​×E_t$$
	​

    Noise:
    $$E_t= \frac{​Y_t}{T_t​×S_t}$$
	​

	​


**Bentuk (Shape)**
- Ciri-ciri noise:
    - acak
    - kecil (relatif)
    - tidak berulang
    - tidak punya pola
- Visual di kepala:
    - zig-zag kecil
    - “getaran” di sekitar garis

**Cara Data Analyst Melihat Noise**
- Intuisi:

    > “Kalau semua pola sudah diambil, sisanya apa?”

- Cara pikir:
    - ambil trend
    - ambil seasonality
    - lihat sisa

    itu noise

**Insight Penting**
- Tujuan analyst:

    bukan menghilangkan noise sepenuhnya

    Tapi:

    memastikan noise itu benar-benar “acak”

- Kenapa?

    Kalau noise masih punya pola:  
    > berarti ada sesuatu yang belum kamu tangkap  

- Contoh:

    Kalau “noise” ternyata:

    selalu naik tiap Jumat

    - itu bukan noise
    - itu seasonality yang belum ketangkap

Kesalahan Umum
- Salah:

    > “Semua fluktuasi kecil = noise”

- Benar:

    cek dulu:

    - ada pola? → bukan noise
    - random? → noise

**Peran Noise dalam Decision**
- Kasus:

    > “Penjualan turun sedikit hari ini”

- Analyst:
    - cek trend → stabil
    - cek seasonality → normal

    berarti:

    > ini noise

- Decision:
    - jangan panik
    - jangan ubah strategi

**Ringkasan**  

- Noise (residual) adalah bagian data yang tidak bisa dijelaskan oleh trend dan seasonality.  
- Ia bersifat acak dan tidak memiliki pola yang jelas.
- Secara intuitif, noise adalah “sisa” dari data setelah semua pola diambil.

#### **Cyclic**

**Definisi**  

> **Cyclic (Siklus)** adalah **fluktuasi jangka panjang** pada data time series yang **tidak memiliki periode tetap**, dan sering kali berkaitan dengan **perubahan kondisi ekonomi, sosial, atau teknologi**.

Dengan kata lain, **pola ini berulang, tapi tidak teratur.**

* Bisa berlangsung **lebih dari satu tahun**, kadang 3, 5, bahkan 10 tahun.
* Tidak selalu sama panjang atau sama tinggi.
* Biasanya **berhubungan dengan fase-fase ekonomi** seperti ekspansi dan resesi.

**Perbedaan Cyclic vs Seasonality**

| Aspek          | **Seasonality (Sₜ)**                                  | **Cyclic (Cₜ)**                            |
| -------------- | ---------------------------------------------------- | ------------------------------------------ |
| **Pola**       | Berulang                                             | Naik-turun                                 |
| **Periode**    | Tetap (misal setiap 12 bulan)                        | Tidak tetap                                |
| **Penyebab**   | Faktor alam atau sosial berulang (musim, hari libur) | Faktor ekonomi, politik, teknologi, dll    |
| **Durasi**     | < 1 tahun (biasanya)                                 | > 1 tahun                                  |
| **Deteksi**    | Mudah terlihat di grafik                             | Butuh analisis jangka panjang              |
| **Model umum** | Sinusoidal periodik                                  | Gelombang tidak beraturan tapi tetap halus |
| **Contoh**     | Penjualan naik tiap Desember                         | Siklus ekonomi naik–turun tiap 4–5 tahun   |

**Intuisi**  

Kalau seasonality itu:

> “terjadi lagi di waktu yang sama”

Maka cyclic itu:

> “terjadi lagi… tapi gak tahu kapan”

Jadi bisa dibilang:  
> Cyclic = “gelombang panjang yang tidak terjadwal”  

dan juga :

> Cyclic itu “fase”

Bukan rutinitas, tapi:

- fase naik
- fase turun
- lalu naik lagi

**Analogi**

Ekonomi:
- lagi booming → semua naik
- lalu krisis → turun
- lalu recovery → naik lagi

tapi:

> gak selalu tiap 5 tahun, bisa beda-beda  

Analogi lain:
- Seasonality = jam kerja harian
- Cyclic = naik-turun karier

**Contoh - Penjualan Properti**  

- 2018–2019 → naik
- 2020 → turun (krisis)
- 2021–2023 → naik lagi

Insight:
- bukan seasonality
- karena tidak berulang tiap periode tetap

ini cyclic

**Bentuk Matematis**


- Additive  
    Dalam model lengkap:

    $$Y_t	​=T_t	​+S_t	​+C_t	​+E_t$$
    	​

    Intuisi:

    > $C_t$	​ adalah “gelombang besar tambahan” di luar trend

- Multiplicative:
    
    $$Y_t	​=T_t	​×S_t	​×C_t	​×E_t$$
	​


**Bentuk (Shape)**
- Ciri-ciri cyclic:
    - gelombang panjang
    - naik → turun → naik
    - tidak konsisten periodenya
- Visual di kepala:
    - seperti ombak besar
    - tapi jaraknya tidak sama

**Cara Data Analyst Mengidentifikasi**

- Intuisi:

    > “Apakah ada pola naik-turun jangka panjang yang tidak terjadwal?”

- Cara pikir:
    - hilangkan trend
    - hilangkan seasonality
    - lihat sisa besar

    kalau masih ada pola panjang → cyclic

**Insight Penting**
- Cyclic biasanya muncul di:
    - ekonomi
    - industri besar
    - data jangka panjang
- Jarang terlihat di:
    - data pendek
    - data harian singkat

- Kesalahan Umum
    - Salah:

        > “Semua naik-turun = seasonality”

    - Benar:
        - kalau terjadwal → seasonality
        - kalau tidak → cyclic

**Peran Cyclic dalam Decision**
- Kasus:

    > “Penjualan turun dalam 1 tahun terakhir”

- Analyst harus cek:
    > ini trend turun?
    > atau bagian dari cycle?

- Decision:

    Kalau cyclic:

    ❌ jangan panik  
    ✅ siapin strategi bertahan

**Ringkasan**  

- Cyclic adalah pola naik-turun jangka panjang yang tidak memiliki periode tetap.  
- Ia merepresentasikan fase atau gelombang besar dalam data.
- Secara intuitif, cyclic adalah “naik-turun yang tidak terjadwal”.

### **Math Concept**

#### **Additive Form**

**Konsep**  

$$Y_t = T_t + S_t + E_t$$

**Artinya:**

> Data yang kita lihat $(Y_t)$ adalah hasil penjumlahan dari beberapa efek:

* $T_t$: arah besar (trend)
* $S_t$: pola berulang (seasonality)
* $E_t$: sisa / gangguan (noise)

Intuisi:  
> Setiap waktu itu ada “penyebab-penyebab kecil” yang dijumlahkan jadi satu angka

Analogi kuat:  
- Bayangin nilai penjualan hari ini:

  > = “nilai normal” + “efek hari ini” + “gangguan kecil”

  Misalnya:

  * normal = 100
  * karena weekend = +30
  * random = -5

- hasil:  
  > 125


**Intuisi**  

- Semua efek itu **konstan dalam bentuk selisih**

    Contoh:

    * weekend selalu +30 (bukan ×1.3)
    * promo kecil selalu +10

    tidak tergantung besar data

- Ini beda dengan multiplicative:

    * additive → tambah tetap
    * multiplicative → kali (ikut skala)

**Contoh - Penjualan E-Commerce**  

Kamu analyst di e-commerce

Pertanyaan:

> “Kenapa penjualan berubah tiap hari?”

Terdapat data:

| Hari   | Sales |
| ------ | ----- |
| Senin  | 100   |
| Selasa | 105   |
| Rabu   | 98    |
| Kamis  | 102   |
| Jumat  | 110   |
| Sabtu  | 140   |
| Minggu | 150   |


Selanjutnya muncul hipotesis:

> “Ini mungkin = trend + efek hari + noise”



- Step 1: Tentukan baseline (trend kasar)

    Misal:

    * rata-rata weekday ≈ 100
    * anggap ini trend awal

    $T_t \approx 100$

    $$ \ $$
    

- Step 2: Cari efek seasonality

    Hitung selisih dari baseline:

    | Hari   | Sales | Efek |
    | ------ | ----- | ---- |
    | Senin  | 100   | 0    |
    | Selasa | 105   | +5   |
    | Rabu   | 98    | -2   |
    | Kamis  | 102   | +2   |
    | Jumat  | 110   | +10  |
    | Sabtu  | 140   | +40  |
    | Minggu | 150   | +50  |

    ini adalah $S_t$

    $$ \ $$
    

- Step 3: Sisa = noise

    $$
    E_t = Y_t - T_t - S_t
    $$

    Biasanya kecil:

    * +3, -2, dll


**Insight dari Additive Model**

- Weekend effect jelas

    * Sabtu +40
    * Minggu +50

    Decision:

    * tambah stok
    * push promo di weekday



- Trend stabil

    * baseline sekitar 100

    Decision:

    * bisnis belum growing signifikan



- Noise kecil

    * fluktuasi kecil

    Decision:

    * jangan overreact ke perubahan kecil


**Kegiatan Data Analyst**


- Exploratory Analysis

    Analyst akan:

    * plot data
    * lihat apakah pola terlihat “penjumlahan”

    Ciri additive:

    * jarak naik turun relatif sama


- Estimasi Komponen

    Biasanya:

    1. ambil trend → moving average
    2. hitung seasonality → rata-rata per periode
    3. hitung noise → sisa


- Forecasting sederhana

    Dengan additive:

    $$
    \hat{Y}_{t+1} = T_{t+1} + S_{t+1}
    $$


     Contoh:

    * trend besok = 105
    * besok Sabtu → +40

    prediksi:  
    145
    


- Decision Making

    - 📈 Kasus 1: “Penjualan naik”

        Cek:

        * trend naik? → growth
        * atau cuma seasonality? → normal

    - 📉 Kasus 2: “Penjualan turun”

        Cek:

        * noise? → abaikan
        * trend turun? → masalah

**Kapan Additive Cocok?**


- Cocok kalau:

    * variasi relatif konstan
    * selisih stabil
    * data tidak “meledak” (growth besar)


- Tidak cocok kalau:

  > semakin lama fluktuasi makin besar $\to$ itu multiplicative

**Ringkasan**

> Additive = “semua efek ditambah”

* setiap komponen punya kontribusi tetap
* mudah diinterpretasi
* sangat cocok untuk analisis bisnis sederhana

#### **Multiplicative Form**

**Konsep**  

$$Y_t = T_t \times S_t \times E_t$$

Artinya:
> Nilai data $(Y_t)$ adalah hasil **perkalian** dari:

* $T_t$: trend
* $S_t$: seasonality (dalam bentuk faktor)
* $E_t$: noise

Intuisi:

> Setiap komponen itu **mengubah skala** data, bukan menambah


Analogi: 

> Bayangin penjualan
> = “nilai dasar” × “efek hari” × “gangguan kecil”

Misalnya:

* trend = 100
* weekend = 1.5 (50% lebih ramai)
* noise = 0.9

Hasil:
$$
100 \times 1.5 \times 0.9 = 135
$$

**Intuisi**  

> Efeknya **ikut membesar atau mengecil sesuai skala**

Misal, awal bisnis:

* trend = 100
* weekend = 1.5 → jadi 150 (+50)

Beberapa bulan kemudian:

* trend = 1000
* weekend = 1.5 → jadi 1500 (+500)

Selisih makin besar, karena:

> efeknya dikali, bukan ditambah

Bandingin langsung:

| Model          | Weekend Effect    |
| -------------- | ----------------- |
| Additive       | +50 terus         |
| Multiplicative | 1.5× (ikut besar) |


**Contoh**  

- Kasus:

    Kamu analyst di e-commerce yang lagi growth

    Data:

    | Hari   | Sales |
    | ------ | ----- |
    | Senin  | 100   |
    | Selasa | 110   |
    | ...    | ...   |
    | Sabtu  | 150   |
    | Minggu | 160   |

    Beberapa bulan kemudian:

    | Hari   | Sales |
    | ------ | ----- |
    | Senin  | 500   |
    | Selasa | 550   |
    | ...    | ...   |
    | Sabtu  | 750   |
    | Minggu | 800   |


    $$ \ $$

- Observasi penting

    * weekend dulu naik ~50
    * sekarang naik ~300

    efeknya **ikut membesar**  

    $$ \ $$

- Insight:

    > Ini bukan additive → ini multiplicative


**Cara Analyst Memecah (Multiplicative Thinking)**



- Step 1: Ambil trend

    Misal:

    * minggu awal ≈ 100
    * minggu akhir ≈ 500

    $$ \ $$


- Step 2: Cari seasonality (dalam bentuk rasio)

    | Hari   | Sales | Rasio |
    | ------ | ----- | ----- |
    | Senin  | 100   | 1.0   |
    | Selasa | 110   | 1.1   |
    | Sabtu  | 150   | 1.5   |
    | Minggu | 160   | 1.6   |

     ini $S_t$

    $$ \ $$

- Step 3: Noise

    $$
    E_t = \frac{Y_t}{T_t \times S_t}
    $$

**Interpretasi**

Insight utama:

- Weekend effect = proporsi

    * Sabtu = 1.5×
    * Minggu = 1.6×

    bukan +50, tapi **+50%**



- Growth bisnis jelas

    * trend naik dari 100 → 500

    Decision:

    * scale up bisnis
    * tambah kapasitas



- Noise relatif kecil

    perubahan kecil = normal


Kegiatan Data Analyst (Praktis)



- Identifikasi model

    Analyst akan cek:

    apakah:

    * selisih tetap → additive
    * rasio tetap → multiplicative



- Transformasi (trik penting)

    Karena multiplicative ribet, biasanya:

    diubah ke additive dengan log:

    $$
    \log(Y_t) = \log(T_t) + \log(S_t) + \log(E_t)
    $$


    Intuisi:

    > perkalian → jadi penjumlahan



- Forecasting

    $$
    \hat{Y}_{t+1} = T_{t+1} \times S_{t+1}
    $$



    Contoh:

    * trend besok = 600
    * besok Minggu = 1.6

    prediksi:
    $$
    960
    $$

 **Kapan Multiplicative Cocok?**



- Cocok kalau:

    * data growth
    * fluktuasi makin besar
    * efek berbentuk persentase



- Tidak cocok kalau:

  perubahan stabil
      $\to$ itu additive

**Ringkasan**

> Multiplicative = “semua efek mengubah skala”

* efek berupa rasio (1.2, 0.8, dll)
* makin besar data → makin besar efek
* cocok untuk bisnis yang berkembang



Perbandingan

|             | Additive | Multiplicative |
| ----------- | -------- | -------------- |
| Bentuk      | +        | ×              |
| Efek        | tetap    | ikut skala     |
| Cocok       | stabil   | growth         |
| Seasonality | selisih  | rasio          |


#### **Autoregressive (AR)**

**Konsep**  

$$Y_t = \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \epsilon_t$$


Artinya:

> Nilai sekarang $(Y_t)$ dipengaruhi oleh nilai-nilai sebelumnya

* $Y_{t-1}$ = kemarin
* $Y_{t-2}$ = kemarin lusa
* $\phi$ = seberapa kuat pengaruhnya
* $\epsilon_t$ = noise


**Intuisi - Data punya “memory”**  

Bayangin:

- Suhu

    > hari ini panas → besok kemungkinan masih panas

    Kenapa?  
    > karena kondisi sebelumnya masih “kebawa”



- Analogi kuat

    Bayangin kamu dorong bola:
    
    * sekarang bola bergerak
    * sebentar lagi masih bergerak
    
    posisi sekarang tergantung posisi sebelumnya

Bentuk Sederhana $(AR(1))$

$$Y_t = \phi Y_{t-1} + \epsilon_t$$

Intuisi:

> “Hari ini = sebagian dari kemarin + gangguan kecil”

Contoh:

* $\phi = 0.8$
* kemarin = 100

hari ini:
$$
Y_t = 0.8 \times 100 + \epsilon_t \approx 80 + noise
$$

Interpretasi:

* kalau kemarin tinggi → hari ini cenderung tinggi
* kalau kemarin rendah → hari ini ikut rendah

**Bentuk Lebih Umum (AR(p))***

$$Y_t = \sum_{i=1}^{p} \phi_i Y_{t-i} + \epsilon_t$$

Intuisi:

> Tidak cuma kemarin, tapi beberapa hari sebelumnya juga berpengaruh

Contoh:

$$
Y_t = 0.5Y_{t-1} + 0.3Y_{t-2} + \epsilon_t
$$

artinya:

* kemarin lebih penting
* tapi 2 hari lalu juga masih berpengaruhB


**Contoh**  

- Kasus:

    > “Kita mau prediksi penjualan besok”

    $$ \ $$


- Data:

    | Hari   | Sales |
    | ------ | ----- |
    | Senin  | 100   |
    | Selasa | 110   |
    | Rabu   | 105   |
    | Kamis  | 108   |


    $$ \ $$

- Analyst mikir:

    > “Penjualan hari ini kemungkinan dipengaruhi hari sebelumnya”

    $$ \ $$

- Model:

    $$
    Y_t = 0.7Y_{t-1} + \epsilon_t
    $$

    $$ \ $$

- Prediksi:

    Kalau hari ini 108:

    $$
    Y_{t+1} \approx 0.7 \times 108 = 75.6 + noise
    $$



    (di real case biasanya ditambah intercept biar gak turun terus)


**Insight Penting**



- AR itu bukan:

    * bukan trend
    * bukan seasonality



- AR itu:

    > cara memodelkan “ketergantungan waktu”


**Kegiatan Data Analyst**



-  Cek apakah data punya memory

    Biasanya:

    > lihat apakah nilai sekarang mirip dengan sebelumnya

    kalau iya → AR cocok



-  Gunakan untuk forecasting

    $$
    \hat{Y}_{t+1} = \phi Y_t
    $$



-  Insight yang bisa didapat



     Kasus:

    > “Apakah data stabil?”

    kalau $\phi$ besar (mendekati 1):

    * data sangat tergantung masa lalu
    * perubahan lambat



    Kalau $\phi$ kecil:

    * data cepat berubah
    * kurang stabil

**Tambahan**  

> AR = “masa lalu mempengaruhi masa sekarang”


* ada efek inertia
* ada memory
* perubahan tidak random sepenuhnya

**Hubungan dengan yang sebelumnya**



- Additive / Multiplicative:

    fokus ke “struktur data”

- AR:

    fokus ke “hubungan antar waktu”





Ringkasan

* AR = model berbasis masa lalu
* pakai nilai sebelumnya untuk prediksi
* cocok untuk data yang punya “memory”

#### **Moving Average**

**Konsep**  

$$\hat{Y}_t = \frac{1}{k} \sum_{i=0}^{k-1} Y_{t-i}$$



Artinya:

> Nilai sekarang = rata-rata dari beberapa nilai sebelumnya

* $k$ = jumlah window (misal 7 hari)
* $\hat{Y}_t$ = nilai yang sudah “dihaluskan”  

Kita “meredam noise”

- Bayangin data kamu:

    > 100, 120, 90, 130, 110

- naik turun tajam → susah lihat pola  

Kayak kamu:

> “gak lihat 1 hari, tapi lihat rata-rata beberapa hari”

biar lebih stabil


Intinya:

> Moving average = bikin data lebih “tenang”


**Contoh**  

- Kasus:

    Penjualan harian:

    | Hari | Sales |
    | ---- | ----- |
    | 1    | 100   |
    | 2    | 120   |
    | 3    | 90    |
    | 4    | 130   |
    | 5    | 110   |


$$ \ $$

- MA 3 hari:

    - Hari ke-3:
        $$
        (100 + 120 + 90)/3 = 103.3
        $$

    - Hari ke-4:
        $$
        (120 + 90 + 130)/3 = 113.3
        $$



- hasilnya:

    * lebih halus
    * gak terlalu loncat


**Bentuk (Shape)**

- Sebelum:

    > zig-zag (noisy)

- Sesudah MA:

    > lebih smooth

Kegiatan Data Analyst (Praktis Banget)



- Melihat trend

    Masalah:

    > data terlalu noisy

    Solusi:  
    > pakai moving average

    Contoh:  
    
    - Manager tanya:  
        > “Penjualan kita naik gak sih?”


    - Kalau lihat raw data:

        > naik turun → bingung

    - Kalau pakai MA:

        > keliatan arah naik

    $$ \ $$

- Bandingkan periode

    - Misal:
    
        > MA 7 hari vs MA 30 hari
    
    - bisa lihat:
    
        > trend jangka pendek vs panjang

    $$ \ $$

- Forecast sederhana

    $$
    \hat{Y}_{t+1} = \text{rata-rata terakhir}
    $$



    Contoh:

    3 hari terakhir:

    > 100, 120, 110

    prediksi:
    $$
    110
    $$







**Trade-off**

- Window kecil (k kecil)

    * respons cepat
    * tapi masih noisy



- Window besar (k besar)

    * sangat smooth
    * tapi lambat respon


    ini trade-off utama

**Hubungan dengan konsep sebelumnya**



- Additive:

    > MA bantu ambil **trend**

- AR:

    > AR pakai nilai sebelumnya

- MA (smoothing):

    > MA ambil **rata-rata masa lalu**

**Moving Average sebagai Model (MA Model)**

- Definisi:  
    MA (Moving Average) sebagai model adalah pendekatan dalam time series yang menyatakan bahwa:

    > Nilai sekarang dipengaruhi oleh shock (error) saat ini dan masa lalu  

    $$ \ $$

- Bentuk lengkap:  

    $$Y_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1} + \cdots + \theta_q \epsilon_{t-q}$$


    Intuisi:
    - $μ$ → kondisi normal / baseline (misalnya rata-rata penjualan)
    - $ϵ_t$	​ → shock hari ini (kejadian tak terduga)
    - $ϵ_{t−1}	​,ϵ_{t−2}$	​,… → efek shock masa lalu
    - $θ$ → seberapa kuat efek shock itu bertahan

    Jadi:

    > Nilai hari ini = kondisi normal + efek kejadian-kejadian (shock)  

    Contoh:
    - Normal penjualan = 100
    - Kemarin ada promo besar → shock +20
    - Hari ini walaupun promo sudah selesai, efeknya masih terasa

    Artinya:
    
    > Data tidak langsung balik ke normal, tapi “pelan-pelan kembali”  

    $$ \ $$

- Kenapa baseline (μ) sering “hilang”?  
    
    Dalam banyak buku / materi, MA sering ditulis:  

    $Y_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \cdots + \theta_q \epsilon_{t-q}$   

    Karena biasanya data sudah diproses dulu:

    > Data di-center (dibuat mean = 0)

    $ Y_t^{'} = Y_t - \mu $  

    Jadi baseline sudah “dikurangi”

    Intuisi:
    
    - Daripada bilang:

        > “penjualan 110 itu karena 100 + shock”

    - Lebih simpel jadi:

        > “+10 itu karena shock”  
    
    Dampaknya
    - Model jadi lebih sederhana
    - Fokus hanya ke pergerakan (fluktuasi), bukan level

    $$ \ $$

- Insight penting

    MA model itu sebenarnya:
    > baseline + efek shock

    Tapi dalam praktik:

    > baseline disembunyikan karena data sudah dinormalisasi

    $$ \ $$

- Perbedaan 2 jenis MA

    |         | MA (Smoothing) | MA (Model)      |
    | ------- | -------------- | --------------- |
    | Fungsi  | haluskan data  | model statistik |
    | Dipakai | analyst        | modeling        |
    | Intuisi | rata-rata      | efek error      |

**Ringkasan**  

> Moving average = cara paling sederhana untuk “melihat pola asli”



* meredam noise
* menonjolkan trend
* sangat praktis di dunia kerja

#### **ARMA**

**Konsep**  

$Y_t = \underbrace{\phi_1 Y_{t-1} + \cdots + \phi_p Y_{t-p}}_{\text{AR}} + \underbrace{\epsilon_t + \theta_1 \epsilon_{t-1} + \cdots + \theta_q \epsilon_{t-q}}_{\text{MA}}$

Nilai sekarang dipengaruhi oleh:
- Nilai masa lalu (trend/ingatan)
- Error masa lalu (shock)

Bayangin:

Harga hari ini dipengaruhi oleh:

- harga kemarin (kebiasaan/trend)
- dan kejutan-kejutan sebelumnya  



**Analogi sederhana**

Bayangin penjualan hari ini dipengaruhi oleh:

- **kemarin laku banyak → hari ini ikut tinggi** (AR)
- **kemarin ada promo besar → efeknya masih terasa** (MA)

dua-duanya digabung

**Bentuk Sederhana: ARMA(1,1)**

$$Y_t = \phi Y_{t-1} + \epsilon_t + \theta \epsilon_{t-1}$$

Intuisi:

> Hari ini = pengaruh kemarin + efek shock kemarin + noise sekarang

**Contoh Konkret**


- Kasus:

    Penjualan harian


- Situasi:

    * kemarin tinggi → hari ini cenderung tinggi
    * kemarin ada promo → hari ini masih terasa


- Model:

    $$
    Y_t = 0.7Y_{t-1} + \epsilon_t + 0.5\epsilon_{t-1}
    $$


    Interpretasi:
    
    * **0.7** → pengaruh masa lalu kuat
    * **0.5** → efek shock masih ada

Kegiatan Data Analyst (Real Scenario)


- Kapan pakai ARMA?

    Saat kamu lihat data:
    
    * ada pola “nyambung antar waktu” → AR
    * ada efek kejadian tertentu → MA
    
    maka:
    
    > AR saja kurang
    > MA saja kurang
    > → pakai ARMA

    $$ \ $$

- Contoh pertanyaan bisnis


    > “Kenapa penjualan masih tinggi walaupun promo sudah selesai?”


    Jawaban dengan ARMA:

    karena:

    * ada efek inertia (AR)
    * ada efek sisa promo (MA)

    Decision:

    * jangan langsung turunkan strategi
    * efek masih akan “decay” perlahan

    $ \ $   

- Forecasting


    Prediksi:
    
    $$
    \hat{Y}_{t+1} = \phi Y_t + \theta \epsilon_t
    $$
    
    
    Contoh:
    
    * hari ini = 100
    * error hari ini = +10
    
    $$
    \hat{Y}_{t+1} = 0.7(100) + 0.5(10) = 70 + 5 = 75
    $$
    
    
    (di praktik biasanya ada intercept biar realistis)


**Insight Penting**

ARMA hanya cocok untuk:

> data yang sudah “stabil” (stationary)


Intuisi:

* tidak ada trend besar
* tidak ada growth terus-menerus

kalau ada → harus diubah dulu (nanti di ARIMA)

**Hubungan dengan konsep sebelumnya**


| Model    | Fokus             |
| -------- | ----------------- |
| Additive | struktur komponen |
| AR       | nilai masa lalu   |
| MA       | error masa lalu   |
| **ARMA** | gabungan keduanya |

**Cara Mikir Sederhana**

> ARMA = “masa lalu + efek kejadian”


* AR → inertia
* MA → shock
* ARMA → real-world behavior


**Ringkasan**

* ARMA = kombinasi AR dan MA
* cocok untuk data yang punya:

  * dependensi waktu
  * efek shock
* dipakai untuk forecasting klasik

#### **ARIMA**

**Latar Belakang Masalah**  

> ARMA itu cuma bisa dipakai kalau data stationary (stabil)

Tapi di dunia nyata:

* penjualan naik terus
* user growth meningkat
* harga berubah trend

ini **tidak stationary**

Solusi ARIMA:

> Ubah dulu datanya → baru pakai ARMA

**Konsep**  

$$ARIMA(p,d,q)$$

- p → bagian AR (pakai nilai masa lalu)
- d → differencing (berapa kali kita ubah data)
- q → bagian MA (pakai error masa lalu)

**Intuisi**  

> ARIMA = “ubah dulu, baru modelkan”


Bayangin kita lihat:

> total uang di tabungan (naik terus) $\to$ susah dianalisis

Tapi kalau kita lihat:

> “berapa kenaikan per hari” $\to$ jadi lebih stabil

Itu yang dilakukan ARIMA:

> ambil **perubahan (difference)**, bukan nilai asli


**Bentuk Matematis**

$$(1 - B)^d Y_t = ARMA(p,q)$$


 Penjelasan:

* (B) = lag operator (geser waktu)
* (d) = jumlah differencing


Contoh differencing:

1st difference:

$$Y'_t = Y_t - Y_{t-1}$$


Intuisi:

> Kita lihat “perubahan”, bukan nilai absolut

Contoh Konkret


- Kasus:

    Penjualan:

    | Hari | Sales |
    | ---- | ----- |
    | 1    | 100   |
    | 2    | 110   |
    | 3    | 120   |
    | 4    | 130   |

    $$ \ $$

     Masalah:

    > naik terus → tidak stationary

    $$ \ $$


- Step 1: Differencing

    | Hari | Difference |
    | ---- | ---------- |
    | 2    | +10        |
    | 3    | +10        |
    | 4    | +10        |


    sekarang:

    * stabil
    * bisa dimodelkan

    $$ \ $$
    
- Step 2: Pakai ARMA

    Misal:

    * perubahan hari ini tergantung kemarin


     Jadi:

    > ARIMA = ARMA pada data yang sudah di-difference

**Kegiatan Data Analyst (Real Scenario)**


- Kasus bisnis

    > “Penjualan kita naik terus, bisa diprediksi gak?”


     - Analyst mikir:- data ada trend → gak bisa langsung ARMA
     - lakukan differencing
     - modelkan dengan AR + MA


    Decision:

    * prediksi growth ke depan
    * tentukan stok
    * planning bisnis


- Kasus lain

    > “Apakah kenaikan ini stabil?”
    
    
    lihat hasil differencing:
    
    * kalau stabil → growth konsisten
    * kalau tidak → fluktuatif

**Insight Penting**


> d (differencing) itu kunci

* d = 0 → ARMA
* d = 1 → beda sekali
* d = 2 → beda dua kali


Intuisi:

> kita “menghilangkan trend”

**Hubungan dengan model sebelumnya**


| Model     | Fungsi          |
| --------- | --------------- |
| AR        | pakai masa lalu |
| MA        | pakai error     |
| ARMA      | gabungan        |
| **ARIMA** | + handle trend  |

**Intuisi**

> ARIMA = “hilangkan trend → modelkan hubungan waktu”


Langkah mental:

- data naik terus → problem
- ubah jadi perubahan → stabil
- pakai AR + MA

**Ringkasan**

* ARIMA = model paling umum
* bisa handle data real (yang gak stabil)
* inti utamanya = **differencing**


## **Basic Forecasting**

### **Naive Forecasting**

**Definisi**

> **Prediksi nilai berikutnya = nilai terakhir yang diketahui**


Bentuk matematis

$$\hat{Y}_{t+1} = Y_t$$


Artinya:

> “Besok sama dengan hari ini”

**Intuisi**  

> Dunia itu punya inertia

- Kalau hari ini:

  > penjualan 100

  tebakan paling masuk akal:

  > besok ≈ 100

- Analogi lain

  > Suhu hari ini 30°C, besok kemungkinan sekitar 30°C

- Intinya:

  > Kalau gak ada info lain, “pakai yang terakhir”

**Contoh**


Data:

| Hari   | Sales |
| ------ | ----- |
| Senin  | 100   |
| Selasa | 110   |
| Rabu   | 105   |


Prediksi:

* Kamis = Rabu = **105**

Versi Naive yang Lebih Realistis


- Simple Naive

    $$
    \hat{Y}_{t+1} = Y_t
    $$

    langsung copy nilai terakhir


- Seasonal Naive

    $$\hat{Y}_{t} = Y_{t-m}$$


    Intuisi:

    > “Gunakan nilai di periode yang sama sebelumnya”


    Contoh:

    > hari ini Minggu, maka pakai Minggu minggu lalu



- Drift Method (sedikit lebih pintar)

    $$\hat{Y}_{t+h} = Y_t + h \cdot \frac{Y_t - Y_1}{t-1}$$
    
    
    Intuisi:
    
    > “lanjutkan arah dari awal sampai sekarang”
    
    
    Contoh:
    
    > dari 100 → 200 dalam 10 hari, naik ~10 per hari

**Kegiatan Data Analyst**


- Sebagai Baseline Model

    Ini penting banget:

    > Naive = patokan minimal performa

    Contoh:

    Kalau model kamu:

    > lebih jelek dari naive → model kamu gagal


    ini sering kejadian 😄


- Quick Forecast

    Kalau butuh cepat:

    * meeting
    * dashboard

> Naive forecasting adalah metode prediksi paling sederhana, di mana nilai masa depan diasumsikan sama dengan nilai terakhir.

Secara matematis:

> * Seasonal naive → pakai periode yang sama

### **Moving Average Forecast**

 **Awalnya: Moving Average (sebagai smoothing)**

Misalnya MA sederhana:

$$
Y_t = \frac{1}{q} (Y_{t-1} + Y_{t-2} + \dots + Y_{t-q})
$$

Ini artinya:

* Kita “meratakan” masa lalu
* Tidak ada konsep probabilitas
* Tidak ada error model
* Murni teknik smoothing

**Insight penting (ini titik transformasinya)**

Dalam time series, kita mulai berpikir:

> “Data itu bukan cuma angka, tapi hasil dari proses + gangguan (noise)”

Artinya kita tulis:

$$
Y_t = \mu + \epsilon_t
$$

di mana:

* (\mu) = baseline (rata-rata)
* (\epsilon_t) = shock / error (acak)


**Substitusi ke dalam smoothing**

Sekarang kita masukkan ke MA smoothing tadi.

Misalnya MA(1) sederhana:

$$
Y_t = \frac{1}{2}(Y_{t-1} + Y_{t-2})
$$

Substitusi:

$$
Y_{t-1} = \mu + \epsilon_{t-1}
$$

$$
Y_{t-2} = \mu + \epsilon_{t-2}
$$

jadi:

$$
Y_t = \frac{1}{2}(\mu + \epsilon_{t-1}) + (\mu + \epsilon_{t-2})
$$

$$
Y_t = \mu + \frac{1}{2}(\epsilon_{t-1} + \epsilon_{t-2})
$$

**Tambahkan shock sekarang**

Realitanya, data sekarang juga kena shock:

$$
Y_t = \mu + \epsilon_t + \frac{1}{2}(\epsilon_{t-1} + \epsilon_{t-2})
$$

Boom — ini jadi MA model

Bentuk umum:

$$
Y_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2}
$$

Ini adalah **MA(q) model**

**Intuisi kenapa bisa berubah**

Perubahan mindset:

| Smoothing                 | Model                     |
| ------------------------- | ------------------------- |
| Rata-rata data masa lalu  | Kombinasi shock masa lalu |
| Deterministik             | Stochastic (acak)         |
| Tidak ada error eksplisit | Error jadi pusat model    |

**Intuisi paling sederhana**

Bayangin:

* Smoothing:
  > “Nilai sekarang = rata-rata masa lalu”

* MA model:
  > “Nilai sekarang = akibat dari kejutan-kejutan sebelumnya”

**Kenapa ini masuk akal?**

Karena dalam dunia nyata:

* Harga saham naik bukan karena rata-rata masa lalu
* Tapi karena **shock informasi (berita, event, dll)**

Jadi lebih masuk akal model:

$$
Y_t = \text{efek shock sekarang} + \text{efek shock masa lalu}
$$


Ringkasan singkat

Transformasinya terjadi karena:

- Kita anggap:  
    $$
   Y_t = \mu + \epsilon_t
   $$

- Kita substitusi ke smoothing  

- Kita tambahkan shock saat ini  

- Hasilnya:
     kombinasi linear dari error → MA model

Proses Transformasi  

- Awal: 1/q itu dari mana?

    Di smoothing:

    $$
    Y_t = \frac{1}{q}(Y_{t-1} + \cdots + Y_{t-q})
    $$

    Kenapa 1/q?
    > Karena kita **memaksa semua bobot sama** $\to$ “Semua masa lalu dianggap sama penting”

    Ini **keputusan manusia**, bukan dari data.  

    $$ \ $$

- Masalah dari 1/q

    Realitanya:

    * Kemarin mungkin lebih penting dari 3 hari lalu
    * Atau justru efeknya beda arah

    Jadi:
    $$
    \text{bobot sama (1/q)} ; \text{= terlalu kaku}
    $$  

    $$ \ $$

- Upgrade ke model (ini titik penting)

    Kita ubah mindset:

    Dari:

    > “Aku tentukan bobotnya (1/q)”

    Menjadi:

    > “Biar data yang menentukan bobot terbaik”  

    $$ \ $$

- Masuk ke MA model

    $$
    Y_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1} + \cdots + \theta_q \epsilon_{t-q}
    $$

    Di sini:

    * (\theta_1, \theta_2, \dots) = **parameter**
    * nilainya **dipelajari dari data (estimasi)**  
    
    $$ \ $$

- Jadi 1/q → θ itu apa?

    Bukan transformasi aljabar langsung seperti:

    $$
    \frac{1}{q} \rightarrow \theta
    $$

    Tapi ini:

    > **Generalisasi / pelonggaran constraint**  

    $$ \ $$

- Analogi sederhana

    - Smoothing:

        $$
        (0.5, 0.5)
        $$

        kamu bilang:

        > “dua hari terakhir sama penting”



    - MA model:

        $$
        (\theta_1, \theta_2) = (0.8, -0.3)
        $$

        data bilang:

        > “hari kemarin penting, tapi 2 hari lalu malah efeknya kebalik”

### **Exponential Smoothing**

#### **Introduction**

**Definisi**

> **Prediksi = kombinasi antara nilai sekarang dan prediksi sebelumnya**


Bentuk matematis (Single Exponential Smoothing)

$$\hat{Y}_{t+1} = \alpha Y_t + (1-\alpha) \hat{Y}_t$$


Artinya:

* $Y_t$ = data terbaru
* $\hat{Y}_t$ = prediksi sebelumnya
* $\alpha$ = bobot (0–1)

**Intuisi**  

> Data “percaya masa lalu, tapi lebih percaya yang terbaru”

Bayangin kamu prediksi penjualan:

* kemarin = 100
* prediksi sebelumnya = 90

kamu gak langsung:

* pakai 100 (naive)
* atau rata-rata semua data


tapi:

> “aku gabung, tapi lebih percaya yang baru”


Intinya:

> data terbaru lebih penting, tapi masa lalu tetap dihitung

**Contoh**


Data:

| Hari | Sales |
| ---- | ----- |
| 1    | 100   |
| 2    | 110   |
| 3    | 105   |


Misal:

* $\alpha = 0.6$
* prediksi hari ke-2 = 100


Prediksi hari ke-3:

$$
\hat{Y}_3 = 0.6(110) + 0.4(100) = 66 + 40 = 106
$$


lebih dekat ke data terbaru (110)

**Peran α (alpha)**



Intuisi:

- α besar (≈ 1)

    > fokus ke data terbaru
    
    * respons cepat
    * sensitif


- α kecil (≈ 0)

    > lebih percaya masa lalu
    
    * lebih smooth
    * lambat respon


Analogi:

* α besar = “reaktif”
* α kecil = “tenang”


**Hubungan dengan Moving Average**


- Moving Average:

    > semua data punya bobot sama


- Exponential Smoothing:

    > bobot **menurun secara eksponensial**

- Kesimpulan:

    > semakin lama → semakin kecil pengaruhnya

**Intuisi “Exponential”**

Kenapa disebut exponential?

karena bobotnya:

* $Y_t$ → α
* $Y_{t-1}$ → α(1-α)
* $Y_{t-2}$ → α(1-α)²

Maka:
> makin lama → makin kecil secara eksponensial

**Kegiatan Data Analyst**  



- Forecast cepat tapi lebih pintar dari naive

    Manager tanya:

    > “Prediksi besok berapa?”


    - Naive:

        > = hari ini

    - Exponential smoothing:

        > = gabungan history + recent


    lebih stabil, tapi tetap responsif


- Data tanpa trend & seasonality

    > ini kondisi ideal SES (Single Exponential Smoothing)


- Dashboard / monitoring

    * smoothing data
    * lihat arah tanpa noise

**Kapan cocok / tidak**


Cocok:

* data relatif stabil
* tidak ada trend kuat
* tidak ada seasonality


Tidak cocok:

* ada trend → butuh Holt
* ada seasonality → butuh Holt-Winters


**Holt’s Method (Trend)**

- Tambah trend:

    $$
    \text{level + trend}
    $$


- Holt-Winters (Trend + Seasonality)

    Tambah seasonality:
    
    $$
    \text{level + trend + seasonality}
    $$
    
    
    ini versi lengkapnya

**Ringkasan**  
> Exponential smoothing adalah metode forecasting yang memberikan bobot lebih besar pada data terbaru, dan bobot yang semakin kecil pada data lama.

Secara matematis:
$$
\hat{Y}_{t+1} = \alpha Y_t + (1-\alpha)\hat{Y}_t
$$

Intuisi:
> Prediksi adalah kombinasi antara nilai terbaru dan estimasi sebelumnya.

Parameter:
> * α besar → respons cepat
> * α kecil → lebih smooth

Intuisi:
> “Exponential smoothing adalah kompromi antara naive (cepat) dan moving average (stabil)”


#### **Advanced**

**Gambaran Besar**

| Model        | Bisa handle                 |
| ------------ | --------------------------- |
| SES          | level saja                  |
| Holt         | level + trend               |
| Holt-Winters | level + trend + seasonality |


Intinya:

> Kita menambahkan “komponen” ke dalam smoothing

##### **Holt's Method**

**Konsep**

> Kita tidak hanya prediksi nilai, tapi juga **arah pergerakan (trend)**

Kalau tadi (SES):

> “pakai nilai terakhir”

Sekarang:

> “pakai nilai terakhir + arah geraknya”


**Analogi**

Bayangin kamu lihat mobil:

* posisinya sekarang di 100
* kecepatannya naik +10 per detik

> kamu tidak hanya lihat posisi
> tapi juga **kecepatan**

Bentuk Matematis

- Level (nilai sekarang)  
    
    $$l_t = \alpha Y_t + (1-\alpha)(l_{t-1} + b_{t-1})$$

    Level sekarang adalah gabungan antara:  
    > "data asli sekarang & prediksi dari sebelumnya (level lama + trend lama)"  

    Keterangan:
    - $Y_t \to$ “kenyataan hari ini”
    - $l_{t-1} + b_{t-1} \to$ “perkiraan kemarin tentang hari ini”  

    Intinya:

    > Kita tidak langsung percaya data sekarang,  
    > tapi juga mempertimbangkan “ekspektasi dari masa lalu”   


    Kayak kamu:

    > punya tebakan sebelumnya
    > lalu lihat data baru

    kamu update keyakinan, bukan langsung ganti total
      
     $$ \ $$

- Trend (arah)  
    $$b_t = \beta (l_t - l_{t-1}) + (1-\beta)b_{t-1}$$

    Trend sekarang adalah gabungan antara:

    > perubahan terbaru (seberapa cepat naik/turun sekarang)
    > trend sebelumnya

    Keterangan:
    - $l_{t}−l_{t−1}$ → “kenaikan terbaru”
    - $b_{t−1}$	​→ “kecepatan lama”  
     
    Intinya:  
    > Kita tidak langsung percaya perubahan terbaru,
    > tapi kita haluskan dengan trend sebelumnya  

    Kayak kamu mengukur kecepatan mobil:
    > 
    > - lihat perubahan sekarang
    > - tapi juga ingat kecepatan sebelumnya

    biar tidak terlalu reaktif

    $$ \ $$

- Forecast

    $$\hat{Y}_{t+h} = l_t + h \cdot b_t$$
    
    Bahasa manusia:

    > Prediksi masa depan = posisi sekarang + lanjutkan arah (trend) ke depan

    Lebih dalam:
    - $l_t$	​ → posisi sekarang
    - $b_t$ → kecepatan
    - $h $→ berapa langkah ke depan

    Intinya:

    > Kita “lanjutkan garis lurus” dari kondisi sekarang

    Kayak kamu lihat mobil:

    - posisi sekarang = 130
    - kecepatan = +10

    maka:

    - 1 detik lagi = 140
    - 2 detik lagi = 150

**Contoh Konkret**


| t | (Y_t) |
| - | ----- |
| 1 | 100   |
| 2 | 112   |
| 3 | 118   |
| 4 | 135   |


Parameter

* (\alpha = 0.6) (level)
* (\beta = 0.4) (trend)


Inisialisasi (awal)

* **Level awal**:  
    
  $
  l_1 = Y_1 = 100
  $

* **Trend awal (kecepatan)**:
  
  $
  b_1 = Y_2 - Y_1 = 112 - 100 = 12
  $


Iterasi Step-by-Step


- t = 2

  - Level

    $
    l_2 = 0.6(112) + 0.4(100 + 12)
    = 67.2 + 44.8 = 112
    $


  - Trend

    $
    b_2 = 0.4(112 - 100) + 0.6(12)
    = 4.8 + 7.2 = 12
    $


- t = 3

  - Level

    $
    l_3 = 0.6(118) + 0.4(112 + 12)
    = 70.8 + 49.6 = 120.4
    $


  - Trend

    $
    b_3 = 0.4(120.4 - 112) + 0.6(12)
    = 3.36 + 7.2 = 10.56
    $


- t = 4

  - Level

    $
    l_4 = 0.6(135) + 0.4(120.4 + 10.56)
    = 81 + 52.384 = 133.384
    $


  - Trend

    $
    b_4 = 0.4(133.384 - 120.4) + 0.6(10.56)
    = 5.1936 + 6.336 = 11.5296
    $


Forecast


  - 1 langkah ke depan (h = 1)

    $
    \hat{Y}_5 = l_4 + b_4 = 133.384 + 11.5296 = 144.9136
    $


  - 2 langkah ke depan (h = 2)

    $
    \hat{Y}_6 = 133.384 + 2(11.5296) = 156.4432
    $

Intuisi dari hasil ini

- Perhatikan yang menarik:


    ```text
    100 → 112 → 118 → 135
    ```

- tidak stabil:

    * awal naik +12
    * lalu +6
    * lalu +17

**Apa yang dilakukan Holt?**

- Level

    * tidak langsung ikut 135
    * tapi “ditarik” ke arah prediksi sebelumnya

    hasilnya: **133.384 (lebih smooth)**


- Trend (kecepatan)

    * awalnya 12
    * lalu turun → 10.56
    * lalu naik lagi → 11.53
    
    tidak lompat-lompat ekstrem


- Intinya:

    > model “menenangkan” data yang noisy

**Cara ngebayanginnya**

Bayangin kamu mengamati mobil:

* kadang ngebut
* kadang melambat

kamu tidak langsung percaya tiap perubahan

tapi kamu:

* haluskan posisi
* haluskan kecepatan

**Insight penting**


- Level ≠ data asli

    > level = versi “halus” dari data


- Trend ≠ selisih langsung

    > trend = kecepatan yang sudah di-smoothing


- Forecast = extrapolation

    > lanjutkan kondisi terakhir (level + trend)

**Ringkasan versi catatan**


Inisialisasi:
> $$
> l_1 = Y_1,\quad b_1 = Y_2 - Y_1
> $$

Iterasi:
>
> * update level
> * update trend

Forecast:
> $$
> \hat{Y}_{t+h} = l_t + h b_t
> $$

Intuisi:
>
> * level = posisi yang sudah dihaluskan
> * trend = kecepatan yang sudah dihaluskan
> * forecast = melanjutkan arah tersebut

##### **Holt-Winters**

**Konsep**

> Tambahkan **pola berulang (seasonality)** ke dalam model

Sekarang model mikir:

> “ada arah (trend) + ada pola berulang (seasonality)”

Bayangin penjualan:

* naik tiap bulan (trend)
* tapi weekend selalu tinggi (seasonality)


model harus tangkap dua-duan

Bentuk Matematis (Additive)


- Level

    $$l_t = \alpha (Y_t - s_{t-m}) + (1-\alpha)(l_{t-1} + b_{t-1})$$


    Level sekarang adalah gabungan antara:

    - data sekarang yang sudah “dibersihkan” dari efek seasonality
    - prediksi sebelumnya (level lama + trend lama)
    
    Intuisi:
    - $Yt−s_{t−m}$→ “nilai asli tanpa efek musim”
    - $l_{t−1}+b_{t−1}$ → “perkiraan kemarin tentang hari ini”
    
    Intinya:  
    > Kita ingin tahu “nilai dasarnya”, jadi kita hilangkan efek seasonality dulu

    Misal:
    - weekend selalu +20 penjualan
    - kalau hari ini 120:

        > nilai “aslinya” ≈ 100

    - itu yang dipakai untuk update level

    $$ \ $$

- Trend

    $$b_t = \beta (l_t - l_{t-1}) + (1-\beta)b_{t-1}$$


    Trend sekarang adalah gabungan antara:

    - perubahan terbaru pada level
    - trend sebelumnya

    Intuisi:
    - $l_t−l_{t−1}$	​ → “seberapa cepat berubah sekarang”
    - $b_{t−1}$	​ → “arah lama”

    Intinya:

    > Kita haluskan arah supaya tidak terlalu reaktif

    Kayak kecepatan mobil:

    - lihat perubahan sekarang
    - tapi tetap ingat kecepatan sebelumnya

    $$ \ $$

- Seasonality

    $$s_t = \gamma (Y_t - l_t) + (1-\gamma)s_{t-m}$$

    Seasonality sekarang adalah gabungan antara:

    - selisih data dengan level (efek musiman terbaru)
    - pola seasonality sebelumnya

    Intuisi:
    - $Y_t​−l_t$ → “berapa tambahan karena musim sekarang”
    - $s_{t−m}$	​ → “pola musim sebelumnya (misal minggu lalu)”

    Intinya:

    > Kita update “kebiasaan musiman” secara bertahap


    Misal:

    > Sabtu biasanya +30

    Tapi hari ini:
    > ternyata +35

    kita update:

    > “oh mungkin sekarang Sabtu lebih tinggi dari biasanya”

    $$ \ $$

- Forecast

    $$\hat{Y}_{t+h} = l_t + h b_t + s_{t+h-m}$$

    Bahasa manusia:

    > Prediksi masa depan = nilai dasar + arah (trend) + pola musiman

    Intuisi:
    - $l_t$	​ → posisi sekarang
    - $h b_t$	​ → gerak ke depan
    - $s_{t+h−m}$ → pola musim di waktu itu

    Intinya:

    > Kita tidak hanya lanjutkan trend,  
    > tapi juga menambahkan pola yang sesuai waktunya

    Bayangin:

    > bisnis lagi naik (trend)
    > dan weekend selalu tinggi (seasonality)

    prediksi Sabtu depan:

    > naik karena trend
    > tambah lagi karena weekend

**Contoh Konkret**

Penjualan:

* naik tiap bulan
* weekend selalu tinggi


Holt-Winters akan:

* tangkap trend naik
* tangkap pola weekend


Prediksi:

* hari Sabtu depan → lebih tinggi dari weekday
* tapi juga mengikuti trend naik

**Additive vs Multiplicative**


- Additive

    $$
    Y_t = T_t + S_t
    $$

    selisih tetap


- Multiplicative

    $$
    Y_t = T_t \times S_t
    $$

    skala berubah


- Intuisi:

    * additive → beda tetap
    * multiplicative → proporsional


**Peran α, β, γ**


| Parameter | Fungsi      |
| --------- | ----------- |
| α         | level       |
| β         | trend       |
| γ         | seasonality |


Intuisi:

* besar → cepat adaptasi
* kecil → lebih smooth

**Kegiatan Data Analyst**


- Forecast penjualan

    * ada trend
    * ada seasonality

    pakai Holt-Winters


- Dashboard

    * smoothing + pattern
    * lebih akurat dari moving average


- Business decision


    Kasus:

    > “Harus stok berapa minggu depan?”

    model:

    * lihat trend naik
    * lihat weekend tinggi


    Decision:

    * tambah stok weekend
    * adjust dengan growth


**Kapan pakai apa**


| Kondisi             | Model        |
| ------------------- | ------------ |
| stabil              | SES          |
| ada trend           | Holt         |
| trend + seasonality | Holt-Winters |

**Cara Mikir Sederhana**


> SES = “nilai sekarang”  
> 
> Holt = “nilai + arah”  
> 
> Holt-Winters = “nilai + arah + pola”

**Ringkasan**  

- Holt method adalah pengembangan exponential smoothing yang menangkap level dan trend.  
- Holt-Winters menambahkan komponen seasonality ke dalam model.

Intuisi:
* Holt → memperhatikan arah perubahan
* Holt-Winters → memperhatikan arah + pola berulang

## **ARIMA**

````python
import matplotlib.pyplot as plt
import numpy as np
````

**Data awal**

Misal data:

| t | Yₜ  |
| - | --- |
| 1 | 100 |
| 2 | 110 |
| 3 | 120 |
| 4 | 125 |
| 5 | 140 |

📌 Masih ada trend → ❌ belum stationary

````python
t   = np.array([1, 2, 3, 4, 5])
y_t = np.array([100, 110, 120, 125, 140])

plt.plot(t, y_t) ;
````

**Differencing (d = 1)**

$$Y'_t = Y_t - Y_{t-1}$$

Hitung:

| t | Yₜ  | Y'ₜ |
| - | --- | --- |
| 2 | 110 | 10  |
| 3 | 120 | 10  |
| 4 | 125 | 5   |
| 5 | 140 | 15  |

Data baru:

> 10, 10, 5, 15

````python
y_t_l = y_t[1:] - y_t[:-1]

plt.plot(t[1:], y_t_l) ;
````

**Bentuk model ARIMA(1,1,1)**

Setelah differencing → jadi:

$$Y'_t = \phi Y'_{t-1} + \epsilon_t + \theta \epsilon_{t-1}$$

Artinya:

* φ → pengaruh nilai sebelumnya
* θ → pengaruh error sebelumnya

**Tentukan parameter (pakai pendekatan sederhana)**

Biar gampang, kita asumsi:

* φ = 0.6
* θ = 0.4

(Tujuannya biar kamu paham alur, bukan fokus ke estimasi statistik)

Hitung error (ε)

Rumus:

$$
\epsilon_t = Y'_t - (\phi Y'_{t-1} + \theta \epsilon_{t-1})
$$

Kita hitung step-by-step:


- t = 2

  Belum ada sebelumnya → asumsi:

  ε₂ = 0


- t = 3

  Prediksi:

  * φ × Y'₂ = 0.6 × 10 = 6
  * θ × ε₂ = 0

  Total prediksi = 6

  Error:

  ε₃ = 10 - 6 = **4**

## **Stationary & Differencing**

### **Stationary**

**Apa itu Stationary?**  

Definisi sederhana:

> Data disebut **stationary** kalau “sifat statistiknya tidak berubah sepanjang waktu”

Yang “tidak berubah” itu:

* rata-rata (mean)  
 
  $E(Y_t) = \mu \to \text{Mean Konstan}$

* variasi (variance)
  
  $Var(Y_t) = \mu \to \text{Varians Konstan}$

* pola hubungan waktu


Bahasa mudahnya:
> Sifat statistik data tidak berubah walaupun waktunya maju

Bayangin kamu lihat data:

- Tidak stationary:

    ```text
    100 → 120 → 150 → 200
    ```

    terus naik


- Stationary:

    ```text
    100 → 105 → 98 → 102 → 101
    ```

    naik turun di sekitar angka yang sama

Intinya:

> Stationary = “data berosilasi di sekitar level yang sama”

**Kenapa Harus Stationary?**

- Problem kalau tidak stationary:

    Model seperti AR, MA, ARMA itu:

    > “belajar dari pola masa lalu”


    Kalau datanya:

    * berubah terus (trend)
    * makin besar variansnya

    > pola lama **tidak relevan lagi**


- Bayangin kamu belajar:

    * hari ini suhu 30°C
    * besok 31°C

    > masuk akal


    Tapi kalau:

    * 30°C → 100°C → 300°C

    gak bisa diprediksi pakai pola lama

- Kesimpulan:

    > Model butuh dunia yang “stabil” → itulah stationary

**Jenis Stationary (biar lengkap tapi tetap masuk akal)**

- Weak Stationary (yang paling sering dipakai)

    > mean dan variance konstan

    ini yang dipakai di ARIMA

- Strong Stationary (lebih ketat)

    > semua distribusi sama
    
    jarang dipakai di praktik


Fokus kita:

> **weak stationary**

**Kenapa Data Jadi Tidak Stationary?**

Penyebab utama:

- Trend

    * naik terus
    * turun terus


- Seasonality

    > pola berulang kuat


- Variance berubah

    > makin lama makin besar fluktuasi

**Cara Cek Stationary**  

- STEP 1 — Plot Time Series
    
    Apa yang dilakukan?

    > Plot data terhadap waktu

    Apa yang dilihat?
    
  - Trend  
    
    > 10 → 12 → 15 → 18 → 22
    
    Artinya:
    > naik terus
    
    Intuisi:
    > “pusat data geser”
    
    Kesimpulan:  
    
    > tidak stationary

  - Musiman (seasonality)
    
    > 10 → 20 → 10 → 20 → 10
    
    Artinya:
    > pola berulang
    
    Intuisi:
    > “pola berubah tergantung waktu”

    Kesimpulan:
    > tidak stationary

  - Variansi berubah


    > kecil → makin besar → makin liar

    Intuisi:
    > “ketidakpastian makin besar”

    Kesimpulan:  
    > tidak stationary

    $$ \ $$

- STEP 2 — Rolling Mean & Variance

    Kenapa pakai rolling?

    Karena:     
    > Kita tidak punya distribusi → kita cek lokal

    Cara kerja

    > Ambil window tetap

    Misal window = 3:
    > [10,12,15] → [12,15,18] → [15,18,22]

    Rumus 
    
    - Rolling mean  

        $$\mu_t = \frac{1}{w} \sum_{i=0}^{w-1} Y_{t-i}$$

    - Rolling Varians

        $$σ_t^2 = \frac{1}{w} \sum_{i=0}^{w-1} (Y_{t-i}- \mu_t)^2$$

    Rolling itu menjawab:

    > “Kalau aku lihat sebagian data di waktu berbeda, apakah sifatnya sama?”

    Cara interpretasi
    
    - Tidak stationary  

        | Window | Mean | Variance |
        | ------ | ---- | -------- |
        | 1      | 12   | 4        |
        | 2      | 15   | 6        |
        | 3      | 18   | 8        |
 
        Keterangan:  
        > naik terus
        
        Artinya:
        
        > “pusat & sebaran berubah”
    
    - Stationary  

        | Window | Mean | Variance |
        | ------ | ---- | -------- |
        | 1      | 10.3 | 1.5      |
        | 2      | 10.6 | 1.5      |
        | 3      | 10.0 | 0.6      |  

        Keterangan:  
        > relatif stabil

        Artinya:  
        > “data konsisten”

    $$ \ $$

- STEP 3 — ADF Test (Formal Check)  
    
    Model dasar

    $$ΔY_t	​=α+βY_{t−1}	​+ϵ_t$$
    

    Pertanyaan:

    > “Apakah masa lalu masih ‘narik kuat’ ke masa sekarang?”

    Hipotesis
    - H0 → tidak stationary
    - H1 → stationary

    Interpretasi
    - p-value < 0.05 → ✅ stationary
    - p-value > 0.05 → ❌ tidak stationary
    
    Kalau:

    > “nilai sekarang terlalu bergantung pada masa lalu”

    Kesimpulan:  
    > data belum stabil 

    $$ \ $$

- STEP 4 — Perbaiki kalau tidak stationary
    
    - Differencing   
        
        $$Y_t′	​=Y_t	​−Y_{t−1}$$
        Intuisi:  

        > “hapus arah (trend)”

    - Seasonal Differencing
        
        $$Y_t	​−Y_{t−12}$$

        Intuisi:

        > “hapus pola berulang”

        Transformasi
        - log
        - sqrt

        Intuisi:

        > “stabilkan fluktuasi”

**Workflow**
- Plot data
 
- Cek trend / seasonality (Optional)
- Rolling mean & variance (Optional)
- ADF test
- Jika tidak stationary → differencing
- Ulangi sampai stationary

**Tambahan**  

- Kalau dekomposisi sudah jelas  
    
    Ada trend / seasonal kuat
    > sudah pasti tidak stationary

    Rolling boleh di-skip (opsional saja)

- Kalau dekomposisi tidak jelas
    
    > Data terlihat “acak”
    
    tapi belum tentu stationary

    > Rolling jadi berguna

    kalau mean / variance berubah → ❌ tidak stationary

- Peran masing-masing
    - Dekomposisi → lihat struktur (trend & seasonal)
    - Rolling → lihat perubahan statistik
    - ADF → keputusan final

### **Differencing**

**Differencing (Solusi Utama)**

- Konsep:

    > Differencing = ambil “perubahan” antar waktu

- Rumus:

    $$Y'_t = Y_t - Y_{t-1}$$

- Intuisi:

    > Daripada lihat “nilai”, kita lihat “perubahan”

- Analogi kuat:

    - Sebelum:
    
        Tabungan:
    
        ```text
        100 → 120 → 140 → 160
        ```
    
        terus naik → tidak stationary
    
    
    - Setelah differencing:
    
        ```text
        +20 → +20 → +20
        ```
        
        stabil → stationary

**Contoh**

```text
100, 110, 120, 130, 140
```

Problem:

* jelas ada trend
* tidak stationary


Step 1 - Differencing:

```text
110-100 = 10  
120-110 = 10  
130-120 = 10  
140-130 = 10  
```

Hasil:

```text
10, 10, 10, 10
```

Insight:

* mean stabil
* tidak ada trend

sekarang stationary


**Kalau Belum Stationary?**

Kadang 1x differencing belum cukup

Second differencing:

$$Y''_t = Y'_t - Y'_{t-1}$$

Intuisi:

> kita ambil perubahan dari perubahan

Contoh:

```text
10, 12, 15, 20
```


Differencing lagi:

```text
2, 3, 5
```


lebih stabil

**Hubungan dengan ARIMA**

$$\text{ARIMA(p, d, q)}$$

keterangan:  
- d = jumlah differencing

Artinya:

> kita ubah data sebanyak d kali sampai stationary


Contoh:

* d = 1 → sekali differencing
* d = 2 → dua kali

**Cara Mikir Analyst**

at lihat data:

Pertanyaan:

* apakah ada trend?
* apakah mean berubah?
* apakah fluktuasi makin besar?

lau iya:

> data tidak stationary

lusi:

> pakai differencing

**Ringkasan**

- Stationary:

    > data stabil, gak berubah sifatnya

- Non-stationary:

    > ada trend / perubahan besar

- Differencing:

    > ubah data jadi perubahan supaya stabil


- Inti:

    > ARIMA itu bekerja dengan cara:  
    > **mengubah data jadi stationary → lalu belajar pola dari sana**
