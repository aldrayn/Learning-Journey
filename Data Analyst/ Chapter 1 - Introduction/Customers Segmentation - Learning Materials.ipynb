# **Customers Segmentation**

## **Introduction**

**Latar Belakang Masalah**  

Dalam bisnis, sering ada asumsi:

> “Semua pelanggan diperlakukan sama”

Padahal realitanya:

* Ada pelanggan yang **sering beli**
* Ada yang **sekali beli lalu hilang**
* Ada yang **belanja besar**
* Ada yang **sensitif harga**

**Masalah utamanya:**

> Strategi marketing jadi “generik” → boros biaya, hasil tidak optimal

Contoh:

* Kasih diskon ke semua orang ❌
* Padahal:

  * pelanggan loyal tidak butuh diskon
  * pelanggan baru butuh dorongan
  * pelanggan hampir churn butuh “diselamatkan”

Jadi inti masalah:  
> **Tidak tahu siapa pelanggan kita sebenarnya**

Contoh Nyata
- Customer A

  * belanja tiap minggu,
  * total spending besar.

- Customer B

  * terakhir aktif 6 bulan lalu.

Strategi bisnisnya tentu berbeda.

**Tujuan**

Dari masalah tadi, kita turunkan tujuan:

- Tujuan Utama:

    > Memahami perbedaan perilaku pelanggan

- Tujuan Spesifik:

    - Mengelompokkan pelanggan berdasarkan karakteristik
    - Menentukan strategi berbeda untuk tiap kelompok
    - Meningkatkan efisiensi marketing (tidak buang-buang budget)
    - Meningkatkan retention (pelanggan tetap bertahan)


Contoh Segment

| Segment        | Karakteristik            |
|----------- |--------------------- |
| Loyal Customer | sering transaksi         |
| High Value     | spending besar           |
| New Customer   | baru bergabung           |
| At Risk        | mulai tidak aktif        |
| Churn Risk     | kemungkinan churn tinggi |

**Kenapa Perlu Metrik?**

Segmentasi itu tidak bisa pakai feeling.

Contoh buruk:

* “Kayaknya dia pelanggan bagus”
* “Sepertinya dia jarang beli”

Ini subjektif

**Solusinya:**

> Kita ubah perilaku pelanggan menjadi angka → itulah metrik

Masalah yang Ingin Dijawab


- Contoh 1

    > “Customer mana paling menguntungkan?”


- Contoh 2

    > “Customer mana yang kemungkinan churn?”


- Contoh 3

    > “Segment mana paling cocok diberi promo?”

Jenis Customer Segmentation

| Jenis         | Fokus            |
|---------- |------------- |
| Demographic   | umur, gender     |
| Geographic    | lokasi           |
| Behavioral    | perilaku user    |
| Transactional | transaksi        |
| Psychographic | minat/gaya hidup |

## **Demographic Segmentation**

Konsep

> Mengelompokkan customer berdasarkan identitas dasarnya, seperti umur, gender, pekerjaan, pendidikan, atau pendapatan.

Intinya:

> “Siapa customer ini?”

Tujuan:
> Supaya bisnis tahu kelompok customer utama mereka dan bisa menyesuaikan produk atau promosi.

Metric yang sering dipakai:

- Age (umur)
- Gender
- Income (pendapatan)
- Education level (tingkat pendidikan)
- Occupation (profesi)
- Marital status (status perkawinan)

Contoh:

- Umur 18–24 lebih sering beli skincare murah.
- Umur 30+ lebih sering beli produk premium.
- Mahasiswa lebih sensitif harga dibanding pekerja tetap.

Insight yang bisa didapat:

- Produk mana cocok untuk kelompok tertentu sehingga bisa menjadi target utama marketing.
- Harga yang sesuai untuk tiap kelompok.
- Strategi marketing per usia/gender karena gender tertentu mungkin saja punya conversion rate lebih tinggi.
- Pendapatan memengaruhi daya beli customer.

## **Geographic Segmentation**


Konsep  
> Mengelompokkan customer berdasarkan lokasi tempat tinggal atau wilayah aktivitas.

Intinya:

> “Customer ini berada di mana?”

Tujuan:
> Karena perilaku customer sering berbeda tergantung daerah, budaya, cuaca, atau kepadatan kota.

Metric yang sering dipakai:

- Country
- Province / city
- Urban (kota) vs rural (desa/pedalaman)
- Time zone
- Climate (iklim)
- Population density

Contoh:

- User Jakarta lebih sering pakai same-day delivery.
- Daerah dingin lebih banyak beli jaket.
- Customer kota besar lebih aktif belanja online malam hari.

Insight yang bisa didapat:
- Waktu aktif user bisa berbeda tiap wilayah.
- Penyesuaian stok barang per daerah.
- Promosi khusus wilayah tertentu.
- Menentukan lokasi gudang atau cabang.

## **Behavioral Segmentation**

### **Introduction**

Konsep 
> Mengelompokkan berdasarkan perilaku user saat menggunakan aplikasi atau membeli produk.

Intinya:

> “Customer ini biasanya ngapain?”

Tujuan:
> Memahami kebiasaan user agar bisnis bisa memprediksi tindakan berikutnya.

Metrik yang digunakan:
- Session count
- Click-through rate (CTR)
- Time spent
- Feature usage
- Purchase frequency
- Add-to-cart rate
- Bounce rate
- Engagement rate
- Retention rate

Contoh:

- User sering buka aplikasi tapi tidak checkout.
- User hanya aktif saat diskon.
- User sering membeli kategori gaming.

Insight yang bisa didapat:
- Bisa mendeteksi user yang mulai tidak aktif.
- Mengetahui fitur paling populer.
- Membantu recommendation system dan churn analysis.
- Produk rekomendasi yang cocok.
- Strategi retargeting iklan.

Behavioral biasanya sangat penting di data science karena datanya paling “hidup”.

### **Metric**

Metric yang sering dipakai:

- Session Count  
    Jumlah sesi penggunaan aplikasi/website oleh user.

   

    $$
    \text{Session\ Count} = \sum \text{Total\ Session}
    $$
    
    Semakin tinggi biasanya user semakin aktif.


- Click-Through Rate (CTR)  
    Persentase user yang mengklik dibanding yang melihat.

   

    $$
    \text{CTR} = \frac{\text{Jumlah\ Klik}}{\text{Jumlah\ Impression}} \times 100%
    $$
    
    Mengukur ketertarikan user terhadap iklan atau konten.


- Time Spent  
    Rata-rata waktu yang dihabiskan user.

   

    $$
    \text{Average\ Time\ Spent} = \frac{\text{Total\ Waktu}}{\text{Jumlah\ User\ atau\ Session}}
    $$
    
    Semakin lama biasanya engagement semakin tinggi.


- Feature Usage  
    Jumlah atau persentase penggunaan fitur tertentu.  

    $$
    \text{Feature\ Usage} = \frac{\text{User\ Pengguna\ Fitur}}{\text{Total\ User}} \times 100%
    $$
    
    Mengetahui fitur yang paling sering dipakai.


- Purchase Frequency  
    Seberapa sering customer melakukan pembelian.

   

    $$
    \text{Purchase\ Frequency} = \frac{\text{Jumlah\ Transaksi}}{\text{Jumlah\ Customer}}
    $$
    
    Mengukur loyalitas dan kebiasaan belanja customer.


- Add-to-Cart Rate  
    Persentase user yang menambahkan produk ke keranjang.

   

    $$
    \text{Add\text{-}to\text{-}Cart\ Rate} = \frac{\text{Jumlah\ Add\ to\ Cart}}{\text{Jumlah\ Visitor}} \times 100%
    $$
    
    Menunjukkan ketertarikan user terhadap produk.


- Bounce Rate  
    Persentase user yang langsung keluar tanpa interaksi lanjutan.

    $$
    \text{Bounce\ Rate} = \frac{\text{Single\ Session}}{\text{Total\ Session}} \times 100%
    $$
    
    Bounce tinggi biasanya menandakan halaman kurang menarik atau tidak relevan.


- Engagement Rate  
    Tingkat interaksi user terhadap platform atau konten.  

    $$
    \text{Engagement\ Rate} = \frac{\text{Total\ Interaction}}{\text{Total\ Reach}} \times 100%
    $$
    
    Mengukur seberapa aktif user berinteraksi.


- Retention Rate  
    Persentase user yang tetap kembali menggunakan layanan.

   

    $$
    \text{Retention\ Rate} = \frac{\text{User\ Aktif\ Akhir\ Periode}}{\text{User\ Awal\ Periode}} \times 100%
    $$
    
    Semakin tinggi retention, biasanya user semakin loyal dan puas.

### **Addition**

Catatan:

- **Total reach**  
  
    artinya:

    > jumlah orang unik yang berhasil melihat konten.

    Jadi fokusnya ke:

    * **berapa orang berbeda** yang melihat,
    * bukan berapa kali halaman dibuka.

    Contoh sederhana:

    Misal:

    * Aldira lihat postingan 5 kali
    * Budi lihat 2 kali
    * Citra lihat 1 kali

    Maka:

    * **Impression** = 8 kali tampil
    * **Reach** = 3 orang unik


- Bedanya Reach vs Impression

    - Reach

        $$
        \text{Jumlah user unik}
        $$

        Contoh:
        
        > 100 orang berbeda melihat iklan
        
        Berarti:
        $$
        Reach = 100
        $$


    - Impression

        $$
        \text{Total tampilan}
        $$

        Kalau 1 orang melihat berkali-kali, tetap dihitung semua.

        Contoh:

        > total tampil 350 kali

        Berarti:
        $$
        Impression = 350
        $$


- Dalam Engagement Rate

    Biasanya:

    $$
    Engagement\ Rate =
    \frac{Total\ Interaction}{Total\ Reach} \times 100%
    $$

    Misal:

    * 1000 orang melihat postingan
    * 120 like/comment/share

    $$
    = \frac{120}{1000} \times 100%
    = 12%
    $$

    Artinya:

    > dari semua orang yang melihat, 12% melakukan interaksi.


- Kalau Website?

    Tergantung definisinya.

    Kadang:

    * reach = unique visitors
    * impression = page views

    Jadi:

    * **unique visitors** → orang unik
    * **page views** → total buka halaman


- Intuisi Mudah

    * **Reach** → “berapa orang yang kena”
    * **Impression** → “berapa kali tampil”
    * **Page view** → “berapa kali halaman dibuka”


## **Transactional Segmentation**

### **Introduction**


Konsep
> Mengelompokkan customer berdasarkan pola transaksi dan uang yang mereka keluarkan.

Intinya:

> “Cara dia belanja seperti apa?”

Fokusnya lebih ke:

- frekuensi belanja,
- total uang,
- nilai transaksi,
- terakhir belanja kapan.

Tujuan:
> Mengetahui nilai bisnis dari setiap customer.

Metric yang sering dipakai:

- Total spending
- Average order value (AOV)
- Purchase frequency
- Recency
- Monetary value
- Customer Lifetime Value (CLV)
- Revenue contribution

Contoh:

- Customer belanja tiap minggu → loyal.
- Customer sekali belanja tapi mahal → high spender.
- Sudah 6 bulan tidak transaksi → risiko churn.

Insight yang bisa didapat:

- Mengetahui High Value Customer.
- Mengetahui Customer loyal dan Customer pasif.
- Mendeteksi customer berpotensi churn.
- Estimasi Customer Lifetime Value (CLV).

Biasanya segmentation ini dekat dengan:

- RFM Analysis,
- CLV,
- churn analysis.

### **Metric**

Metric yang digunakan:

- Total Spending  
    Total uang yang dihabiskan customer selama periode tertentu.  

    $$
    \text{Total\ Spending} = \sum \text{Seluruh\ Transaksi\ Customer}
    $$

    
    Mengukur nilai total customer bagi bisnis.

- Average Order Value (AOV)  
    Rata-rata nilai transaksi setiap pembelian.  

    $$
    \text{AOV} = \frac{\text{Total\ Revenue}}{\text{Jumlah\ Order}}
    $$

    
    Semakin tinggi AOV, semakin besar rata-rata pengeluaran customer per transaksi.

- Purchase Frequency  
    Seberapa sering customer melakukan pembelian.  

    $$
    \text{Purchase\ Frequency} = \frac{\text{Jumlah\ Transaksi}}{\text{Jumlah\ Customer}}
    $$

    
    Mengukur loyalitas dan kebiasaan belanja customer.

- Recency  
    Jarak waktu sejak transaksi terakhir customer.  

    $$
    \text{Recency} = \text{Tanggal\ Sekarang - Tanggal\ Pembelian\ Terakhir}
    $$

    
    Semakin kecil recency, semakin aktif customer tersebut.

- Monetary Value  
    Total nilai uang yang dihasilkan customer.  

    $$
    \text{Monetary\ Value} = \sum \text{Total\ Pembelian}
    $$

    
    Digunakan untuk mencari high-value customer.

- Customer Lifetime Value (CLV)  
    Estimasi total revenue yang dihasilkan customer selama menjadi pelanggan.

    $$
    \text{CLV} = \text{AOV} \times \text{Purchase\ Frequency} \times \text{Customer\ Lifespan}
    $$

    
    Customer dengan CLV tinggi biasanya diprioritaskan bisnis karena paling menguntungkan.

- Revenue Contribution  
    Persentase kontribusi revenue dari customer atau segmen tertentu.  

    $$
    \text{Revenue\ Contribution} = \frac{\text{Revenue\ Customer}}{\text{Total\ Revenue}} \times 100%
    $$

    
    Mengetahui customer atau segmen mana yang paling besar menyumbang pendapatan bisnis.


### **RFM Analysis**

**Metrik**  
Di sini kita pakai konsep klasik yang sangat kuat:

> **RFM + CLV**

Recency (R) → “Seberapa baru dia aktif?”

- **Definisi:**  
    Waktu sejak terakhir pelanggan melakukan transaksi

- **Intuisi:**
    > Semakin baru → semakin “hangat”  

    Analoginya:
    * Orang yang baru beli kemarin → masih ingat brand kamu
    * Orang yang terakhir beli 6 bulan lalu → hampir lupa
- **Makna bisnis:**
    * Recency kecil → pelanggan aktif
    * Recency besar → pelanggan mulai hilang
- **Kontribusi ke tujuan:**  
        ✔ Deteksi pelanggan aktif vs hampir churn

Frequency (F) → “Seberapa sering dia beli?”

- **Definisi:**  
    Jumlah transaksi dalam periode tertentu

- **Intuisi:**
    > Seberapa “loyal” pelanggan itu

    Contoh:
    * 1x beli → belum tentu loyal
    * 10x beli → sangat loyal

- **Makna bisnis:**  
    * Frequency tinggi → pelanggan repeat
    * Frequency rendah → pelanggan pasif
     
- **Kontribusi ke tujuan:**  
     ✔ Identifikasi loyal customer  
     ✔ Evaluasi engagement

Monetary (M) → “Seberapa besar dia kontribusi uang?”

- **Definisi:**   
    Total uang yang dibelanjakan pelanggan
- **Intuisi:**  
    > Tidak semua pelanggan punya nilai yang sama

    Contoh:
    * Customer A: 10x beli, total 1 juta
    * Customer B: 2x beli, total 5 juta
    Siapa lebih penting? Bisa jadi B

- **Makna bisnis:**  
    Monetary tinggi → high value customer

- **Kontribusi ke tujuan:**  
    ✔ Fokus ke pelanggan yang paling menghasilkan uang

Kombinasi RFM

Yang penting bukan masing-masing metrik, tapi **kombinasinya**.

Contoh:

| Recency | Frequency | Monetary | Arti                |
| ------- | --------- | -------- | ------------------- |
| rendah  | tinggi    | tinggi   | ⭐ pelanggan terbaik |
| tinggi  | tinggi    | tinggi   | ⚠️ mulai churn      |
| rendah  | rendah    | rendah   | 🆕 pelanggan baru   |
| tinggi  | rendah    | rendah   | ❌ hampir hilang     |

Di sini mulai terbentuk **segmentasi nyata**

Customer Lifetime Value (CLV)

- **Definisi:**  
    Estimasi total nilai yang akan diberikan pelanggan selama hubungan dengan bisnis

- **Intuisi:**

    > Jangan hanya lihat masa lalu, tapi masa depan


    Secara konsep:

    $$
    CLV = \text{Average Purchase Value} \times \text{Purchase Frequency} \times \text{Customer Lifespan}
    $$

- **Makna:**

    * CLV tinggi → pelanggan sangat berharga jangka panjang
    * CLV rendah → tidak terlalu prioritas

- **Kontribusi ke tujuan:**

    * ✔ Prioritas strategi retention
    * ✔ Alokasi budget marketing

 Dari Metrik → Segmentasi Nyata

Sekarang kita ubah angka jadi kategori:



- High Value Customer

    * Recency: rendah (baru beli)
    * Frequency: tinggi
    * Monetary: tinggi

    **Intuisi:**

    > Ini “aset utama” bisnis

    **Strategi:**

    * VIP treatment
    * loyalty program
    * jangan terlalu banyak diskon (mereka sudah suka)

    $$ \ $$


- Loyal Customer

    * Frequency tinggi
    * Monetary cukup

    **Intuisi:**

    > Mereka sering balik, tapi belum maksimal nilai

    **Strategi:**

    * upselling
    * bundling

    $$ \ $$

- Churn Risk Customer

    * Recency tinggi (lama tidak beli)

    **Intuisi:**

    > Mereka “hampir hilang”

    **Strategi:**

    * retargeting
    * diskon khusus
    * email reminder

    $$ \ $$
    

- New Customer

    * Recency rendah
    * Frequency rendah

    **Intuisi:**

    > Masih tahap kenalan

    **Strategi:**

    * onboarding
    * edukasi produk


Dari Segmentasi → Keputusan

- Contoh Kasus:

    Data menunjukkan:

    * Banyak pelanggan masuk kategori **Churn Risk**
    * High Value Customer jumlahnya sedikit tapi kontribusinya besar



- Insight:

    * Bisnis kehilangan banyak pelanggan diam-diam
    * Revenue bergantung pada sedikit pelanggan

    

- Keputusan:

    * Fokus campaign untuk **retention**
    * Buat program khusus untuk **High Value Customer**
    * Kurangi budget ke pelanggan low value



Intuisi

> Customer segmentation itu bukan sekadar “mengelompokkan”, tapi “memahami siapa yang paling penting dan bagaimana memperlakukan mereka secara berbeda”


Alur Berpikir

-  **Masalah:**  
   Semua pelanggan diperlakukan sama

-  **Tujuan:**  
   Memahami perbedaan pelanggan

-  **Solusi:**  
   Gunakan metrik (RFM + CLV)

-  **Peran Metrik:**  

   * Mengukur perilaku pelanggan
   * Membentuk segmentasi
   * Menjadi dasar strategi

-  **Output:**  
   Segment → Strategy → Profit

## **Psychographic Segmentation**

### **Introduction**

Konsep
> Mengelompokkan customer berdasarkan gaya hidup, minat, kepribadian, atau cara berpikir.

Intinya:

> “Customer ini orang seperti apa?”

Ini lebih dalam dibanding demographic.

Tujuan:
> Memahami motivasi emosional customer.

Metric yang sering dipakai:

- Interests
- Lifestyle category
- Personality traits
- Hobbies
- Values/beliefs
- Social media activity
- Preferred content category

Contoh:

- Suka hidup sehat.
- Pecinta teknologi.
- Suka produk ramah lingkungan.
- Suka gaya hidup minimalis.

Insight yang bisa didapat:

- Cara marketing yang lebih personal.
- Brand positioning jadi lebih kuat.
- Brand image yang cocok.
- Produk bisa disesuaikan dengan gaya hidup customer.

Contoh:
> Orang yang peduli lingkungan biasanya lebih tertarik produk eco-friendly walaupun lebih mahal.

### **Metric**

- Interests  
    Topik atau kategori yang paling diminati user.

    **Rumus sederhana:**

    $$
    \text{Interest\ Score} =
    \frac{\text{Interaksi\ pada\ Kategori}}{\text{Total\ Interaksi}}
    $$


    Mengetahui minat utama user, misalnya gaming, olahraga, atau teknologi.


- Lifestyle Category  
    Kategori gaya hidup user berdasarkan pola aktivitas dan preferensi.

    **Contoh kategori:**

    * healthy lifestyle
    * traveler
    * gamer
    * minimalist

    **Rumus (skoring sederhana):**

    $$
    \text{Lifestyle\ Score} =
    \frac{\text{Aktivitas\ Terkait\ Lifestyle}}{\text{Total\ Aktivitas}}
    $$

    Membantu personalisasi produk dan marketing.


- Personality Traits  
    Karakter atau kecenderungan perilaku user.

    Contoh:

    * impulsive buyer
    * risk taker
    * introvert/extrovert

    Biasanya diukur menggunakan:

    * survey,
    * behavioral scoring,
    * atau machine learning.

    **Rumus sederhana:**

    $$
    \text{Trait\ Score} =
    \frac{\text{Perilaku\ Terkait\ Trait}}{\text{Total\ Perilaku}}
    $$


    Membantu memahami pola keputusan customer.


- Hobbies  
    Aktivitas yang sering dilakukan atau disukai user.

    **Rumus sederhana:**

    $$
    \text{Hobby\ Preference} =
    \frac{\text{Interaksi\ pada\ Hobi}}{\text{Total\ Interaksi}}
    $$


    Digunakan untuk recommendation system dan targeted ads.


- Values / Beliefs  
    Nilai hidup atau prinsip yang dianggap penting oleh user.

    Contoh:

    * peduli lingkungan,
    * hemat,
    * premium quality,
    * kesehatan.

    Biasanya diperoleh dari:

    * survey,
    * riwayat aktivitas,
    * pola pembelian.

    **Rumus sederhana (scoring):**

    $$
    \text{Value\ Score} =
    \frac{\text{Aktivitas\ Sesuai\ Nilai}}{\text{Total\ Aktivitas}}
    $$


    Membantu brand positioning dan emotional marketing.


- Social Media Activity  
    Tingkat aktivitas user di media sosial.

    **Rumus:**

    $$
    \text{Social\ Activity\ Rate} =
    \frac{\text{Total\ Interaksi}}{\text{Total\ Post/View}}
    $$

    Interaksi bisa berupa:

    * like,
    * comment,
    * share,
    * save.


    Mengukur keaktifan dan engagement user.


- Preferred Content Category  
    Kategori konten yang paling sering dikonsumsi user.

    **Rumus:**

    $$
    \text{Preference\ Ratio} =
    \frac{\text{View\ Kategori}}{\text{Total\ View}}
    $$


    Membantu sistem rekomendasi memahami preferensi user.
