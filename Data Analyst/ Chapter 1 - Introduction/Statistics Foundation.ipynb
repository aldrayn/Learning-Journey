## **Statistics**

## **Foundation**



Karena:

> Naik sedikit belum tentu benar-benar lebih baik.

Bisa saja hanya kebetulan/random, oleh karena itu dibutuhkan analisis statistik.

**Konsep Dasar - Hipotesis**


- Null Hypothesis $(H_0)$

    Tidak ada perbedaan antara A dan B.  

    $$H_0​:μ_A​=μ_B​$$

- Alternative Hypothesis $(H_1)$

    Ada perbedaan antara A dan B.  

    $$H_1	​:μ_A \ne μ_B$$
    
    Kadang juga:
    
    $$H1	​:μ_B	​>μ_A$$
    
    ​(jika ingin membuktikan versi baru lebih baik)

##### **P-value**  

Untuk:
> Mengukur probabilitas hasil terjadi secara kebetulan.


Kalau:  

$
p < 0.05
$  

Artinya:
* kemungkinan hasil terjadi karena kebetulan sangat kecil,
* maka kita percaya treatment memang berbeda.


Decision Rule

- Jika:  

    $
    p < 0.05
    $

    Maka:
    - Tolak (H_0)
    - Ada perbedaan signifikan.
- Jika:
    
    $
    p \geq 0.05
    $  

    Maka:  
    - Gagal menolak (H_0)  
    - Belum ada bukti cukup.

##### **Standard Error (SE)**  

Untuk:

> Mengukur seberapa stabil hasil sampel kita.

SE membantu menjawab:

> “Kalau eksperimen diulang terus, hasilnya bakal berubah banyak atau tidak?”  

Rumus:  

$$
SE =
\sqrt{
\frac{p(1-p)}{n}
}
$$

Keterangan:
- $p$ : Proporsi



Contoh:

Misal conversion rate:

> p = 0.05

Jumlah user:

> n = 100


Perhitungan:

$$
SE =
\sqrt{
\frac{0.05(1-0.05)}{100}
}=\sqrt{\frac{0.0475}{100}}=\sqrt{0.000475}=0.0218
$$

Standard Error sekitar:

2.18%

Jadi kalau eksperimen diulang berkali-kali:

> conversion rate kemungkinan akan “goyang” sekitar ±2%.

Misalnya:

* hari ini: 5%
* besok: 7%
* lusa: 4%

Masih cukup wajar karena SE masih lumayan besar.

Kalau user diperbesar:

$$n = 10{,}000$$

maka:

$$
SE =
\sqrt{
\frac{0.05(0.95)}{10000}
}
=0.00218=0.218%
$$

Artinya:

> hasil jauh lebih stabil dan tidak mudah berubah-ubah.


Interpretasi
- Kalau user sangat banyak:

    - hasil lebih konsisten,
    - tidak gampang berubah.

    Berarti SE kecil.

- Kalau $n$ besar, SE kecil → hasil lebih terpercaya.
- Kalau $n$ kecil, SE besar → hasil masih noisy/tidak stabil.

##### **Confidence Interval (CI)**

###### <span style = "font-size:16px">**Introduction**</span>


Untuk:
> Memberikan range perkiraan nilai sebenarnya di populasi.

Karena data yang kita ambil cuma sampel, kita tidak pernah benar-benar tahu nilai asli populasi.  

CI membantu menjawab:

> “Kemungkinan nilai aslinya ada di sekitar mana?”  

Rumus

$$
CI =
\bar{x}
\pm
Z \cdot \frac{\sigma}{\sqrt{n}}
$$
Keterangan:
- $Z$ : Nilai dari distribusi normal $\to$ digunakan untuk menentukan tingkat keyakinan.


Interpretasi
- CI itu seperti:

    > “zona kemungkinan nilai asli.”

- Semakin sempit CI:

    - semakin yakin,
    - estimasi makin stabil.

###### <span style="font-size: 16px;">**Example**</span>

Data Sampel

Misal hasil conversion user:

| User | Purchase |
| ---- | -------- |
| U1   | 1        |
| U2   | 0        |
| U3   | 0        |
| U4   | 1        |
| U5   | 0        |
| U6   | 0        |
| U7   | 1        |
| U8   | 0        |
| U9   | 0        |
| U10  | 0        |

Proses Perhitungan

- Step 1 — Hitung Conversion Rate Sampel

    Total purchase:

    $$
    3
    $$

    Jumlah user:

    $$
    10
    $$

    Maka:

    $$
    \bar{x} = \frac{3}{10}
    =0.3
    $$

    Conversion rate sampel:

    $$
    30%
    $$

- Step 2 — Hitung Standard Error

    Rumus:

    $$
    SE =
    \sqrt{
    \frac{p(1-p)}{n}
    }
    $$

    Substitusi:

    $$
    SE =
    \sqrt{
    \frac{0.3(1-0.3)}{10}
    }
    =
    \sqrt{
    \frac{0.21}{10}
    }
    \sqrt{0.021}
    =0.145
    $$

- Step 3 — Hitung Confidence Interval

    Gunakan 95% confidence:

    $$
    Z=1.96
    $$

    Rumus:

    $$
    CI =
    \bar{x}
    \pm
    Z \cdot SE
    $$

    Substitusi:

    $$
    CI =
    0.3
    \pm
    1.96(0.145)=
    0.3
    \pm
    0.284
    $$

    Hasil

    Lower bound:

    $$
    0.016
    $$

    Upper bound:

    $$
    0.584
    $$

- Confidence Interval

    $$1.6\%, \ 58.4\%$$

    Artinya

    Walaupun sampel menunjukkan conversion rate:

    $$
    30%
    $$

    nilai asli populasi kemungkinan berada di sekitar:

    $$
    1.6% \text{ sampai } 58.4%
    $$

    Intuisi

    Karena jumlah user sangat sedikit:

    $$
    n=10
    $$

    range menjadi sangat lebar.

    Artinya:

    > estimasi masih belum stabil dan sangat noisy.


###### <span style = "font-size:16px;">**Addition**</span>

**Hubungan CI dan SE**

> CI dibangun menggunakan SE.

Karena:

$$CI=\text{Estimasi}±(Z×SE)$$

Artinya:

- SE menentukan lebar CI.
- Jika SE besar

    - CI lebar
    - ketidakpastian tinggi.

- Jika SE kecil

    - CI sempit
    - estimasi lebih yakin/stabil.

##### **Z-Score**



> Mengukur seberapa jauh perbedaan dari rata-rata.  

Rumus
$$
Z =
\frac{\hat{p}_1 - \hat{p}_2}
{SE}
$$
Keterangan:
- $\hat{p}_1 \to$ Conversion rate group pertama.
- $\hat{p}_2 \to$ Conversion rate group kedua.

Intuisi:

$$Z= \frac{\text{Perbedaan Asli}}{\text{Noise}}$$

Z-score mencoba menjawab:

> “Perbedaan conversion ini besar banget atau sebenarnya kecil dibanding variasi/noise data?”

Data

| ---- | ----- | ----------- | -------- | ------ |
| U1   | A     | Fashion     | 0        | 0      |
| U2   | A     | Fashion     | 1        | 120    |
| U3   | A     | Electronics | 0        | 0      |
| U4   | A     | Beauty      | 1        | 150    |
| U5   | A     | Fashion     | 0        | 0      |
| U6   | A     | Electronics | 1        | 180    |
| U7   | B     | Fashion     | 1        | 140    |
| U8   | B     | Beauty      | 1        | 170    |
| U9   | B     | Electronics | 1        | 220    |
| U10  | B     | Fashion     | 0        | 0      |

#### **T-test**

Definisi

- T-Test digunakan untuk membandingkan:

    > rata-rata dua group.

- Sekarang pertanyaannya berubah.

    Bukan:
    
    > “berapa orang membeli?”
    
    Tapi:
    
    > “berapa besar rata-rata transaksi?”

Rumus T-Test

$$t = \frac{\bar{x}_1-\bar{x}_2}{\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}}}$$

T-Test mengukur:

> “berapa besar perbedaan mean dibanding variasi data.”

Proses Perhitungan

- Fokus Data

    Sekarang hanya memakai:

    | Group | Amount |
    | ----- | ------ |
    | A     | 0      |
    | A     | 120    |
    | A     | 0      |
    | A     | 150    |
    | A     | 0      |
    | A     | 180    |
    | B     | 140    |
    | B     | 170    |
    | B     | 220    |
    | B     | 0      |
    | B     | 200    |
    | B     | 250    |

    Karena sekarang fokusnya:

    > nilai numerik.

    Bukan binary lagi.

- Hitung Mean

    Group A
    
    $$
    \bar x_A=\frac{450}{6}
    =75
    $$
    
    
    Group B
    
    $$
    \bar x_B=\frac{980}{6}
    =163.3
    $$

    Keterangan:
    > Rata-rata transaksi group B lebih besar.


    Pertanyaan Besarnya

    > apakah selisih ini cukup besar?

    atau:

    > hanya kebetulan?


    Hipotesis

    $$
    H_0 : \mu_A=\mu_B
    $$

    $$
    H_1 : \mu_A \ne \mu_B
    $$


- Hitung Variance  

    Variance mengukur:

    > seberapa menyebar data dari rata-ratanya.


    Rumus Variance

    $$s^2 = \frac{\sum (x_i-\bar{x})^2}{n-1}$$


    Variance menghitung:

    - jarak tiap data ke mean,
    - lalu dikuadratkan,
    - lalu dirata-ratakan.

    Kalau variance besar:

    > data noisy.

    Kalau variance kecil:

    > data stabil.

- Hitung T-test

    Misal didapat:

    $$
    t=-1.78
    $$


- Interpretasi

    Biasanya:

    $$
    |t|>2
    $$

    > baru signifikan.

    Karena:

    $$
    1.78 < 2
    $$

    maka:

    > belum signifikan.

- Kesimpulan
  
  Walaupun mean naik:

  > 75 → 163

  datanya masih terlalu noisy.


Kapan T-Test Dipakai?

| Kasus                    | Cocok? |
| ------------------------ | ------ |
| average revenue          | ✅      |
| average session duration | ✅      |
| conversion rate          | ❌      |

#### **Mann–Whitney U**

##### **Introduction**

Latar Belakang

- Uji t-test punya masalah ketika data tidak normal atau banyak outlier, karena dia bergantung pada rata-rata yang mudah “ketarik” nilai ekstrem.

    Oleh karena itu, cara berpikirnya diubah:
    > yang awalnya membandingkan rata-rata, diganti menjadi membandingkan posisi (ranking).

    Artinya, kita tidak lagi peduli seberapa besar nilainya,

- tapi lebih ke:

    > “data di grup A secara umum lebih tinggi dari grup B atau tidak?”
    
    Dengan cara ini, hasil jadi lebih stabil dan tidak mudah tertipu outlier.

Intuisi super sederhana

Bayangin lo lagi lomba lari:

* Semua peserta (A & B) digabung
* Dikasih ranking 1, 2, 3, …

Lalu lo tanya:

> “Anak A sering finish di depan atau belakang?”

Kalau A sering di depan
- → ranking tinggi
- → ( R_1 ) besar
- → U kecil (atau besar tergantung sisi)
- → beda signifikan

Kalau campur → gak beda

Kenapa ranking bisa ganti nilai asli?

- Masalah di nilai asli (real data)

    Contoh sales:

    - A = [10, 20, 30]  
    - B = [15, 25, **1000**] ← outlier gila

    Kalau pakai mean:

    * B jadi kelihatan “jauh lebih besar”
    * padahal cuma 1 data aneh

- Solusi: ubah ke ranking

    Gabung & ranking:  

    10(1), 15(2), 20(3), 25(4), 30(5), 1000(6)

    > Yang 1000 cuma jadi ranking **6**  
    > gak “merusak sistem”

    Intuisi mendalam

    > Ranking itu cuma peduli **urutan**, bukan “jarak”

- Bandingin:

    | Nilai      | Ranking  |
    | ---------- | -------- |
    | 30 vs 25   | beda 5   |
    | 1000 vs 25 | beda 975 |

    Keterangan:
    - di mean: beda jauh banget
    - di ranking: cuma beda 2 posisi

Rumus 

$$U_1 = n_1 n_2 + \frac{n_1(n_1+1)}{2} - R_1$$


Keterangan:
- $n_1 n_2 \to $ **Total kemungkinan perbandingan antar grup**

  Bahasa manusia:

  > “Kalau semua orang di grup A dibandingin satu-satu sama grup B, ada berapa pasangan?”

  Contoh:

  * A = 3 data
  * B = 3 data
  
  total perbandingan = 9

- $\frac{n_1(n_1+1)}{2}\to$ **Total ranking ideal kalau grup A selalu paling kecil**

  Bahasa manusia:

  > “Kalau semua data A ada di posisi paling bawah (ranking kecil semua), total rankingnya harusnya berapa?”

  Ini semacam:

  * baseline / kondisi “A kalah total”

- $R_1$ (jumlah ranking grup A) $\to$ **seberapa tinggi posisi data A di dunia nyata**

  Bahasa manusia:

  > “Setelah semua data digabung dan diurutkan, total posisi (ranking) A itu tinggi atau rendah?”

- Gabungan semuanya

  > U itu ngukur:
  > **“Seberapa jauh posisi ranking A dari kondisi ‘A selalu kalah’”**

Interpretasi matematis

U bisa dipahami sebagai:

> jumlah kemenangan A saat dibandingkan dengan B

Bayangin:

* bandingin semua pasangan A vs B
* hitung:

  * A > B → menang
  * A < B → kalah

👉 U = total kemenangan

##### **Addition**

Insight penting

> Mann-Whitney sebenarnya nanya:

> “Seberapa sering nilai di A lebih besar dari nilai di B?”

Bukan:

> “Seberapa besar perbedaannya?”

Kenapa ini powerful buat data real?

Karena di dunia nyata:

* data **jarang normal**
* banyak:

  * outlier
  * skew
  * noise

👉 Ranking bikin:

* lebih stabil
* lebih robust
* gak gampang dibohongi data ekstrem

#### **Chi-Squared Test**

##### **Introduction**

Definisi

Chi-Square digunakan untuk menguji:

> apakah dua kategori saling berhubungan.


Contoh Pertanyaan

> apakah group memengaruhi purchase?

Rumus Chi-Square

$$\chi^2 = \sum \frac{(O-E)^2}{E}$$

Keterangan

* (O) = observed value (data asli)
* (E) = expected value (nilai harapan)

Chi-Square mengukur:

> “seberapa jauh data asli dari kondisi normal/tidak ada hubungan.”




Proses Perhitungan

- Data berubah menjadi contingency table.
  
    | Group | Purchase | No Purchase |
    | ----- | -------- | ----------- |
    | A     | 3        | 3           |
    | B     | 5        | 1           |


    Kelihatannya:

    > group B lebih banyak purchase.

    Tapi:

    > apakah benar ada hubungan?


    Hipotesis

    - $H_0 :$ group dan purchase tidak berhubungan.
    - $H_1 :$ group dan purchase berhubungan.
- Expected Value

    Expected value adalah:

    > angka yang diharapkan jika tidak ada hubungan.

    Rumus

    $$Expected = \frac{Row\ Total \times Column\ Total}{Grand\ Total}$$


    Contoh

    Expected untuk A-Purchase:

    $$
    E=\frac{6 \times 8}{12}
    =4
    $$

    Kalau tidak ada hubungan:

    > harusnya A punya sekitar 4 purchase.

- Hitung Chi-square
  Hasil

    $$
    \chi^2=1.5
    $$


- Interpretasi

    Threshold:

    $$
    3.84
    $$

    Karena:

    $$
    1.5 < 3.84
    $$

    maka:

    > belum signifikan.

    Artinya:
    > Perbedaan purchase masih mungkin terjadi karena kebetulan.

Kapan Chi-Square Dipakai?

| Kasus              | Cocok? |
| ------------------ | ------ |
| gender vs purchase | ✅      |
| country vs churn   | ✅      |
| average revenue    | ❌      |

##### **Addition**

Konsep

Expected value adalah:

> jumlah yang seharusnya muncul jika kedua variabel benar-benar tidak berhubungan.
bahasa

Atau:

> “kalau semuanya random/tidak ada pengaruh, harusnya angkanya berapa?”

Hipotesis Null

> $H_0$ : Group dan purchase tidak berhubungan.

Artinya:

> group A dan B harus punya proporsi purchase yang mirip dengan proporsi global.

Contoh 1 - Tidak Ada Hubungan

- Data
    | Group | Purchase | No Purchase | Total |
    | ----- | -------- | ----------- | ----- |
    | A     | 40       | 60          | 100   |
    | B     | 40       | 60          | 100   |

    Keterangan:
    - Total purchase: 80
    - Grand total: 200
- Proporsi Global

    Purchase global:

    $$\frac{80}{200}=0.4$$

    Artinya:

    > secara global, 40% orang membeli.

- Expected Value

    Kalau group dan purchase tidak berhubungan:

    > maka setiap group harus mengikuti proporsi global 40%.

    Maka:
    - Untuk Group A:

        $100×0.4=40$

    - Untuk Group B:

        $100×0.4=40$

- Hasil
    | Group      | Observed | Expected |
    | ---------- | -------- | -------- |
    | A Purchase | 40       | 40       |
    | B Purchase | 40       | 40       |

    Keterangan 
  
    > Observed = Expected.

    Artinya:

    > data nyata sesuai dengan kondisi “tidak ada hubungan”.

    Maka:

    > kemungkinan memang independen/random.

Contoh 2 - Ada Hubungan

| Group | Purchase | No Purchase | Total |
| ----- | -------- | ----------- | ----- |
| A     | 20       | 80          | 100   |
| B     | 60       | 40          | 100   |

Keterangan:
- Total purchase tetap = 80
- Grand total tetap = 200

Proporsi Global Tetap Sama

$$\frac{20+60}{200}=\frac{80}{200}=0.4$$

Jadi expected value tetap:

| Group | Expected Purchase |
| ----- | ----------------- |
| A     | 40                |
| B     | 40                |

Tapi Observed Berbeda Jauh

| Group      | Observed | Expected |
| ---------- | -------- | -------- |
| A Purchase | 20       | 40       |
| B Purchase | 60       | 40       |

Kalau benar tidak ada hubungan:

- A harusnya sekitar 40 purchase,
- B harusnya sekitar 40 purchase.

Tapi nyata:

- A cuma 20,
- B malah 60.

Artinya:

> distribusi nyata menyimpang jauh dari kondisi independen.

Kesimpulan:
> Purchase ternyata tidak tersebar “normal” karena group B jauh lebih sering membeli.

Berarti:

> kemungkinan group memang memengaruhi purchase.

#### **Z-Test**

Z-Test digunakan untuk menguji:

> apakah dua proporsi berbeda secara signifikan.

Biasanya dipakai untuk:

- conversion rate,
- CTR,
- open rate email,
- fraud rate.

Bayangkan:

| Group | Conversion |
| ----- | ---------- |
| A     | 50%        |
| B     | 83%        |

Pertanyaannya:

> apakah selisih ini benar-benar besar?

atau:

> cuma kebetulan karena sample kecil?

Jadi statistik mencoba mengukur:

> “berapa jauh perbedaan data dibanding noise normal?”

Rumus

$$z = \frac{p_1-p_2}{\sqrt{p(1-p)(\frac1{n_1}+\frac1{n_2})}}$$

Formula ini mengukur:

> “selisih proporsi dibanding variasi normal data.”

Proses perhitungan

- Transformasi Data  
    Dari data awal:

    | User | Group | Purchase |
    | ---- | ----- | -------- |

    menjadi

    | Group | Purchase | Total |
    | ----- | -------- | ----- |
    | A     | 3        | 6     |
    | B     | 5        | 6     |

    Data berubah strukturnya karena Z-Test fokus pada:

    > proporsi keberhasilan.

    Bukan transaksi individual lagi.

- Hitung Proporsi
  
  Matriks
  - Group A
    $$p_A = \frac{3}{6} = 0.5$$

    Artinya:
    > 50% user di group A membeli.

  - Group B
    $$p_B = \frac{5}{6} = 0.833$$

    Artinya:
    > 83.3% user di group B membeli.

   Hipotesis


    - Null Hypothesis

        $$
        H_0 : p_A = p_B
        $$

        Artinya:

        > conversion sebenarnya sama.


    - Alternative Hypothesis

        $$
        H_1 : p_A \ne p_B
        $$

        Artinya:

        > conversion berbeda.

- Pooled Proportion

    $$
    p=\frac{3+5}{6+6}
    =\frac{8}{12}
    =0.667
    $$

    Pooled proportion adalah:
    
    > estimasi conversion global seluruh data.


- Standard Error

    $$
    SE=\sqrt{0.667(1-0.667)(\frac16+\frac16)}
    =\sqrt{0.074}
    =0.272
    $$

    Standard Error mengukur:

    > seberapa besar noise alami data.


- Hitung Z Score

    $$
    z=\frac{0.5-0.833}{0.272}
    =-1.22
    $$


    Perbedaan conversion hanya:

    > 1.22 kali noise normal.


- Interpretasi

    Biasanya threshold:

    $$
    |z| > 1.96
    $$

    baru signifikan.

    Karena:

    $$
    1.22 < 1.96
    $$

    maka:

    > belum signifikan.


- Intuisi Besar

    Walaupun conversion naik:

    > 50% → 83% (sample terlalu kecil)

    Statistik berkata:

    > “belum cukup bukti.”






Kapan Z-Test Dipakai?

| Kasus           | Cocok? |
| --------------- | ------ |
| conversion rate | ✅      |
| CTR             | ✅      |
| fraud rate      | ✅      |
| average revenue | ❌      |

#### **ANOVA**

###### <span style="font-size:18px">**Introduction**</span>

Definisi

- ANOVA digunakan untuk membandingkan:

    > lebih dari dua rata-rata.

- Pertanyaan

    > apakah kategori produk memiliki rata-rata transaksi berbeda?

Ide Besarnya

- Kalau:

    - mean antar kategori jauh berbeda,
    - tapi data dalam kategori stabil,

- maka:

    kemungkinan kategori memang berpengaruh.

Hipotesis

- $H_0 :\mu_1=\mu_2=\mu_3 \to$ semua rata-rata sebenarnya sama.
- $H_1 :$ minimal ada satu mean berbeda.

Konsep 1 - Variasi Antar Group (SS Between)

- Menggambarkan:

    > seberapa besar perbedaan karakteristik antar group.

- Pada ANOVA biasanya:

    > perbedaan rata-rata antar group.

- Rumus
  $$SS_{between} = \sum n_i(\bar{x}_i-\bar{x})^2$$

  Keterangan:
  - $\bar{x}_i =$ rata-rata suatu kelompok
  - $\bar{x} =$ rata-rata global
- Logikanya

    Kalau tiap group memang berbeda:

    | Group | Mean |
    | ----- | ---- |
    | A     | 50   |
    | B     | 100  |
    | C     | 150  |

    maka jarak tiap mean ke grand mean besar.

    Akibatnya:

    > $SS_{between}$ besar.

    Artinya

    > Signal/perbedaan antar group kuat.

    Kemungkinan:

    > group memang punya pengaruh terhadap nilai data.

Konsep 2 - Variasi Dalam Group (SS Within)

- Menggambarkan:

    > seberapa besar variasi/noise di dalam masing-masing group.

- Rumus
  $$SS_{within}=\sum (x_{ij}-\bar{x}_i)^2$$

  Keterangan:
  - $\bar{x}_i =$ rata-rata nilai suatu kelompok
  - $x_{ij} =$ nilai-nilai di dalam kelompok

- Logikanya  
  
  Misal group A  

  $[49,50,51]$ (sangat stabil) berarti $SS_{within}$ kecil.
  
  Tapi kalau 
  
  $[10,50,90]$ (berantakan/noisy) maka  $ SS_{within}$ besar.

    Artinya

    * kecil → data stabil
    * besar → data noisy/random

Konsep 3 - Mean Square

- Mean Square Between

    Menggambarkan:

    > rata-rata kekuatan perbedaan antar group.

    Karena:
    $$
    MS_{between}
    =\frac{SS_{between}}{df_{between}}
    $$

    

- Mean Square Within

    Menggambarkan:

    > rata-rata noise/penyebaran dalam group.

    Karena:
    $$
    MS_{within}
    =\frac{SS_{within}}{df_{within}}
    $$

Konsep 4 - Derajat Kebebasan

Degree of freedom sebenarnya mengukur:

> “berapa banyak informasi independen yang benar-benar bisa bervariasi?”

Di ANOVA ada dua df.

- Between Group
  
    $$df_{between}=k−1$$

    Karena:  
    jika ada k mean group, setelah grand mean tetap/fixed:

    - hanya k−1 mean yang benar-benar bebas bergerak.
    - Mean terakhir “terikat” agar total tetap konsisten.
    
    Artinya:
    > berapa banyak mean group yang benar-benar bisa berbeda bebas.


- Within Group
    
    $$df_{within}	​=N−k$$

    Karena:

    - setiap group punya constraint mean sendiri.
    - Dalam tiap group:
        - satu nilai “dikunci” oleh mean group,
        - sehingga tiap group kehilangan 1 kebebasan.

    Total hilang: 
    
    $k$

    Maka:

    $N−k$

    Artinya:
    > berapa banyak variasi data independen di dalam group.

F Ratio

Menggambarkan:

> apakah signal antar group lebih besar daripada noise dalam group.

Rumus

$$F = \frac{MS_{between}}{MS_{within}}$$

Logikanya

- Kalau:

    - antar group beda jauh,
    - tapi dalam group stabil,

    maka:
    > F besar → kemungkinan perbedaannya nyata.
    
- Tapi:
  > kalau antar group terlihat beda namun tiap group sangat noisy,

    maka:
    > perbedaan itu bisa saja cuma akibat randomness/data kacau.

ANOVA sebenarnya sedang bertanya:

> “Apakah perbedaan antar group benar-benar nyata, atau cuma akibat variasi random data?”


Tambahan

Perbedaan antar group baru dianggap “adil” atau bermakna jika dibandingkan dengan:

> seberapa kacau/noisy data di dalam masing-masing group.

Karena bisa saja mean terlihat berbeda hanya karena:

- ada nilai ekstrem,
- data berantakan,
- variasi random besar.

#### **Kruskal Wallis**

##### **Introduction**

Latar Belakang

- ANOVA punya masalah yang sama ketika data tidak normal atau tidak stabil, karena juga bergantung pada rata-rata.
- Oleh karena itu, pendekatannya diubah seperti Mann–Whitney, dari membandingkan rata-rata antar banyak grup, menjadi membandingkan penyebaran ranking antar grup.

    Artinya, kita melihat:

    > “apakah ada grup yang secara konsisten berada di posisi lebih tinggi atau lebih rendah?”

- Dengan pendekatan ini, perbandingan jadi lebih adil dan robust, terutama untuk data nyata seperti sales yang sering tidak normal.

Intuisi super sederhana

Bayangin:

* Semua data dari semua grup digabung
* Diurutkan → dikasih ranking

Terus lo lihat:

> “Apakah ada grup yang isinya kebanyakan ranking atas / bawah?”

Kalau:

* Semua grup campur → H kecil → gak beda
* Ada grup dominan → H besar → beda signifikan



Contoh simpel

- Kasus gak beda

    A = [10, 50, 90]  
    B = [20, 60, 80]  
    C = [30, 40, 70]

    Keterangan: 
    - Ranking campur semua
    - gak ada yang dominan
    - ❌ tidak signifikan

- Kasus beda

    A = [10, 20, 30]  
    B = [40, 50, 60]  
    C = [70, 80, 90]  

    Ranking:  

    * A bawah
    * B tengah
    * C atas

    jelas beda → ✅ signifikan

Intuisi

Kruskal–Wallis sebenarnya nanya:

> “Apakah ada minimal 1 grup yang secara konsisten
> lebih tinggi / lebih rendah dibanding yang lain?”

Rumus Kruskal–Wallis

$$H = \frac{12}{N(N+1)} \sum_{i=1}^{k} \frac{R_i^2}{n_i} - 3(N+1)$$

Keterangan

- $N \to$ total semua data (gabungan semua grup)

    > “Total peserta lomba”

- $k \to$ jumlah grup

    > “Ada berapa tim yang dibandingkan”

- $R_i \to$ total ranking di grup ke-i

    > “Kalau semua data digabung & diranking, total posisi grup ini tinggi atau rendah?”

- $n_i \to$ jumlah data di grup ke-i

    > “Jumlah anggota di grup itu”

- $\frac{R_i^2}{n_i} \to$ “kekuatan ranking rata-rata grup”

    Bahasa manusia:

    > “Apakah grup ini cenderung punya ranking tinggi secara konsisten?”

    Kenapa dikuadratkan?
    > biar **perbedaan makin kelihatan jelas** (yang tinggi makin kelihatan dominan)

- $\sum \frac{R_i^2}{n_i} \to$ akumulasi semua grup

    > “Secara keseluruhan, ada grup yang dominan gak?”

- $\frac{12}{N(N+1)}$ dan $-3(N+1) \to$ faktor normalisasi

    Bahasa manusia:

    > “Biar hasilnya bisa dibandingin secara adil (scale-nya rapi)”

- Gabungan semuanya

    > H itu ngukur:
    > **“Seberapa berbeda distribusi ranking antar grup?”**

##### **Addition**

Kenapa ranking bisa ganti nilai asli (lagi, tapi lebih dalam)

- Problem data asli (real case sales)

  Campaign A = [10, 20, 30]  
  Campaign B = [15, 25, 35]  
  Campaign C = [20, 30, **5000**]

  mean:

  * C jadi keliatan paling bagus (karena 5000)
  - > padahal cuma 1 outlier

  Pakai ranking

  - Gabung:
    > 10,15,20,20,25,30,30,35,5000

  - Ranking:
    > 1,2,3,4,5,6,7,8,9

  Keterangan:
  - 5000 cuma jadi ranking 9
  - gak “menghancurkan” distribusi

- Insight inti

  > Ranking itu “menyederhanakan dunia” jadi cuma urutan

  Bukan:

  * seberapa jauh beda

  Tapi:

  * siapa lebih besar dari siapa

Cara mikir paling benar

ANOVA (parametric):

> “Mean grup beda gak?”

Kruskal–Wallis:

> “Distribusi posisi (ranking) beda gak?”

Kapan ini penting banget?

Kruskal-Wallis itu GOAT kalau:

* data sales 💸
* data skew
* banyak outlier
* distribusi beda bentuk (bukan cuma mean)


Kesimpulan yang harus nempel

* Kruskal–Wallis = versi “ranking” dari ANOVA
* Fokus ke:

  > **posisi relatif, bukan nilai absolut**
* H besar → ada perbedaan antar grup
* Robust terhadap outlier & data aneh

- Insight terakhir  

  > Parametric test itu “percaya angka”  
  > Non-parametric itu “percaya urutan”

- Dan di dunia nyata:

  > **urutan sering lebih jujur daripada angka**



Kalau lo mau next step, gue bisa:

* lanjut ke **post-hoc test (habis Kruskal → grup mana yang beda?)**
* atau bikin **studi kasus dashboard sales + Kruskal vs ANOVA (biar keliatan efeknya real)**

Mau lanjut ke mana?


## **Effect Size**


Konsep

- Effect size mengukur:

    > seberapa besar pengaruh/perbedaan antar grup.

- P-value cuma menjawab:

    > “Apakah ada perbedaan?”

- Tapi effect size menjawab:

    > “Perbedaannya BESAR atau kecil?”  

    Misalnya:
    
    | Group | Conversion |
    |-- |------- |
    | A     | 5.00%      |
    | B     | 5.01%      |
    
    Dengan user jutaan 
    
    > perbedaan bisa signifikan.
    
    Tapi…
    
    > peningkatannya hampir tidak berguna secara bisnis.
    
    Nah effect size membantu melihat:
    
    > apakah efeknya meaningful.




**P-value vs Effect Size**  

| P-value                          | Effect Size                    |
|----------------------------- |--------------------------- |
| Apakah signifikan?               | Seberapa besar efeknya?        |
| Dipengaruhi sample size          | Fokus ke besar efek            |
| Bisa signifikan walau efek kecil | Mengukur meaningful atau tidak |

**Jenis**

##### **Cohen's d (T-test)**


**Cohen’s d**

> Salah satu effect size paling populer.

Rumus

$$
d=
\frac{μ_1−μ_2}{σ}
$$	​

	​

Keterangan Variabel
- $μ_1$ : Rata-rata grup pertama $\to$ Biasanya control/A.

- $μ_2$ : Rata-rata grup kedua $\to$ Biasanya treatment/B.

- σ : Standard deviation $\to$ Mengukur variasi data.

Intuisi Formula

$$d= \frac{\text{perbedaan}}{\text{variasi}}$$
	​

Cohen’s d mencoba melihat:

> “Perbedaannya besar dibanding noise data atau tidak?”

atau

> “Selisih rata-rata dibandingkan dengan seberapa ‘nyebar’ datanya”

Interpretasi Cohen’s d  

| Cohen’s d | Interpretasi |
|------ |--------- |
| 0.2       | efek kecil   |
| 0.5       | efek sedang  |
| 0.8       | efek besar   |

Misalnya:

| Group | Mean Revenue |
|-- |--------- |
| A     | 100          |
| B     | 120          |

Dengan standar deviasi $(σ)$ adalah 40, maka:  

$$d=\frac{120−100}{40}	​=0.5$$

Artinya:  
> efek perubahan cukup sedang/moderat.

##### **Rank-Biserial Correlation (Mann–Whitney)**

Definisi

> Effect size untuk uji non-parametrik (Mann–Whitney)

Rumus

$$
r = \frac{U}{n_1 \times n_2}
$$

Keterangan Variabel

* $U$ : Statistik Mann–Whitney
  → Jumlah “kemenangan” satu grup terhadap grup lain

* $n_1$ : Jumlah data grup pertama

* $n_2$ : Jumlah data grup kedua

Intuisi Formula

$$
r = \frac{\text{berapa kali A menang}}{\text{total perbandingan}}
$$

Rank-Biserial mencoba melihat:

> “Seberapa sering nilai di grup A lebih tinggi dibanding grup B?”

atau

> “Dominasi satu grup terhadap grup lain”

Interpretasi

| r   | Interpretasi |
| --- | ------------ |
| 0.1 | kecil        |
| 0.3 | sedang       |
| 0.5 | besar        |

Contoh

Misal:

* #U = 70$
* #n_1 = 10$, $n_2 = 10$

$$
r = \frac{70}{100} = 0.7
$$

Artinya:

> Grup A cukup dominan dibanding grup B

##### **Eta Squared (ANOVA)**

> Effect size untuk ANOVA (parametrik)

Rumus

$$
\eta^2 = \frac{SS_{between}}{SS_{total}}
$$

Keterangan Variabel

* ( SS_{between} ) : Variasi antar grup
  → Seberapa beda rata-rata antar grup

* ( SS_{total} ) : Total variasi data
  → Variasi keseluruhan

Intuisi Formula

$$
\eta^2 = \frac{\text{variasi karena grup}}{\text{total variasi}}
$$

Eta squared mencoba melihat:

> “Berapa persen variasi data dijelaskan oleh perbedaan grup?”

Interpretasi

| η²   | Interpretasi |
| ---- | ------------ |
| 0.01 | kecil        |
| 0.06 | sedang       |
| 0.14 | besar        |

Contoh

Misal:

* ( SS_{between} = 200 )
* ( SS_{total} = 1000 )

$$
\eta^2 = \frac{200}{1000} = 0.2
$$

Artinya:

> 20% variasi data dijelaskan oleh perbedaan grup (efek besar)

##### **Epsilon Squared (Kruskal–Wallis)**

Definisi

> Effect size untuk ANOVA non-parametrik

Rumus

$$
\epsilon^2 = \frac{H - k + 1}{N - k}
$$

Keterangan Variabel

* $H$ : Statistik Kruskal–Wallis
* $k$ : Jumlah grup
* $N$ : Total jumlah data

Intuisi Formula

$$
\epsilon^2 = \frac{\text{perbedaan ranking antar grup}}{\text{total data}}
$$

Epsilon squared mencoba melihat:

> “Seberapa besar perbedaan antar grup jika dilihat dari ranking data”

Interpretasi

| ε²   | Interpretasi |
| ---- | ------------ |
| 0.01 | kecil        |
| 0.08 | sedang       |
| 0.26 | besar        |

Contoh

Misal:

* ( H = 10 )
* ( k = 3 )
* ( N = 30 )

$$
\epsilon^2 = \frac{10 - 3 + 1}{30 - 3} = \frac{8}{27} \approx 0.296
$$

Artinya:

> Efek besar (perbedaan antar grup cukup kuat)

##### **Cramer’s V (Chi-Square)**


Definisi

> Effect size untuk data kategori (chi-square)

Rumus

$$
V = \sqrt{\frac{\chi^2}{n(k-1)}}
$$

Keterangan Variabel

* ( \chi^2 ) : Nilai chi-square

* ( n ) : Total jumlah data

* ( k ) : Jumlah kategori terkecil (baris/kolom)

Intuisi Formula

$$
V = \sqrt{\frac{\text{hubungan kategori}}{\text{skala data}}}
$$

Cramer’s V mencoba melihat:

> “Seberapa kuat hubungan antar kategori?”

Interpretasi

| V   | Interpretasi |
| --- | ------------ |
| 0.1 | kecil        |
| 0.3 | sedang       |
| 0.5 | besar        |

Contoh

Misal:

* ( \chi^2 = 20 )
* ( n = 100 )
* ( k = 2 )

$$
V = \sqrt{\frac{20}{100(2-1)}} = \sqrt{0.2} \approx 0.447
$$

Artinya:

> Hubungan antar kategori cukup kuat

## **Post Hoc Test**

**Post-Hoc Test**

> Digunakan setelah uji utama (ANOVA / Kruskal) menunjukkan **ada perbedaan**,  
> 
> untuk mencari:  
> “Perbedaan itu terjadi di grup mana saja?”

### **Tukey HSD (Setelah ANOVA)**



> Post-hoc paling umum untuk ANOVA (parametrik)


Rumus

$$
q = \frac{\bar{X}_i - \bar{X}_j}{SE}
$$


Keterangan Variabel

* $\bar{X}_i$, $\bar{X}_j$ : rata-rata dua grup yang dibandingkan
* $SE$ : standard error (error gabungan dari semua grup)


Intuisi Formula

$$
q = \frac{\text{selisih rata-rata}}{\text{error}}
$$

Tukey mencoba melihat:

> “Apakah selisih antar dua grup cukup besar dibanding noise?”


Kapan dipakai

* Setelah **ANOVA signifikan**
* Data **normal & homogen**


Contoh

Misal ada 3 grup:

* A = 100
* B = 120
* C = 105

Hasil:

* A vs B → beda signifikan
* A vs C → tidak
* B vs C → beda signifikan

Insight:

> Grup B yang paling berbeda

### **Bonferroni Correction**



> Versi lebih “aman” (ketat) untuk banyak perbandingan

Rumus

$$
p_{adjusted} = p \times k
$$

Keterangan Variabel

* ( p ) : p-value asli
* ( k ) : jumlah perbandingan

Intuisi Formula

$$
\text{p makin besar kalau banyak tes}
$$

Bonferroni mencoba:

> “Menghindari false positive karena terlalu banyak perbandingan”

Kapan dipakai

* Banyak pairwise comparison
* Mau hasil lebih konservatif

Contoh

* p awal = 0.01
* jumlah perbandingan = 5

$$
p = 0.01 \times 5 = 0.05
$$

👉 Bisa jadi tadinya signifikan → jadi tidak

### **Dunn Test (Setelah Kruskal–Wallis)**



> Post-hoc untuk non-parametrik (ranking)

Rumus

$$
Z = \frac{R_i - R_j}{SE}
$$

Keterangan Variabel

* $R_i$, $R_j$ : rata-rata ranking grup
* $SE$ : standard error ranking

Intuisi Formula

$$
Z = \frac{\text{selisih ranking}}{\text{error}}
$$

Dunn test mencoba:

> “Grup mana yang ranking-nya beda jauh?”

Kapan dipakai

* Setelah **Kruskal-Wallis signifikan**
* Data **tidak normal**

Contoh

Misal ranking:

* A = 10
* B = 20
* C = 12

Hasil:

* A vs B → beda
* A vs C → tidak
* B vs C → beda

👉 Lagi-lagi:

> Grup B paling beda

## **Statistic Power** 

Power mengukur:

> “Seberapa besar kemungkinan eksperimen berhasil mendeteksi perbedaan nyata.”

Kenapa efek kecil sulit dideteksi? karena di data nyata selalu ada:

- noise,
- variasi user,
- randomness.

Kalau perubahannya kecil banget:

| Group | CR    |
|-- |-- |
| A     | 5.00% |
| B     | 5.05% |

Selisihnya cuma: 

> 0.05%

Masalahnya noise data bisa lebih besar dari selisih itu. Misalnya:
- hari ini conversion naik,
- besok turun,
- user behavior berubah,
- traffic berbeda.

Akhirnya:
> sinyal efek kecil “tertutup” oleh noise.

Maka untuk efek kecil biasanya perlu:

- sample lebih besar
- durasi eksperimen lebih lama
- data lebih stabil

Supaya:

- noise relatif mengecil,
- sinyal efek mulai terlihat.

Umumnya target power adalah 80% karena dianggap kompromi yang bagus. Ini artinya adalah:

> Kalau efek nyata memang ada 80% kemungkinan eksperimen berhasil menemukannya.

**Hubungan dengan Type II Error**

Type II Error (β) terjadi ketika:

> efek sebenarnya ada, tapi test bilang tidak ada.

Rumus

$$\text{Power}=1−β$$

Contoh

- Kalau:

    $$β=0.2$$

- Maka:

    $$Power=0.8=80\%$$

Faktor yang Mempengaruhi Power
- Sample Size

    Semakin besar sample:

    > power naik.

    Karena:

    - noise lebih kecil,
    - hasil lebih stabil.  
    
- Effect Size

    Semakin besar efek:
    > semakin mudah dideteksi.
    
    Misal:

    > conversion naik dari 5% → 20% $\to$ mudah terlihat.

    Tapi kalau:

    > 5% → 5.2% $\to$ lebih sulit.

- Noise / Variasi Data

    > Semakin noisy data $\to$ semakin sulit mendeteksi efek.

- Significance Level (α)

    Biasanya:

    0.05

    Kalau threshold terlalu ketat:

    > lebih sulit mendapatkan signifikan, power turun.

**Contoh Nyata**

Misalnya:

|Group	|CR|
|-------|--|
|A	    |5%|
|B	    |6%|

Kelihatannya naik, tapi:

> user cuma 50.

Akhirnya:

- p-value besar,
- tidak signifikan.

Bukan berarti:

> tombol baru jelek.

Bisa jadi:

> data belum cukup.

## **Addition - Z-test vs T-test**

### **Introduction**

Kapan Pakai Z-Test?

Gunakan Z-Test ketika:

* data berbentuk proporsi/persentase,
* hasilnya binary:

  * beli/tidak,
  * klik/tidak,
  * fraud/tidak.

Biasanya dipakai untuk:

* conversion rate,
* CTR,
* open rate,
* success rate.

Contoh:

| Group | Purchase | Total User |
| ----- | -------- | ---------- |
| A     | 50       | 1000       |
| B     | 70       | 1000       |

Pertanyaan:

> apakah proporsi purchase berbeda?

Kapan Pakai T-Test?

Gunakan T-Test ketika:

* membandingkan rata-rata numerik,
* datanya berupa angka kontinu.

Biasanya dipakai untuk:

* average revenue,
* average session duration,
* average salary,
* average transaction.

Contoh:

| Group | Revenue |
| ----- | ------- |
| A     | 120     |
| A     | 150     |
| B     | 180     |
| B     | 200     |

Pertanyaan:

> apakah rata-rata revenue berbeda?

Intuisi Singkat

| Test   | Fokus          |
| ------ | -------------- |
| Z-Test | beda proporsi  |
| T-Test | beda rata-rata |





Cara Cepat Membedakan

- Kalau datanya:

    * 0/1,
    * yes/no,
    * success/fail,

    → biasanya Z-Test.

- Kalau datanya:

    * angka transaksi,
    * waktu,
    * nilai numerik,

    → biasanya T-Test.


### **Example**



Data Eksperimen A/B Testing

| User | Group | Purchase | Revenue |
| ---- | ----- | -------- | ------- |
| U1   | A     | 1        | 120     |
| U2   | A     | 0        | 0       |
| U3   | A     | 1        | 150     |
| U4   | A     | 0        | 0       |
| U5   | A     | 1        | 130     |
| U6   | B     | 1        | 180     |
| U7   | B     | 1        | 200     |
| U8   | B     | 0        | 0       |
| U9   | B     | 1        | 170     |
| U10  | B     | 1        | 210     |

Pertanyaan 1 — Z-Test

- Pertanyaan Bisnis

    > “Apakah conversion rate Group B lebih tinggi dari Group A?”

- Kenapa Pakai Z-Test?

    Karena yang dibandingkan:

    $$
    Purchase = 0 \text{ atau } 1
    $$

    Artinya data berbentuk:

    * binary,
    * proporsi.

- Hitung Conversion Rate

    - Group A
    
        Purchase:
    
        $$
        3
        $$
    
        Total user:
    
        $$
        5
        $$
    
        $$
        p_A=\frac{3}{5}=0.6
        $$
    
    - Group B
    
        Purchase:
    
        $$
        4
        $$
    
        Total user:
    
        $$
        5
        $$
    
        $$
        p_B=\frac{4}{5}=0.8
        $$
    
    Pertanyaan Statistik
    
    > apakah perbedaan 60% vs 80% benar-benar signifikan?

Pertanyaan 2 — T-Test

- Pertanyaan Bisnis

    > “Apakah rata-rata revenue Group B lebih tinggi dari Group A?”

- Kenapa Pakai T-Test?

    Karena yang dibandingkan:

    $$
    Revenue
    $$

    berupa angka kontinu/numerik.

- Hitung Mean Revenue

    Group A

    $$
    \frac{120+0+150+0+130}{5}
    =80
    $$


    Group B

    $$
    \frac{180+200+0+170+210}{5}
    =152
    $$

- Pertanyaan Statistik

    > apakah perbedaan mean revenue 80 vs 152 benar-benar signifikan?

## **Summary**

**FLOW ANALISIS STATISTIK (DARI BISNIS → INSIGHT)**

- Pertanyaan Bisnis

    > Mulai dari sini, bukan dari statistik

    Contoh:

    * “Apakah campaign B lebih baik dari A?”
    * “Apakah ada perbedaan revenue antar channel?”
    * “Apakah diskon berpengaruh ke conversion?”

    Output:   
    > **tujuan jelas (compare / hubungan / pengaruh)**
    
- Lihat KPI (Data Asli)  
    Misal:  
    - revenue
    - conversion
    - CTR
    - dll
    
    bisa pakai:
    - mean / median
    - summary per grup

    Tujuan:

    > “Ngerti dulu gambaran data, jangan langsung uji statistik”


- Confidence Interval (CI)  
    lihat range estimasi mean / difference

    Tujuan:

    > “Seberapa pasti angka KPI ini?”

    Insight penting:

    - Kalau CI overlap banyak → kemungkinan gak beda
    - Tapi tetap perlu uji statistik

- Cek Asumsi / Distribusi
    - normal atau tidak
    - variansi sama atau tidak

    Tujuan:

    “Pilih test yang benar”

- **Ubah ke Hipotesis**

    Contoh:

    * H0 (null): tidak ada perbedaan
    * H1 (alternatif): ada perbedaan

    Contoh real:

    * H0: mean revenue A = B
    * H1: mean revenue A ≠ B

    Output:   
    > **statement yang bisa diuji**

- **Pilih Uji Statistik**

    Berdasarkan:

    * jumlah grup
    * jenis data
    * normal / tidak

    - Perbedaan rata-rata

        | Kondisi              | Test           |
        | -------------------- | -------------- |
        | 2 grup normal        | t-test         |
        | 2 grup tidak normal  | Mann–Whitney   |
        | >2 grup normal       | ANOVA          |
        | >2 grup tidak normal | Kruskal–Wallis |

    - Kategori

        | Test       |
        | ---------- |
        | Chi-Square |


    Output:   
    > **p-value**

- **Keputusan (Signifikan atau Tidak)**

    * p < 0.05 → ada perbedaan
    * p ≥ 0.05 → tidak ada bukti perbedaan

    Output:
  
  > 
    > “Secara statistik beda / tidak”

    ⚠️ Tapi ini belum cukup!

- **Effect Size (Seberapa Besar Bedanya?)**

    > Ini yang sering dilupakan tapi penting di bisnis

    Mapping cepat

    | Test         | Effect Size     |
    | ------------ | --------------- |
    | t-test       | Cohen’s d       |
    | Mann–Whitney | Rank-biserial   |
    | ANOVA        | Eta squared     |
    | Kruskal      | Epsilon squared |
    | Chi-square   | Cramer’s V      |


    Output:
  
  > 
    > “Bedanya kecil / sedang / besar”



- **Post-Hoc Test (Kalau >2 Grup)**

    > Hanya dilakukan kalau hasil signifikan


    | Test utama | Post-hoc |
    | ---------- | -------- |
    | ANOVA      | Tukey    |
    | Kruskal    | Dunn     |


    Output:

    > “Grup mana yang beda”

- **Insight Bisnis (INI YANG PALING PENTING)**

    Gabungkan semua:

    Contoh:

    > “Campaign B signifikan lebih tinggi dari A (p<0.05), dengan efek sedang (d=0.5), dan estimasi peningkatan revenue antara 5–35%. Campaign B bisa diprioritaskan.”



**Ringkasan super singkat (biar nempel)**

- Business Question
- Hipotesis
- Uji statistik → p-value
- Effect size → besar kecil
- CI → range
- Post-hoc → grup mana
- Insight bisnis

Cara mikir yang bener

* p-value → “ada beda?”
* effect size → “penting gak?”
* CI → “seberapa pasti?”
* post-hoc → “beda di mana?”
