# **Fraud Detection**

## **Introduction**

**Latar Belakang**

Dalam sistem digital (e-commerce, fintech, dll):

* Ribuan transaksi terjadi setiap detik
* Tidak semua transaksi dilakukan oleh user “jujur”

Contoh fraud:

| Industri   | Contoh Fraud                 |
|------- |------------------------- |
| Bank       | transaksi kartu kredit palsu |
| E-commerce | fake order                   |
| Insurance  | klaim palsu                  |
| Fintech    | account abuse                |
| Game       | cheating/top-up ilegal       |

**Masalah utamanya:**

> Fraud terlihat seperti transaksi normal… sampai terlambat

Dampaknya:

* Kerugian finansial langsung
* Kehilangan customer trust
* Reputasi rusak
* Sistem jadi tidak terpercaya

**Tujuan**

- Tujuan Utama:

    > Mendeteksi dan mencegah aktivitas fraud secepat mungkin

- Tujuan Spesifik:

    - Mengidentifikasi transaksi mencurigakan
    - Mengurangi kerugian akibat fraud
    - Menjaga keseimbangan: aman tapi tidak mengganggu user normal




**Masalah yang Ingin Dijawab**

- Contoh 1

    > “Apakah transaksi ini mencurigakan?”

- Contoh 2

    > “User mana kemungkinan melakukan fraud?”

- Contoh 3

    > “Apakah ada pola transaksi abnormal?”

**Kenapa Perlu Metrik?**

Fraud itu tricky karena:

* Tidak selalu jelas
* Bisa menyerupai user normal

Maka:

> Kita butuh metrik untuk menangkap “pola yang tidak wajar”

Jenis Fraud Detection

| Jenis                      | Fokus                    |
|----------------------- |--------------------- |

## **Metrik**

### **Introduction**

Metrik


- Fraud Rate    
    > Persentase transaksi yang teridentifikasi sebagai fraud

    Konsep:

    $$Fraud\ Rate = \frac{\text{Fraudulent Transactions}}{\text{Total Transactions}} \times 100\% $$


    **Intuisi:**

    > Seberapa besar “ancaman” di sistem
    * Tinggi → sistem rawan
    * Rendah → relatif aman (atau deteksi belum bagus 👀)

    **Kontribusi ke tujuan:**

    ✔ Mengukur tingkat fraud secara umum

    $$ \ $$

- False Positive Rate (FPR)    
    > Persentase transaksi normal yang salah dianggap fraud

    Konsep:

    $$FPR = \frac{\text{Normal Transactions Flagged as Fraud}}{\text{Total Normal Transactions}}$$


    **Intuisi:**

    > “Seberapa sering kita salah tuduh”

    Ini sangat penting!

    Kalau FPR tinggi:

    * user normal terganggu
    * bisa kehilangan pelanggan

        Trade-off:

        * terlalu ketat → banyak false alarm
        * terlalu longgar → fraud lolos

    **Kontribusi ke tujuan:**

    ✔ Menjaga user experience

    $$ \ $$

- Transaction Frequency    
    > Jumlah transaksi dalam periode waktu tertentu

    **Intuisi:**

    > Manusia punya batas, bot tidak

    Contoh:

    * 50 transaksi dalam 1 menit → sangat mencurigakan

    Ini sinyal kuat fraud

    **Kontribusi ke tujuan:**

    ✔ Deteksi pola tidak wajar

    $$ \ $$

- Anomaly Detection Score    
    > Skor yang menunjukkan seberapa “aneh” suatu transaksi dibanding pola normal

    **Intuisi:**

    > Semakin berbeda dari kebiasaan → semakin mencurigakan

    Contoh:

    * User biasa belanja 100k
    * Tiba-tiba transaksi 10 juta

    skor anomali tinggi

    **Kontribusi ke tujuan:**

    ✔ Deteksi fraud yang tidak obvious


### **Addition**

Hubungan Antar Metrik



- Fraud Detection = Balance Problem

  Kita ingin:

  * Fraud tertangkap ✔
  * User normal tidak terganggu ✔


  - Konflik utama:

  * Naikkan sensitivity → fraud tertangkap
    ❌ tapi FPR naik

  * Turunkan sensitivity → user aman
    ❌ tapi fraud lolos



  Ini inti fraud system:

    > **trade-off antara security vs user experience**



- Pola Umum Fraud

  Biasanya ada kombinasi:
  
  * Transaction frequency tinggi
  * Anomaly score tinggi
  * Behavior berubah tiba-tiba
  
  Fraud jarang terlihat dari 1 metrik saja




**Dari Metrik → Insight Nyata**

- Contoh kamu:

    User:

    > 50 transaksi dalam 1 menit



- Insight:

    > Sangat tidak normal untuk manusia

    Kemungkinan:

    * bot
    * scripted attack
    * abuse sistem

**Dari Insight → Keputusan**

- Action yang bisa diambil:

  - **Blokir sementara akun**
  - Minta **verifikasi tambahan (OTP / CAPTCHA)**
  - Tandai sebagai **high-risk user**
  - Monitor aktivitas lanjutan


- Contoh lain:

    * Banyak transaksi dengan anomaly score tinggi
    * Tapi FPR juga tinggi

    Insight:

    * Model terlalu agresif

    Keputusan:

    * tuning threshold
    * improve model
    * tambahkan fitur (device, lokasi, dll)


- Intuisi:

    > Fraud detection itu bukan mencari yang “jelas salah”, tapi menemukan yang “tidak wajar”


**Alur Berpikir**

- **Masalah:**  
   Fraud tersembunyi dalam transaksi normal

- **Tujuan:**  
   Deteksi dan cegah fraud

- **Solusi:**  
   Gunakan metrik perilaku & anomali

- **Insight:**  
   Temukan pola tidak normal

   Blokir, verifikasi, atau investigasi

**Insight Lanjutan**

Kalau kamu masuk ke ML, ini jadi sangat menarik:

- Pendekatan:

   - **Rule-based**  

      * contoh: > 10 transaksi/menit → flag

   - **Machine Learning**  

      * classification (fraud vs normal)
      * anomaly detection (unsupervised)

   - **Deep Learning**  

      * sequence behavior (RNN/LSTM)


- Feature penting:

   * waktu transaksi
   * lokasi
   * device
   * pola pembelian


**Ringkasan**

| Metrik        | Menjawab                      |
| ------------- | ----------------------------- |
| Fraud Rate    | Seberapa besar masalah        |
| FPR           | Seberapa sering salah deteksi |
| Frequency     | Seberapa abnormal aktivitas   |

Output akhirnya:

> **Menemukan aktivitas mencurigakan tanpa merusak pengalaman user normal**


## **Rule Based Fraud Detection**

Definisi

- Rule-Based Fraud Detection adalah metode deteksi kecurangan yang bekerja menggunakan:

    > sekumpulan aturan bisnis (business rules) yang telah ditentukan sebelumnya.

- Apabila suatu aktivitas memenuhi kondisi tertentu, maka sistem akan memberikan tanda (flag) sebagai indikasi fraud atau aktivitas mencurigakan.

- Metode ini merupakan pendekatan paling dasar dan paling banyak digunakan pada tahap awal pembangunan sistem fraud detection.

Cara Kerja

Secara umum, alurnya adalah:


> ```text
> Data transaksi masuk
>         ↓
> Dicek terhadap aturan fraud
>         ↓
> Apakah memenuhi kondisi mencurigakan?
>         ↓
> Ya → Tandai fraud / suspicious
> Tidak → Transaksi normal
> ```

Contoh Aturan Fraud
- Nominal Transaksi Sangat Besar  
    > Jika transaksi > Rp50.000.000
    → tandai sebagai suspicious  
    
    Tujuan  
    > Mendeteksi transaksi dengan nilai tidak wajar dibanding perilaku normal customer.

- Perubahan Lokasi Drastis
    > Jika login sebelumnya di Jakarta  
    > dan 10 menit kemudian transaksi di luar negeri
    → tandai fraud  
    
    Tujuan  
    > Mendeteksi kemungkinan akun dibajak (account takeover).

- Terlalu Banyak Transaksi dalam Waktu Singkat  
    > Jika jumlah transaksi > 20 dalam 1 menit
    → tandai fraud  
    
    Tujuan  
    > Mendeteksi aktivitas bot atau pencurian kartu.

- Login dari Banyak Device    
    > Jika 1 akun login dari >5 device berbeda dalam 24 jam
    → tandai suspicious  
    
    Tujuan  
    > Mendeteksi akun yang dibagikan atau dicuri.

- Aktivitas pada Jam Tidak Normal
    > Jika transaksi dilakukan pukul 02:00–04:00  
    > dan customer biasanya aktif siang hari
    → tingkatkan risk score  
    
    Tujuan  
    > Mendeteksi perilaku yang tidak sesuai pola normal pengguna.

Komponen Utama dalam Rule-Based System
- Rules Engine

    Bagian sistem yang bertugas menjalankan aturan fraud.

    Contoh:

    > IF transaksi > threshold  
    > THEN fraud_flag = TRUE

- Threshold

    Batas nilai tertentu yang digunakan sebagai indikator fraud.

    Contoh:

    - transaksi > Rp10 juta
    - gagal login > 5 kali
    - transaksi > 30 kali/jam

    Threshold biasanya ditentukan berdasarkan:

    - pengalaman bisnis,
    - historical data,
    - regulasi,
    - analisis statistik.
     
- Alert / Flagging System

    Jika aturan terpenuhi, sistem akan:

    - memberi notifikasi,
    - memblokir transaksi,
    - meminta verifikasi tambahan,
    - atau mengirim ke tim investigasi.

Jenis Rule dalam Fraud Detection
- Static Rules

    Aturan tetap dan jarang berubah.

    Contoh:

    > Jika kartu blacklist digunakan
    > → tolak transaksi

- Dynamic Rules

    Aturan yang menyesuaikan perilaku customer.

    Contoh:

    > Jika transaksi > 5x rata-rata transaksi user
    > → suspicious

    Aturan ini lebih fleksibel dibanding static rule.

Sumber Data yang Digunakan

Rule-based system biasanya memanfaatkan berbagai data, seperti:

| Data           | Contoh                   |
| -------------- | ------------------------ |
| Data transaksi | nominal, waktu, merchant |
| Data lokasi    | GPS, IP address          |
| Data device    | device ID, browser       |
| Data perilaku  | frekuensi transaksi      |
| Data akun      | umur akun, riwayat login |


Contoh di Industri
- Perbankan
    - transaksi luar negeri mendadak,
    - penarikan ATM berulang,
    - transfer nominal besar.
- E-Commerce
    - banyak checkout dalam waktu singkat,
    - penggunaan voucher berulang,
    - akun baru langsung transaksi besar.
- Fintech
    - pinjaman dari banyak akun/device,
    - identitas ganda,
    - aktivitas mencurigakan pada KYC.

Kelebihan Rule-Based Fraud Detection
- Mudah Dipahami
- Cepat Diimplementasikan
- Cocok Sebagai Baseline System
- Mudah Dikontrol  
  Perusahaan dapat langsung:

    - menambah rule,
    - menghapus rule,
    - mengubah threshold.

Kekurangan Rule-Based Fraud Detection
- Sulit Mendeteksi Fraud Baru  
  Sistem hanya bisa mendeteksi pola yang:

    > sudah diketahui sebelumnya.

    Jika fraudster menggunakan pola baru, aturan lama mungkin gagal mendeteksi.
- Terlalu Kaku
  Aturan tidak fleksibel terhadap perubahan perilaku user.  
  Contoh:
  - user sedang traveling
  - user tiba-tiba transaksi besar karena kebutuhan tertentu.
  
  Sistem bisa salah menganggap fraud.
- Tinggi False Positive  
  False positive adalah:

    > transaksi normal tetapi dianggap fraud.

    Akibatnya:

    - customer terganggu,
    - transaksi sah diblokir,
    - pengalaman pengguna menurun.
- Maintenance Semakin Sulit  
  Semakin besar sistem:  

    - rule semakin banyak,
    - konflik antar-rule muncul,
    - debugging menjadi sulit.

## **Statistical Fraud Detection**

### **Introduction**

Definisi

- Statistical Fraud Detection adalah metode deteksi fraud yang menggunakan:

    > pendekatan statistik untuk mengidentifikasi pola transaksi yang tidak normal (abnormal pattern).

- Metode ini bekerja dengan menganalisis distribusi data historis, kemudian mencari transaksi atau aktivitas yang memiliki karakteristik sangat berbeda dibanding perilaku normal.

- Pendekatan statistik merupakan salah satu fondasi penting dalam fraud detection sebelum berkembangnya machine learning modern.

Konsep Dasar

- Intuisi utama dari statistical fraud detection adalah:

    > data yang sangat jauh dari pola normal memiliki kemungkinan lebih besar sebagai fraud.

- Dengan kata lain:

    > transaksi normal biasanya berkumpul pada pola tertentu,
    > sedangkan fraud sering muncul sebagai anomali (anomaly) atau outlier.

Contoh Sederhana

- Misalkan rata-rata transaksi customer adalah:

    > Rp500.000

    Namun tiba-tiba muncul transaksi:

    > Rp50.000.000

    Secara statistik:
    > nilai tersebut sangat jauh dari pola normal,

- sehingga dapat dianggap sebagai:
    - outlier,
    - aktivitas abnormal,
    - atau potensi fraud.

Cara Kerja Statistical Fraud Detection

Secara umum:

> ```text
> Kumpulkan data historis
>         ↓
> Hitung statistik penting
> (mean, std, quartile, dll)
>         ↓
> Bandingkan transaksi baru
> dengan distribusi normal
>         ↓
> Jika terlalu abnormal
> → tandai suspicious
> ```

Statistik yang Umum Digunakan

| Statistik             | Fungsi                         |
| --------------------- | ------------------------------ |
| Mean                  | Mengukur rata-rata             |
| Median                | Nilai tengah data              |
| Standard Deviation    | Mengukur penyebaran data       |
| Variance              | Mengukur variasi data          |
| Z-Score               | Mengukur jarak dari rata-rata  |
| IQR                   | Mendeteksi outlier             |
| Percentile            | Menentukan posisi data         |
| Distribution Analysis | Menganalisis bentuk distribusi |


Kelebihan Statistical Fraud Detection
- Mudah Diimplementasikan

    Tidak memerlukan model machine learning kompleks.

- Cepat Diproses

    Perhitungan statistik relatif ringan secara komputasi.

- Cocok untuk Baseline Detection

    Sering digunakan sebagai:

    - sistem awal,
    - anomaly screening,
    - atau preprocessing sebelum ML.
- Interpretasi Mudah

    Tim bisnis dapat memahami alasan transaksi dianggap abnormal.

Kekurangan Statistical Fraud Detection
- Sulit Menangani Pola Fraud Kompleks

    Fraud modern sering:

    - tersembunyi,
    - adaptif,
    - dan tidak selalu muncul sebagai outlier sederhana.
     
- Sensitif terhadap Distribusi Data

    Jika distribusi data tidak normal:  
    > mean dan standard deviation bisa menyesatkan.

- Banyak False Positive

    Transaksi besar belum tentu fraud.

    Contoh:

    - customer premium,
    - promo besar,
    - musim liburan.
     
- Tidak Belajar Otomatis

    Metode statistik tradisional:

    > tidak melakukan pembelajaran otomatis seperti machine learning.

Penggunaan di Dunia Nyata

- Perbankan
    - deteksi transfer abnormal,
    - transaksi kartu kredit tidak biasa,
    - penarikan ATM ekstrem.
- E-Commerce
    - lonjakan pembelian,
    - aktivitas bot,
    - transaksi massal mendadak.
- Cybersecurity
    - login abnormal,
    - traffic tidak normal,
    - akses sistem yang menyimpang.

Perbandingan dengan Rule-Based Detection

| Aspek                | Rule-Based    | Statistical     |
| -------------------- | ------------- | --------------- |
| Dasar deteksi        | aturan manual | distribusi data |
| Fleksibilitas        | rendah        | lebih fleksibel |
| Perlu data historis  | sedikit       | ya              |
| Bisa deteksi outlier | terbatas      | sangat baik     |
| Mudah dijelaskan     | sangat mudah  | mudah           |


### **Z-Score Method**

Z-Score digunakan untuk mengukur:

> seberapa jauh suatu data dari rata-rata dalam satuan standar deviasi.

Rumus:

$$Z = \frac{x - \mu}{\sigma}$$

Semakin besar nilai:

> |Z|

maka:

- data semakin jauh dari normal,
- kemungkinan anomali semakin tinggi.

interpretasi:

| Z-Score       | Interpretasi       |
| ------------- | ------------------ |
| 0             | Tepat di rata-rata |
| ±1            | Masih normal       |
| ±2            | Mulai tidak biasa  |
| ±3 atau lebih | Sangat abnormal    |

Dalam banyak sistem fraud detection:

> |Z| > 3

sering dianggap sebagai indikator anomaly.

### **IQR Method**

IQR digunakan untuk:

> mendeteksi outlier berdasarkan penyebaran kuartil data.

Metode ini sangat populer karena:

- sederhana,
- robust terhadap extreme value,
- tidak terlalu sensitif terhadap distribusi miring (skewed distribution).

Rumus IQR adalah:

$$IQR=Q3−Q1$$


Suatu data dianggap outlier jika:

$$x>Q3+1.5(IQR)$$

atau

$$x<Q1−1.5(IQR)$$


## **Anomaly Detection**

### **Introduction**

Definisi

- Anomaly Detection adalah teknik untuk mendeteksi:

    > perilaku, pola, atau data yang berbeda secara signifikan dari kondisi normal.

- Dalam konteks fraud detection, anomaly detection digunakan untuk menemukan aktivitas yang:

    - tidak biasa,
    - menyimpang dari pola umum,
    - atau memiliki karakteristik langka (rare behavior).

- Pendekatan ini sangat penting karena banyak kasus fraud belum pernah terlihat sebelumnya.

Intuisi Dasar

- Sebagian besar aktivitas dalam sistem biasanya bersifat:

    > normal

    Sedangkan fraud sering muncul sebagai:

    > anomali / penyimpangan

- Karena itu, sistem mencoba mempelajari:

    - seperti apa perilaku normal,
    - lalu mencari data yang berbeda jauh dari normal tersebut.

Contoh Anomaly dalam Fraud Detection
- Login dari Negara Berbeda
    > User biasa login dari Indonesia, tetapi tiba-tiba login dari Rusia.

    Perilaku ini dianggap abnormal.

- Transaksi Terlalu Cepat
    > 10 transaksi dalam 5 detik.

    Aktivitas ini sulit dilakukan manusia normal.

- Pola Pembelian Tidak Wajar
    > Customer biasa belanja Rp100 ribu, tiba-tiba membeli Rp50 juta.  
- Aktivitas Device Tidak Normal
    > 1 akun digunakan di 20 device berbeda.

Kapan Anomaly Detection Digunakan?

Anomaly detection sangat berguna ketika:

- Label Fraud Sedikit

    Pada banyak kasus:

    > data normal sangat banyak, tetapi data fraud sangat sedikit.

    Kondisi ini disebut:

    > imbalanced data

- Fraud Sulit Diketahui

    Banyak fraud:

    - belum pernah terjadi,
    - terus berubah,
    - atau memiliki pola baru.

    Karena itu:

    > model supervised biasa sulit mengenali fraud baru.

- Ingin Mendeteksi Unknown Fraud

    Anomaly detection mampu mendeteksi:

    > fraud yang belum pernah dilihat sebelumnya

    karena fokusnya adalah:

    > mencari penyimpangan, bukan menghafal label fraud lama.

Cara Kerja Anomaly Detection
> ```text
> Pelajari pola normal
>         ↓
> Bangun batas normalitas
>         ↓
> Data baru masuk
>         ↓
> Jika terlalu berbeda
> → anomaly
> ```

Jenis Anomaly
- Point Anomaly  
    Satu data sangat berbeda dari lainnya.

    Contoh:

    > 1 transaksi Rp100 juta di antara transaksi Rp100 ribu

- Contextual Anomaly

    Data abnormal pada kondisi tertentu.

    Contoh:

    > Login jam 3 pagi untuk user yang biasanya aktif siang
- Collective Anomaly

    Sekumpulan aktivitas tampak mencurigakan.

    Contoh:

    > 100 transaksi kecil berturut-turut

Teknik yang Sering Digunakan

| Model                      | Fungsi                                 |
| -------------------------- | -------------------------------------- |
| Isolation Forest           | Mendeteksi anomaly dengan isolasi data |
| One-Class SVM              | Mempelajari boundary data normal       |
| Local Outlier Factor (LOF) | Mengukur kepadatan lokal data          |
| Autoencoder                | Deep learning untuk anomaly detection  |
| Clustering                 | Mencari data jauh dari cluster         |
| Statistical Methods        | Deteksi outlier statistik              |


### **Isolation Forest**

#### **Introduction**

Definisi Formal

Isolation Forest adalah algoritma anomaly detection yang bekerja dengan cara:

> mengisolasi observasi menggunakan pembagian (split) fitur secara acak.

Ide utamanya:

* data normal biasanya berada dalam kelompok padat,
* sedangkan anomaly berada jauh/sendirian,
* sehingga anomaly lebih cepat terisolasi.

Analogi

- Bayangin ada kerumunan orang.

    Sebagian besar:

    * tinggi mirip,
    * berat mirip,
    * perilaku mirip.

    Tapi ada satu orang:

    * sangat tinggi,
    * pakai baju mencolok,
    * berdiri sendirian.

    Kalau kita membagi orang secara random:

    * orang aneh ini biasanya cepat kepisah,
    * sedangkan orang normal masih bercampur dengan banyak orang lain.

- Nah, Isolation Forest mencari:

    > “siapa yang paling gampang dipisahkan dari kerumunan.”

Intuisi Utama
- Data Normal

    Biasanya:

    * berkumpul,
    * rapat,
    * mirip satu sama lain.

    Akibatnya:

    * sulit dipisahkan,
    * butuh banyak split.

    
- Data Anomaly

    Biasanya:

    * sendirian,
    * jauh dari kelompok,
    * berbeda ekstrem.

    Akibatnya:

    * cepat terisolasi,
    * hanya butuh sedikit split.

Konsep Inti: Isolation

> Isolation = memisahkan satu data hingga sendirian.

Semakin cepat data menjadi sendirian:  
→ semakin mencurigakan.


Cara Kerja Isolation Forest


- Step 1 — Pilih Fitur Secara Random

    Misal dataset punya:

    * umur,
    * transaksi,
    * pendapatan.

    Algoritma memilih salah satu fitur secara acak.


- Step 2 — Pilih Threshold Random

    Misal fitur transaksi dipilih.

    Jika:

    * minimum = 100
    * maksimum = 5000

    Maka threshold dipilih random di antara keduanya.


    Formula Matematikanya

    Threshold dipilih menggunakan distribusi uniform:

    $$
    p \sim Uniform(min(X), max(X))
    $$


    Artinya:

    > komputer mengambil angka acak di antara nilai terkecil dan terbesar.

    Misal:

    * min = 100
    * max = 5000

    Maka threshold bisa:

    * 200
    * 1500
    * 4200
    * dll.

    Kenapa Random?

    Karena filosofi Isolation Forest adalah:

    > anomaly akan tetap mudah dipisahkan meskipun pembagian dilakukan secara acak.


- Step 3 — Split Data

    Misal threshold:
    $$
    p = 1000
    $$

    Maka:

    $$
    X < 1000
    $$

    masuk kiri.

    Dan:

    $$
    X \ge 1000
    $$

    masuk kanan.


    Contoh

    | User | Transaksi |
    | ---- | --------- |
    | A    | 100       |
    | B    | 120       |
    | C    | 110       |
    | D    | 130       |
    | E    | 5000      |

    Jika threshold:
    $$
    1000
    $$

    Maka:

    Kiri:

    * 100
    * 120
    * 110
    * 130

    Kanan:

    * 5000

    User E langsung sendirian.

    Bahasa Manusianya

    > 5000 terlalu jauh dari kerumunan.

    Jadi bahkan split random pun:

    * langsung memisahkan dia.

Path Length


- Path length adalah:

    > jumlah split yang diperlukan hingga suatu data terisolasi.

- Analogi:

    > Bayangin main tebak-tebakan.

    Kalau seseorang sangat unik:

    * cuma perlu 1–2 pertanyaan untuk menemukannya.

    Kalau orangnya mirip banyak orang:

    * perlu banyak pertanyaan.

- Intuisi

    - Anomaly

        Cepat ketemu:
        $$
        \text{path\ length} \rightarrow \text{pendek}
        $$

    - Normal

        Susah dipisahkan:
        $$
        \text{path\ length} \rightarrow \text{panjang}
        $$

- Formula Path Length

    Misal:

    $h(x)$ adalah panjang path data (x).

    Isolation Forest menghitung:

    $$
    E(h(x))
    $$

    yaitu:

    * rata-rata panjang path dari banyak tree.


Kenapa Banyak Tree?

Karena satu tree bisa:

* kebetulan jelek,
* split random tidak bagus.

Dengan banyak tree:

* hasil lebih stabil,
* seperti voting.

Makanya disebut:

- Forest

  karena terdiri dari banyak isolation tree.


- Anomaly Score

  Isolation Forest mengubah panjang path menjadi anomaly score.


  Formula

  $$
  s(x,n)=2^{-\frac{E(h(x))}{c(n)}}
  $$

  Keterangan:
  
  | Simbol    | Arti                  |
  | --------- | --------------------- |
  | (s(x,n))  | anomaly score         |
  | (E(h(x))) | rata-rata path length |
  | (c(n))    | normalisasi           |
  | (n)       | jumlah data           |

  Skor ini mengukur:

> seberapa cepat data bisa diisolasi dibanding data normal.



Interpretasi

- Jika:

    $$
    s(x) \approx 1
    $$

    Artinya:

    * sangat mudah dipisahkan,
    * kemungkinan anomaly tinggi.


- Jika:

    $$
    s(x) \approx 0.5
    $$

    Artinya:

    * normal.


- Jika:

    $$
    s(x) < 0.5
    $$
    
    Artinya:
    
    * sangat normal.


Kenapa Path Pendek = Anomaly?

Karena anomaly biasanya:

* jauh,
* sendirian,
* berada di pinggir distribusi.

Maka:

* split random sering langsung mengenainya.

Ringkasan Isolation Forest
- Ambil data.
- Pilih fitur random.
- Pilih angka split random.
- Split data menjadi kiri dan kanan.
- Data yang masih bercampur:
- di-split lagi.
- Data yang sudah sendirian:
    > berhenti (tidak diproses lagi di cabang itu).
- Hitung berapa banyak split sampai data sendirian:
    > path length
- Buat banyak tree dengan split random berbeda.
- Rata-rata semua path length.

#### **Contoh**

Langkah Algoritma
- Ambil data  
  
    | User | Transaksi |
    | ---- | --------- |
    | A    | 100       |
    | B    | 120       |
    | C    | 140       |
    | D    | 5000      |

- Pilih split random.

    Misal:  
    > 1000

- Split data.

    - Kiri (<1000)
        - A
        - B
        - C
    - Kanan (≥1000)  
        D

- D langsung sendirian.

    Maka:

    > D berhenti, tidak di-split lagi.

    Path D:

    > 1

- Grup kiri di-split lagi.

    Misal threshold:

    > 130

    Kiri (<130)
    - A
    - B
    Kanan (≥130)
    - C

    C sekarang sendirian.

    Path C:

    > 2

- Split lagi untuk A dan B.

    Misal threshold:

    > 110

    - Kiri (<110)
        > A
    - Kanan (≥110)
        > B

    Path:

    - A=3
    - B=3

- Hasil
  
    | User | Path Length |
    | ---- | ----------- |
    | D    | 1           |
    | C    | 2           |
    | A    | 3           |
    | B    | 3           |

    Tree dibuat banyak kali dengan split random berbeda.

    Semua path length:

    - dirata-ratakan,
    - lalu dijadikan anomaly score.


#### **Implementasi**

````python
import pandas as pd
from sklearn.ensemble import IsolationForest

# -----------------------------
# Dataset sederhana
# -----------------------------
data = pd.DataFrame({
    'transaksi': [100, 120, 140, 160, 180, 5000]
})

print(data)

# -----------------------------
# Membuat model Isolation Forest
# -----------------------------
iso = IsolationForest(
    n_estimators=100,   # jumlah tree
    contamination=0.1, # perkiraan proporsi anomaly
    random_state=42
)

# -----------------------------
# Training model
# -----------------------------
iso.fit(data)

# -----------------------------
# Prediksi
# -----------------------------
prediksi = iso.predict(data)

# Hasil:
#  1  -> normal
# -1  -> anomaly

data['prediksi'] = prediksi

print(data)
````

False Negative (FN)

- Fraud terjadi tetapi:

    ❌ account takeover

    ```text id="40m9z6"
    FN sangat berbahaya
    ```

- User normal dianggap fraud.
- Dampak FP

    ❌ transaksi valid diblokir  
    ❌ customer marah  
    ❌ pengalaman pengguna buruk  

### **Metrics**

Recall

- Recall mengukur:

    > seberapa banyak fraud berhasil ditangkap.

- Rumus Recall

    $$\text{Recall} = \frac{TP}{TP + FN}$$


- Intuisi Recall

    Semakin tinggi recall:

    > semakin sedikit fraud yang lolos.

    Karena itu recall sangat penting dalam fraud detection.

Precision

- Precision mengukur:

    > seberapa banyak prediksi fraud yang benar-benar fraud.


- Rumus Precision

    $$\text{Precision} = \frac{TP}{TP + FP}$$

- Intuisi Precision

    Semakin tinggi precision:

    > semakin sedikit false alarm.

Trade-Off Precision vs Recall

Biasanya:

| Fokus            | Dampak                        |
| ---------------- | ----------------------------- |
| Recall tinggi    | lebih banyak fraud tertangkap |
| Precision tinggi | lebih sedikit user terganggu  |

Meningkatkan recall sering menyebabkan:

> ```text id="z7ucy6"
> false positive naik
> ```

F1-Score

- F1-score adalah harmonic mean antara:

    * precision,
    * recall.


- Rumus F1-Score

    $$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$$

ROC-AUC

- ROC-AUC mengukur:

    > kemampuan model membedakan fraud dan normal.


- Interpretasi ROC-AUC

    | Nilai | Interpretasi |
    | ----- | ------------ |
    | 0.5   | random       |
    | 0.7   | cukup baik   |
    | 0.8   | baik         |
    | 0.9+  | sangat baik  |



### **Handling Imbalance Data**

#### **Oversampling**

##### **Introduction**

Oversampling adalah teknik untuk:

> menambah jumlah data kelas minoritas agar distribusi kelas lebih seimbang.

Contoh Imbalance

| Kelas  | Jumlah |
| ------ | ------ |
| Normal | 990    |
| Fraud  | 10     |

Model cenderung:

* lebih sering belajar pola normal,
* mengabaikan fraud.

Karena secara statistik:

> probabilitas melihat data normal jauh lebih besar.

Dasar Statistikanya

- Probabilitas suatu kelas:

    $$
    P(class)=\frac{jumlah\ class}{total\ data}
    $$

- Contoh

    Sebelum Oversampling

    Fraud:

    $$
    P(Fraud)=\frac{10}{1000}=0.01
    $$

    Normal:

    $$
    P(Normal)=\frac{990}{1000}=0.99
    $$

    Artinya:

    * 99% data yang dilihat model adalah normal,
    * hanya 1% fraud.

    Maka model “terbiasa” melihat normal.


- Model bisa belajar:

    > “Kalau saya selalu jawab normal, saya hampir selalu benar.”

Tujuan Oversampling

> Menaikkan representasi minoritas.

Contoh Setelah Oversampling

Misal fraud di-copy hingga 990 data.

| Kelas  | Jumlah |
| ------ | ------ |
| Normal | 990    |
| Fraud  | 990    |

Distribusi Baru

$$
P(Fraud)=\frac{990}{1980}=0.5
$$

$$
P(Normal)=\frac{990}{1980}=0.5
$$
Sekarang:

* model melihat fraud sama seringnya dengan normal,
* sehingga lebih “memperhatikan” fraud.

Cara Kerja Oversampling Sederhana

Misal data fraud:

| ID | Nilai |
| -- | ----- |
| F1 | 5000  |
| F2 | 6000  |

Oversampling bisa menjadi:

| ID      | Nilai |
| ------- | ----- |
| F1      | 5000  |
| F2      | 6000  |
| F1-copy | 5000  |
| F2-copy | 6000  |

Intuisi

Minoritas seperti:

* suara kecil di keramaian.

Oversampling:

> memperbesar jumlah suaranya.

Pengaruh ke Training

Karena data fraud muncul lebih sering:

* gradient update lebih sering dipengaruhi fraud,
* model lebih sensitif terhadap fraud.


Kelebihan

- Tidak membuang data  
   Semua data mayoritas tetap dipakai.

- Recall minoritas biasanya naik  
    Karena model lebih sering belajar pola minoritas.


Kekurangan

> Overfitting

Karena data hanya di-copy.

Model bisa:

* menghafal fraud,
* bukan belajar pola fraud.

Kalau fraud yang sama diulang terus:

* model terlalu hafal contoh tertentu,
* sulit generalisasi ke fraud baru.




Solusi yang Lebih Baik

Gunakan:

> SMOTE

Karena:

* membuat data sintetis,
* bukan sekadar copy-paste.

##### **SMOTE**

Synthetic Minority Oversampling Technique

- SMOTE adalah teknik oversampling yang:

    > membuat data minoritas baru secara sintetis.

    Jadi:

    * bukan copy-paste data,
    * tapi membuat “data baru yang mirip”.

- Oversampling biasa punya masalah:

    > hanya menggandakan data lama.

    Akibatnya:

    * model bisa menghafal,
    * overfitting.

    SMOTE mencoba:

    > membuat variasi baru.

- Contoh Imbalance

    | Kelas  | Jumlah |
    | ------ | ------ |
    | Normal | 990    |
    | Fraud  | 10     |
    
    SMOTE akan:
    
    * membuat fraud sintetis,
    * hingga distribusi lebih seimbang.

Ide Inti SMOTE

- SMOTE membuat titik baru:

    > di antara data minoritas yang mirip.


- Contoh Sederhana

    Data fraud:

    | Titik | X   | Y   |
    | ----- | --- | --- |
    | F1    | 100 | 200 |
    | F2    | 120 | 220 |

    SMOTE membuat titik baru:

    * di antara F1 dan F2.

- Dasar Statistika  
    SMOTE menggunakan:
    
    * nearest neighbors,
    * interpolasi linear.

Step-by-Step Matematikanya


- Ambil Titik Minoritas

    Misal:
    $$
    x_i=(100,200)
    $$


- Cari Tetangga Terdekat

    Misal nearest neighbor:
    $$
    x_{nn}=(120,220)
    $$


- Hitung Selisih
    $$
    x_{nn}-x_i
    $$
    $$
    =(120,220)-(100,200)
    $$
    $$
    =(20,20)
    $$

    Kita mencari:

    > arah menuju tetangga terdekat.


- Ambil Angka Random

    Misal:
    $$
    \lambda=0.3
    $$

    dengan:
    $$
    0 \leq \lambda \leq 1
    $$


    Kita memilih:

    > seberapa jauh bergerak menuju tetangga.


- Kalikan
    $$
    0.3 \times (20,20)
    $$
    $$
    =(6,6)
    $$


- Tambahkan ke Titik Awal
    $$
    (100,200)+(6,6)
    $$
    $$
    =(106,206)
    $$


- Hasil

    Data sintetis baru:
    $$
    (106,206)
    $$



Intuisi

SMOTE berkata:

> “Saya akan membuat fraud baru yang berada di sekitar fraud asli.”

Visualisasi Intuisi

Misal:

> ```text
> F1 ---------------- F2
> ```

SMOTE membuat titik:

* di tengah,
* atau dekat salah satu sisi.

Contoh:

> ```text
> F1 ----- NEW ------- F2
> ```

Formula Umum SMOTE

$$x_{new}=x_i+\lambda(x_{nn}-x_i)$$


Penjelasan Formula

| Simbol    | Arti                 |
| --------- | -------------------- |
| (x_i)     | titik minoritas awal |
| (x_{nn})  | nearest neighbor     |
| (\lambda) | angka random         |
| (x_{new}) | data sintetis baru   |


Kenapa Ini Bagus?

Karena:

* tidak sekadar copy data,
* membuat variasi,
* boundary model lebih general.

Kelebihan

- Mengurangi overfitting  
    Karena bukan duplicate murni.


- Boundary lebih baik  
   Minoritas jadi punya area representasi lebih luas.


Kekurangan

- Bisa membuat data aneh

    Kadang interpolasi menghasilkan:

    > kombinasi tidak realistis.


- Sensitif terhadap noise

    Kalau data minoritas jelek:

    > SMOTE bisa memperbanyak noise.


Pengaruh ke Training

Dengan data sintetis:

* model lebih sering melihat pola minoritas,
* tetapi tetap punya variasi.


Kesimpulan

- Bayangin ada sedikit contoh fraud.

    SMOTE mencoba berkata:

    > “Kalau fraud seperti ini ada, kemungkinan fraud lain di sekitar pola ini juga mungkin ada.”



- SMOTE tidak hanya:

    > memperbanyak minoritas,
    
    tetapi:
    
    > memperluas area representasi minoritas agar model lebih memahami pola mereka.

#### **Undersampling**

Konsep

Undersampling adalah teknik untuk:

> mengurangi jumlah data kelas mayoritas agar distribusi data lebih seimbang.

Contoh Imbalance

| Kelas  | Jumlah |
| ------ | ------ |
| Normal | 990    |
| Fraud  | 10     |

Masalah:

* model terlalu sering melihat normal,
* fraud hampir tidak diperhatikan.

Ide Utama Undersampling

Daripada:

* menambah fraud,

undersampling memilih:

> mengurangi data normal.

Contoh
- Sebelum

  | Kelas  | Jumlah |
  | ------ | ------ |
  | Normal | 990    |
  | Fraud  | 10     |

- Setelah Undersampling

  | Kelas  | Jumlah |
  | ------ | ------ |
  | Normal | 10     |
  | Fraud  | 10     |




Dasar Statistikanya

- Sebelum

    $$
    P(Fraud)=\frac{10}{1000}=0.01
    $$

    $$
    P(Normal)=\frac{990}{1000}=0.99
    $$


- Setelah Undersampling

    $$
    P(Fraud)=\frac{10}{20}=0.5
    $$

    $$
    P(Normal)=\frac{10}{20}=0.5
    $$


Keterangan:
- Awalnya:

  > model hampir selalu melihat normal.

- Setelah undersampling:

    > fraud dan normal muncul sama sering.

Cara Kerja

- Biasanya:

    > data mayoritas dipilih secara random untuk dibuang.


- Contoh

  Data normal:
    
    | ID |
    | -- |
    | N1 |
    | N2 |
    | N3 |
    | N4 |
    | N5 |
    | N6 |
    
    Fraud:
    | ID |
    | -- |
    | F1 |
    | F2 |
    
  
  Setelah Undersampling
  
  Normal yang dipilih random:
    
    | ID |
    | -- |
    | N2 |
    | N5 |
    
    Fraud tetap:
    
    | ID |
    | -- |
    | F1 |
    | F2 |


Pengaruh ke Training

Karena data normal dikurangi:

* model tidak terlalu bias ke mayoritas,
* lebih memperhatikan fraud.


Kelebihan
- Cepat

    > Dataset jadi lebih kecil.

    Training:
    
    * lebih ringan,
    * lebih cepat.

- Mengurangi bias mayoritas

    Karena distribusi lebih seimbang.

Kekurangan Besar

> Kehilangan Informasi

Karena:

> banyak data normal dibuang.

Padahal mungkin:

* data tersebut penting,
* punya pola unik.


Misal, kalau dari 990 orang normal, kamu hanya simpan 10:

> banyak variasi perilaku hilang.

Model bisa:

* kurang memahami pola normal,
* performa turun.



Kapan Cocok Dipakai?

Biasanya cocok jika:

* dataset sangat besar,
* mayoritas terlalu banyak,
* training terlalu berat.

Kombinasi yang Sering Dipakai

Kadang digunakan:

* undersampling ringan,
* lalu SMOTE.

Agar:

* tidak terlalu banyak data dibuang,
* tapi tetap balance.



Kesimpulan  

Undersampling mencoba membuat distribusi lebih adil dengan:

> mengurangi dominasi kelas mayoritas.

Tapi trade-off-nya:

> informasi dari mayoritas bisa hilang.

> Fraud Prediction
>       ↓

> ```
