# **Financial Analysis**

## **Introduction**

**Latar Belakang Masalah**  

Perusahaan bisa terlihat “sibuk”:

- banyak penjualan
- banyak pelanggan
- traffic tinggi

Tapi ada realita yang sering terjadi:

> Bisnis terlihat ramai, tapi sebenarnya tidak sehat secara finansial

Contoh:

- Revenue naik, tapi profit turun
- Banyak penjualan, tapi cash habis

Masalah utama:  
Tidak memahami kondisi keuangan secara utuh

**Tujuan**

- Tujuan Utama:

    > Menjaga bisnis tetap sehat dan berkelanjutan

- Tujuan Spesifik:
    - Mengetahui apakah bisnis untung atau rugi
    - Mengontrol pengeluaran
    - Memastikan cash cukup untuk operasional
    - Mendukung keputusan strategis (investasi, ekspansi)

**Konsep**  

Financial Analysis adalah:

> analisis data keuangan untuk memahami kondisi, performa, dan kesehatan bisnis.


**Kenapa Perlu Metrik?**

- Keuangan itu kompleks:

    - banyak jenis biaya
    - banyak sumber pendapatan
    - waktu pembayaran tidak selalu sama

- Tanpa metrik:

    Tidak bisa membedakan “ramai” vs “menguntungkan”

Jenis Financial Analysis

| Jenis                  | Fokus               |
|------------------- |---------------- |
| Revenue Analysis       | pendapatan          |
| Profitability Analysis | keuntungan          |
| Cost Analysis          | biaya               |
| Cash Flow Analysis     | arus kas            |
| Budget vs Actual       | realisasi vs target |
| Financial Forecasting  | prediksi keuangan   |

**Metrik Utama**

- Revenue (Pendapatan)

    - Definisi:
        > Total uang yang masuk dari penjualan

    - Intuisi:

        > Ini adalah “top line” — gambaran besar aktivitas bisnis

    - Tapi hati-hati:

        > Revenue tinggi ≠ profit tinggi

    - Kontribusi ke tujuan:

        ✔ Mengukur skala bisnis  
        ✔ Tracking pertumbuhan  

    $$ \ $$

- Profit Margin

    - Definisi:
        Persentase keuntungan setelah dikurangi biaya

    - Konsep:

        $$ \text{Profit Margin} = \frac{\text{Net Profit}}{\text{Revenue}} ×100%$$

    - Intuisi:

        > Seberapa “efisien” bisnis menghasilkan keuntungan

    - Contoh:

        - Margin 30% → sehat
        - Margin 5% → tipis, riskan

        Dua bisnis bisa punya revenue sama, tapi:

        yang margin tinggi jauh lebih sehat

    - Kontribusi ke tujuan:

        ✔ Menilai kesehatan finansial  
        ✔ Evaluasi efisiensi

    $$ \ $$

- Operating Expense (Biaya Operasional)

    - Definisi:  
        Biaya untuk menjalankan bisnis sehari-hari

    - Contoh:

        - gaji karyawan
        - sewa
        - marketing
        - teknologi

    - Intuisi:

        > Ini “kebocoran” utama uang

        Kalau terlalu besar:

        > profit akan tergerus

    - Kontribusi ke tujuan:

        ✔ Kontrol pengeluaran  
        ✔ Identifikasi pemborosan  

    $$ \ $$

- Cash Flow

    - Definisi:
        Arus uang masuk dan keluar

        Intinya:

        - Cash In (masuk)
        - Cash Out (keluar)

    - Intuisi:

        > Profit itu teori, cash itu realita

    - Contoh:

        > Kamu punya profit di laporan
        > Tapi uangnya belum dibayar customer

        Bisa bangkrut walau “profit”

    - Makna penting:

       - Cash positif → aman
       - Cash negatif → bahaya

    - Kontribusi ke tujuan:

        ✔ Menjaga kelangsungan bisnis  
        ✔ Perencanaan jangka pendek

**Hubungan Antar Metrik**  

Kasus Penting:
- Revenue Naik, Profit Turun

    - Ini contoh klasik  
    - Analisis  

        Kemungkinan:

        - Operating Expense naik drastis
        - Diskon terlalu besar
        - Cost produksi naik
    - Intuisi:

        > “Kita jual lebih banyak, tapi tidak lebih untung”  

    $$ \ $$

- Profit Ada, Tapi Cash Flow Negatif

    - Lebih berbahaya

    - Artinya:

        > Secara akuntansi untung
        > Tapi uang belum masuk
    - Intuisi:

        > “Kita terlihat kaya di atas kertas, tapi tidak punya uang”

    $$ \ $$
    
- Revenue Tinggi, Margin Rendah

    - Artinya:

        - Banyak penjualan
        - Tapi keuntungan kecil

    - Risiko:

        Sedikit perubahan biaya → langsung rugi

**Metrik → Insight Nyata**  

- Contoh Kasus:
    - Revenue naik
    - Profit margin turun
    - Operating expense naik
- Insight:
    - Pertumbuhan tidak efisien
    - Biaya operasional tidak terkontrol

**Insight → Keputusan**

Action yang bisa diambil:
- Audit biaya operasional
    - potong biaya tidak penting
    - optimasi proses
- Evaluasi strategi pricing
    > apakah terlalu banyak diskon?

- Tingkatkan efisiensi
    - otomatisasi
    - optimasi supply chain

**Intuisi**  

> Revenue itu “seberapa besar bisnis kamu”  
> Profit itu “seberapa sehat bisnis kamu”  
> Cash flow itu “apakah bisnis kamu bisa bertahan hidup”

**Alur Berpikir**
- Masalah:  
    Tidak tahu kondisi finansial sebenarnya
- Tujuan:  
    Menjaga kesehatan bisnis
- Solusi:  
    Gunakan metrik finansial
- Insight:  
    Temukan ketidakseimbangan (revenue vs cost vs cash)
- Action:  
    Optimasi biaya & strategi

**Rangkuman**  

| Metrik            | Menjawab                 |
| ----------------- | ------------------------ |
| Revenue           | Seberapa besar bisnis    |
| Profit Margin     | Seberapa sehat           |
| Operating Expense | Seberapa boros           |
| Cash Flow         | Bisa bertahan atau tidak |


## **Revenue Analysis**

### **Introduction**

Revenue Analysis adalah proses menganalisis pendapatan bisnis untuk memahami:

- dari mana revenue berasal,
- faktor yang memengaruhi revenue,
- produk/customer/channel terbaik,
- dan bagaimana meningkatkan pendapatan bisnis.

Tujuan utamanya adalah:

> memahami bagaimana bisnis menghasilkan uang.

Kenapa Revenue Analysis Penting?

> Revenue adalah salah satu KPI utama perusahaan.

Tanpa revenue:

- bisnis tidak bisa bertahan,
- operasional tidak berjalan,
- profit tidak mungkin terjadi.

Revenue analysis membantu perusahaan:

- memonitor performa bisnis,
- mencari sumber revenue terbesar,
- memahami perilaku customer,
- mengevaluasi strategi marketing,
- membuat forecasting.

Pertanyaan Umum dalam Revenue Analysis

- Revenue naik atau turun?
- Produk paling menghasilkan?
- Customer paling profitable?
- Channel marketing terbaik?
- Revenue tertinggi terjadi kapan?
- Faktor apa yang memengaruhi penjualan?
- Apakah campaign meningkatkan revenue?

Metric Penting

| Metric               | Fungsi              |
|----------------- |---------------- |
| Total Revenue        | total pendapatan    |
| Revenue Growth       | pertumbuhan revenue |
| Revenue per Customer | revenue/customer    |
| Revenue per Product  | revenue produk      |

### **Metrics**





Rumus 

- Revenue  
    Total pendapatan dari seluruh penjualan.

    $$
    Revenue =
    Price \times Quantity
    $$

    Contoh

    | Harga  | Qty |
    |--------|-----|
    | 50.000 | 100 |

    Hitung

    $$

    5.000.000
    $$

- Revenue Growth  
    Mengukur pertumbuhan revenue antar periode.

    Rumus

    $$
    Growth =
    \frac{
    Current - Previous
    }{
    Previous
    }
    \times100%
    $$

    Contoh

    | Bulan    | Revenue  |
    |----- |----- |
    | Januari  | 100 juta |
    | Februari | 120 juta |

    Hitung

    $$

    \frac{
    120-100
    }{
    100
    }
    \times100\%
    =

    20\%
    $$

    Interpretasi

    > Revenue naik 20%.

- Revenue per Customer  
  Rata-rata revenue yang dihasilkan tiap customer.  

  $$\text{Revenue per Customer} = \frac{\text{Total Revenue}}{\text{Jumlah Customer}}$$

  Insight
    - Mengetahui nilai rata-rata customer.
    - Membantu segmentasi high-value customer.

- Revenue per Product  
    Revenue yang dihasilkan tiap produk. 

## **Profitability Analysis**

### **Introduction**

Konsep

- Profitability Analysis adalah proses menganalisis kemampuan bisnis dalam menghasilkan keuntungan.

- Fokus utamanya bukan hanya:

    > revenue besar,

    melainkan:

    > apakah bisnis benar-benar untung.

- Tujuan utamanya:

    > memahami seberapa efisien bisnis menghasilkan profit.




Kenapa Profitability Analysis Penting?

- Revenue tinggi belum tentu berarti bisnis sehat.

    Contoh:

    > revenue Rp1 miliar, tapi biaya Rp980 juta.

    Artinya profit sangat kecil.

- Profitability analysis membantu perusahaan:

    - mengukur kesehatan bisnis,
    - memahami efisiensi operasional,
    - mengevaluasi biaya,
    - menentukan strategi pricing,
    - mencari produk/customer paling profitable.

Pertanyaan Umum  

Beberapa pertanyaan yang sering dijawab:

- Bisnis benar-benar untung?
- Margin bisnis sehat?
- Produk paling profitable?
- Cost terlalu besar?
- Customer tertentu menghasilkan profit lebih besar?
- Profit meningkat atau turun?
- Revenue naik tapi profit turun kenapa?

Metric Penting

| Metric        | Fungsi                |
| ------------- | --------------------- |
| Gross Profit  | laba kotor            |
| Net Profit    | laba bersih           |
| Profit Margin | persentase keuntungan |

Gambaran Besar

Profitability analysis membantu bisnis memahami:

* apakah bisnis benar-benar untung,
* apakah biaya terlalu besar,
* seberapa efisien operasional bisnis,
* dan apakah pertumbuhan revenue benar-benar menghasilkan profit.

Intinya:

> “Revenue besar belum tentu bisnis sehat, yang penting adalah profitnya.”

### **Metrics**

Metric yang sering digunakan


- Gross Profit

    > Keuntungan setelah dikurangi biaya produksi langsung.


    Rumus

    $$
    \text{Gross\ Profit =
    Revenue - COGS}
    $$


    Keterangan

    | Variabel | Arti                  |
    | -------- | --------------------- |
    | Revenue  | total pendapatan      |
    | COGS     | biaya produksi/barang |


    Contoh

    | Revenue  | COGS    |
    | -------- | ------- |
    | 100 juta | 60 juta |

    Hitung:

    $$
    100 - 60
    =
    40\ juta
    $$


    Interpretasi

    > Setelah biaya produksi, bisnis masih mendapat laba Rp40 juta.


    Insight

    * Mengukur efisiensi produksi.
    * Mengetahui apakah harga jual cukup menguntungkan.


- Net Profit


    > Keuntungan akhir setelah semua biaya dikurangi.

    Biaya bisa meliputi:

    * operasional,
    * marketing,
    * gaji,
    * pajak,
    * dll.


    Rumus

    $$
    \text{Net\ Profit =
    Revenue - Total\ Expense}
    $$


    Contoh

    | Revenue  | Total Expense |
    | -------- | ------------- |
    | 100 juta | 85 juta       |

    Hitung:

    $$
    100 - 85
    =
    15\ juta
    $$


    Interpretasi

    > Keuntungan bersih bisnis adalah Rp15 juta.


    Insight

    * Mengukur keuntungan nyata perusahaan.
    * Revenue besar belum tentu net profit besar.


- Profit Margin


    > Persentase keuntungan dibanding revenue.


    Rumus

    $$
    \text{Profit\ Margin} =
    \frac{
    \text{Profit}
    }{
    \text{Revenue}
    }
    \times100%
    $$


    Contoh

    | Revenue  | Profit  |
    | -------- | ------- |
    | 100 juta | 20 juta |

    Hitung:

    $$
    \frac{
    20
    }{
    100
    }
    \times100%
    =
    20%
    $$


    Interpretasi

    > Setiap Rp100 revenue menghasilkan Rp20 profit.


    Insight

    * Mengukur kesehatan bisnis.
    * Margin tinggi biasanya menunjukkan bisnis lebih efisien.
    * Digunakan untuk membandingkan performa antar produk atau perusahaan.





## **Cost Analysis**

### **Introduction**

Konsep

> Cost Analysis adalah proses menganalisis seluruh pengeluaran bisnis untuk memahami:

- uang keluar untuk apa,
- biaya terbesar berasal dari mana,
- apakah biaya masih efisien,
- dan bagaimana mengontrol pengeluaran.

Tujuan utamanya:  
> memahami dan mengoptimalkan pengeluaran bisnis.

Kenapa Penting?  

Revenue besar belum tentu menghasilkan profit besar.

Karena:

$$\text{Profit=Revenue−Cost}$$

Jika cost terlalu besar:

- profit kecil,
- bahkan bisnis bisa rugi.

Cost analysis membantu perusahaan:

- mengontrol pengeluaran,
- meningkatkan efisiensi,
- mengurangi pemborosan,
- menentukan harga produk,
- meningkatkan profit.

Pertanyaan Umum dalam Cost Analysis

Beberapa pertanyaan yang sering dijawab:

- Biaya terbesar berasal dari mana?
- Cost meningkat kenapa?
- Operasional sudah efisien?
- Produk mana paling mahal diproduksi?
- Marketing terlalu mahal?
- Delivery cost meningkat?
- Cost memengaruhi profit?

Konsep Dasar Cost

Cost adalah seluruh pengeluaran yang dibutuhkan bisnis.

Secara sederhana:

$$\text{Total Cost=Fixed Cost+Variable Cost+Operational Cost}$$


Jenis Cost

| Jenis            | Definisi                                        | Contoh                       |
| ---------------- | ----------------------------------------------- | ---------------------------- |
| Fixed Cost       | biaya tetap meski penjualan berubah             | gaji, sewa                   |
| Variable Cost    | biaya yang berubah mengikuti produksi/penjualan | bahan baku, packaging        |
| Operational Cost | biaya operasional bisnis sehari-hari            | marketing, listrik, internet |



Metric Penting

| Metric                 | Fungsi                     |
| ---------------------- | -------------------------- |
| Total Cost             | total seluruh biaya        |
| Cost Growth            | pertumbuhan biaya          |
| Cost per Unit          | biaya per produk           |
| Operational Cost Ratio | proporsi biaya operasional |

Insight Contoh

| Temuan                   | Insight                         |
| ------------------------ | ------------------------------- |
| Marketing cost tinggi    | campaign kurang efisien         |
| Delivery cost naik       | masalah supply chain            |
| Cost per unit turun      | produksi makin efisien          |
| Operational ratio tinggi | biaya operasional terlalu besar |

Gambaran Besar

Cost analysis membantu bisnis memahami:

* uang habis di mana,
* biaya mana yang paling besar,
* apakah operasional efisien,
* dan bagaimana mengoptimalkan profit.

Intinya:

> “Bukan hanya mencari revenue besar, tapi juga mengontrol pengeluaran.”

### **Metrics**

Metric yang sering digunakan

- Total Cost

    > Total seluruh pengeluaran bisnis.

    Rumus

    $$
    \text{Total\ Cost} =
    \text{Fixed\ Cost + Variable\ Cost + Operational\ Cost}
    $$

    Insight

    * Mengetahui total pengeluaran perusahaan.
    * Dasar analisis profit.

- Cost Growth

    > Mengukur kenaikan biaya antar periode.

    Rumus

    $$
    \text{Cost\ Growth} =
    \frac{\text{
    Current\ Cost - Previous\ Cost
    }}{
    \text{Previous\ Cost}
    }
    \times100%
    $$

    Contoh

    | Bulan    | Cost    |
    | -------- | ------- |
    | Januari  | 50 juta |
    | Februari | 65 juta |

    Hitung:

    $$
    \frac{
    65-50
    }{
    50
    }
    \times100%
    =
    30%
    $$

    Interpretasi

    > Cost meningkat 30%.

    Insight

    * Membantu mendeteksi pemborosan.
    * Kenaikan cost belum tentu buruk jika revenue ikut naik.

- Cost per Unit  
  
    > Biaya rata-rata untuk menghasilkan satu produk.
  
    Rumus

    $$
    \text{Cost\ per\ Unit} =
    \frac{
    \text{Total\ Cost}
    }{
    \text{Jumlah\ Produk}
    }
    $$

        Contoh

    | Total Cost | Produk |
    | ---------- | ------ |
    | 10 juta    | 500    |

    Hitung:

    $$
    \frac{
    10.000.000
    }{
    500
    }
    =
    20.000
    $$

    Interpretasi

    > Biaya produksi tiap produk adalah Rp20.000.

    Insight

    * Membantu menentukan harga jual.
    * Mengukur efisiensi produksi.

- Operational Cost Ratio
  > Persentase biaya operasional terhadap revenue.

    Rumus

    $$
    \text{Operational\ Cost\ Ratio} =
    \frac{
    \text{Operational\ Cost}
    }{
    \text{Revenue}
    }
    \times100\%
    $$

    Insight

    * Mengukur efisiensi operasional bisnis.
    * Ratio terlalu tinggi bisa menekan profit.

- Cost Contribution  
    > Kontribusi suatu jenis biaya terhadap total cost.

    Rumus
    $$\text{Cost Contribution}=\frac{\text{Jenis Cost}}{\text{Total Cost}}×100%$$


    Insight

    Mengetahui cost terbesar dalam bisnis.






## **Cash Flow Analysis**

### **Introduction**

Konsep

> Cash Flow Analysis adalah proses menganalisis arus uang masuk dan keluar perusahaan.

Fokus utamanya:

- apakah perusahaan memiliki cukup uang,
- uang masuk berasal dari mana,
- uang keluar digunakan untuk apa,
- dan apakah bisnis mampu bertahan secara finansial.

Tujuan utamanya:

> memahami kondisi cash nyata perusahaan.

Kenapa Cash Flow Analysis Penting?

Bisnis bisa:

> profit, tetapi tetap kehabisan uang.

Contoh:

> customer belum membayar,
> tetapi perusahaan tetap harus membayar gaji dan operasional.

Karena itu:

> profit ≠ cash.

Cash flow analysis membantu perusahaan:

- menjaga likuiditas,
- menghindari kekurangan cash,
- merencanakan pengeluaran,
- mengontrol operasional,
- mendukung keputusan investasi.

Pertanyaan Umum dalam Cash Flow Analysis

Beberapa pertanyaan yang sering dijawab:

- Cash flow positif atau negatif?
- Perusahaan kekurangan cash?
- Operasional menghasilkan cash?
- Pengeluaran terlalu besar?
- Perusahaan terlalu bergantung pada pinjaman?
- Cash cukup untuk ekspansi?
- Kapan cash flow mulai bermasalah?

Jenis Cash Flow

| Jenis               | Definisi                         | Contoh                   |
| ------------------- | -------------------------------- | ------------------------ |
| Operating Cash Flow | cash dari aktivitas utama bisnis | penjualan produk         |
| Investing Cash Flow | cash untuk investasi             | beli mesin, aset         |
| Financing Cash Flow | cash dari pendanaan              | pinjaman, modal investor |


Metric Penting

| Metric              | Fungsi                        |
| ------------------- | ----------------------------- |
| Net Cash Flow       | total arus kas bersih         |
| Operating Cash Flow | cash dari operasional         |
| Free Cash Flow      | cash yang benar-benar tersisa |
| Cash Flow Growth    | pertumbuhan cash flow         |

Insight Contoh

| Temuan                          | Insight                         |
| ------------------------------- | ------------------------------- |
| Profit tinggi tapi cash negatif | pembayaran customer lambat      |
| Operating cash flow negatif     | operasional tidak sehat         |
| Financing cash flow tinggi      | terlalu bergantung pinjaman     |
| Free cash flow besar            | perusahaan punya ruang ekspansi |


Gambaran Besar

Cash flow analysis membantu bisnis memahami:

* apakah perusahaan punya cukup uang,
* uang masuk dari mana,
* uang keluar untuk apa,
* dan apakah bisnis bisa bertahan secara finansial.

Intinya:

> “Profit itu akuntansi, cash flow itu kenyataan uang di tangan.”

### **Metrics**








Metric



- Net Cash Flow  

    Selisih total uang masuk dan uang keluar.


    Rumus

    $$
    \text{Net\ Cash\ Flow} =
    \text{Cash\ Inflow - Cash\ Outflow}
    $$
    
    
    Contoh
    
    | Cash In  | Cash Out |
    | -------- | -------- |
    | 100 juta | 80 juta  |
    
    Hitung:
    
    $$
    100 - 80
    =
    20\ juta
    $$
    
    
    Interpretasi
    
    > Perusahaan memiliki cash flow positif Rp20 juta.


    Insight

    * Positif → cash bertambah.
    * Negatif → cash berkurang.


- Operating Cash Flow  

    Cash yang dihasilkan dari aktivitas utama bisnis.


    Rumus Sederhana

    $$
    \text{Operating\ Cash\ Flow} =
    \text{Cash\ dari\ Penjualan - Operational\ Expense}
    $$


    Insight

    * Mengukur kesehatan operasional bisnis.
    * Jika negatif terus, bisnis bisa bermasalah.


- Free Cash Flow (FCF)  

    > Cash yang tersisa setelah biaya operasional dan investasi.


    Rumus

    $$
    \text{Free\ Cash\ Flow} =
    \text{Operating\ Cash\ Flow - Capital\ Expenditure}
    $$


    Keterangan

    | Variabel            | Arti                 |
    | ------------------- | -------------------- |
    | Capital Expenditure | biaya investasi/aset |


    Insight

    * Cash yang benar-benar bisa dipakai perusahaan.
    * Penting untuk ekspansi atau pembayaran utang.


- Cash Flow Growth  

    Pertumbuhan cash flow antar periode.


    Rumus

    $$
    \text{Cash\ Flow\ Growth} =
    \frac{
    \text{Current - Previous
    }}{
    \text{Previous}
    }
    \times100\%
    $$


    Insight

    * Mengukur perkembangan kondisi cash perusahaan.
    * Growth positif biasanya menandakan bisnis makin sehat.





## **Budget vs Actual Analysis**

### **Introduction**


Konsep

Budget vs Actual Analysis adalah proses membandingkan:

- target/rencana (budget),
- dengan hasil realisasi sebenarnya (actual).

Tujuan utamanya:

> mengetahui apakah performa bisnis sesuai target.

Kenapa Budget vs Actual Analysis Penting?

Perusahaan selalu membuat:

- target revenue,
- target profit,
- target cost,
- target penjualan.

Namun realisasi di lapangan bisa berbeda.

Analysis ini membantu perusahaan:

- mengevaluasi performa,
- mengontrol pengeluaran,
- mengukur efektivitas strategi,
- memperbaiki perencanaan,
- mendeteksi masalah lebih cepat.

Pertanyaan Umum dalam Budget vs Actual Analysis

Beberapa pertanyaan yang sering dijawab:

- Target tercapai?
- Revenue di bawah target?
- Cost melebihi budget?
- Divisi mana paling efisien?
- Variance terlalu besar?
- Forecast terlalu optimis?
- Pengeluaran masih terkendali?

Insight Contoh

| Temuan               | Insight                   |
| -------------------- | ------------------------- |
| Actual < Budget      | target tidak tercapai     |
| Actual > Budget      | performa melebihi target  |
| Cost melebihi budget | pengeluaran tidak efisien |
| Variance kecil       | perencanaan cukup akurat  |

Gambaran Besar

Budget vs Actual Analysis membantu bisnis memahami:
- apakah target tercapai,
- area yang over budget,
- performa bisnis terhadap rencana,
- dan kualitas perencanaan perusahaan.

Intinya:

> “Seberapa dekat realisasi dengan target yang direncanakan?”

### **Metrics**

Contoh

| Metric  | Budget   | Actual  |
| ------- | -------- | ------- |
| Revenue | 100 juta | 90 juta |

Perhitungan Metric  

- Variance

    > Selisih antara actual dan budget.

    Rumus

    $$
    \text{Variance =Actual - Budget}
    $$

    Hitung

    $$90 - 100=-10\ \text{juta}
    $$

    Interpretasi

    > Target revenue tidak tercapai sebesar Rp10 juta.

    Insight

    * Variance negatif → performa di bawah target.
    * Variance positif → melebihi target.

- Variance Percentage

    > Persentase selisih dibanding budget.

    Rumus

    $$
    \text{Variance \%} =
    \frac{
    \text{Actual - Budget}
    }{
    \text{Budget}
    }
    \times100\%
    $$

    Hitung

    $$
    \frac{90-100}{100}\times100\%=-10\%
    $$

    Interpretasi

    > Revenue 10% di bawah target.

    Insight

    * Memudahkan membandingkan performa antar divisi/periode.
    * Membantu evaluasi target bisnis.


- Budget Achievement  
  
  > Persentase pencapaian target.

    Rumus
    $$\text{Budget Achievement} = \frac{\text{Actual}}{\text{Budget}}×100\%$$


## **Financial Forecasting**

### **Introduction**


Konsep

> Financial Forecasting adalah proses memprediksi kondisi finansial bisnis di masa depan menggunakan data historis.

Biasanya digunakan untuk memprediksi:

- revenue,
- sales,
- profit,
- cash flow,
- demand,
- biaya.

Tujuan utamanya:

> membantu bisnis membuat keputusan berdasarkan prediksi masa depan.

Kenapa Financial Forecasting Penting?

Bisnis perlu mengetahui kemungkinan kondisi masa depan untuk:

- membuat target,
- mengatur budget,
- menyiapkan stok,
- mengurangi risiko,
- menentukan strategi bisnis.

Tanpa forecasting:

> keputusan lebih banyak berdasarkan tebakan.

Pertanyaan Umum dalam Financial Forecasting

Beberapa pertanyaan yang sering dijawab:

- Revenue bulan depan berapa?
- Sales akan naik atau turun?
- Kapan peak demand terjadi?
- Cash flow aman?
- Stok perlu ditambah?
- Profit akan meningkat?
- Apakah bisnis mengalami seasonality?

Dasar Financial Forecasting

Forecasting menggunakan:

- pola historis,
- trend,
- seasonality,
- siklus,
- dan noise.

Konsep sederhananya:

> data masa lalu digunakan untuk memperkirakan masa depan.

Komponen Time Series dalam Forecasting
|Komponen	|Definisi|
|-----------|--------|
|Trend	    |arah naik/turun jangka panjang|
|Seasonality	|pola berulang periodik|
|Cyclic	    |siklus ekonomi/bisnis|
|Noise	    |data acak|

### **Metrics**

Metrik yang digunakan
- Forecast Value

    > Nilai hasil prediksi model.

    Contoh

    | Bulan | Forecast Sales |
    | ----- | -------------- |
    | Mei   | 150 juta       |

    Insight

    > Model memprediksi sales bulan Mei sebesar Rp150 juta.

- Forecast Error


    > Selisih antara nilai aktual dan prediksi.

    Rumus

    $$
    Forecast\ Error =
    Actual - Forecast
    $$

    Contoh

    | Actual   | Forecast |
    | -------- | -------- |
    | 160 juta | 150 juta |

    Hitung:

    $$
    160 - 150
    =

    10\ juta
    $$

    Insight

    * Error kecil → prediksi bagus.
    * Error besar → model kurang akurat.

- Mean Absolute Error (MAE)


    > Rata-rata selisih absolut prediksi.

    Rumus

    $$
    MAE =
    \frac{1}{n}
    \sum |y - \hat{y}|
    $$

    Insight

    * Mudah dipahami.
    * Semakin kecil semakin baik.

- Mean Squared Error (MSE)


    > Rata-rata error kuadrat.

    Rumus

    $$
    MSE =
    \frac{1}{n}
    \sum (y - \hat{y})^2
    $$

    Insight

    * Menghukum error besar lebih kuat.
    * Sensitif terhadap outlier.

- Root Mean Squared Error (RMSE)


    > Akar dari MSE.

    Rumus

    $$
    RMSE =
    \sqrt{MSE}
    $$

    Insight

    * Satuan kembali ke data asli.
    * Sangat umum dipakai dalam forecasting.

- Mean Absolute Percentage Error (MAPE)


    > Error prediksi dalam bentuk persen.

    Rumus

    $$
    MAPE =
    \frac{100}{n}
    \sum
    \left|
    \frac{
    y-\hat{y}
    }{
    y
    }
    \right|
    $$

    Insight

    * Mudah diinterpretasikan.
    * Cocok untuk membandingkan model.

    Contoh:

    > MAPE 10% berarti rata-rata error prediksi sebesar 10%.
