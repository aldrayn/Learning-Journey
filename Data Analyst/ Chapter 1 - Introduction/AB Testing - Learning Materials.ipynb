# **A/B Testing**

## **Introduction**

**Masalah**

> Perusahaan ingin mengetahui apakah perubahan tertentu benar-benar memberikan peningkatan performa atau tidak.

Contoh:

- Tombol checkout baru
- Landing page baru
- Algoritma rekomendasi baru
- Harga baru
- Email marketing baru

Masalah utamanya:

> “Apakah perubahan ini benar-benar lebih baik, atau cuma kebetulan saja?”


**Tujuan**

Membandingkan:

- Control (A) → versi lama
- Treatment (B) → versi baru

Untuk menentukan:

- apakah ada peningkatan,
- seberapa besar peningkatannya,
- dan apakah peningkatan tersebut signifikan secara statistik.

**Pendekatan** 

| Pendekatan           | Fokus                                      |
|----------------- |--------------------------------------- |
| Descriptive Analysis | melihat perubahan/performa                 |
| Statistical Analysis | memastikan perubahan signifikan atau tidak |


## **Metric**

#### **Conversion Rate (CR)**  



> Mengukur persentase user yang melakukan aksi tertentu.

Contoh:

- beli produk,
- klik tombol,
- subscribe,
- register.

Rumus  

$$CR = \frac{\text{Jumlah Conversion}}{\text{Jumlah User}}$$

Contoh

- 1000 user masuk.
- 50 membeli produk.  
- Maka conversion ratenya adalah:  
    
  $CR = \frac{50}{1000} = 0.05 = 5\% $



#### **Click Through Rate (CTR)**



> Mengukur persentase klik terhadap jumlah tampilan.

Rumus

$$
CTR = \frac{\text{Clicks}}{\text{Impressions}}
$$


Contoh

- 10000 orang melihat iklan.
- 500 klik.
- Maka CTR-nya adalah:  
    
    $
    CTR = \frac{500}{10000} = 5%
    $


#### **Bounce Rate**



> Persentase user yang langsung keluar tanpa interaksi.

Rumus

$$
\text{Bounce\ Rate} = \frac{\text{Single Page Session}}{\text{Total Session}}
$$

Contoh  
- 1000 user masuk website.
- 400 langsung keluar tanpa klik apa pun.
- Maka Bounce Ratenya adalah :
  
  $\text{Bounce\ Rate} = \frac{400}{1000} = 0.4 = 40 \%$

Interpretasi
- Bounce rate tinggi → landing page kurang menarik
- Bounce rate rendah → user tertarik menjelajah

#### **Revenue Per User (RPU)**



> Rata-rata revenue per user.

Rumus

$$
RPU = \frac{\text{Total Revenue}}{\text{Total User}}
$$

Contoh  
- Total revenue:  
    
    $Rp50.000.000$

- Jumlah user:

    $1000$

- Maka:  
  
  $\text{RPU} = \frac{50.000.000}{1000} = 50.000$  

Interpretasi

- Rata-rata setiap user menghasilkan revenue Rp50.000.

#### **Average Order Value (AOV)**



> Rata-rata nilai transaksi.

Rumus

$$
AOV = \frac{\text{Total Revenue}}{\text{Jumlah Order}}
$$

Contoh

- Total revenue:

    $Rp50.000.000$

- Jumlah order:

    $250$

- Maka:  
  
    $$
    AOV = \frac{50.000.000}{250} = 200.000
    $$

Interpretasi

- Rata-rata setiap transaksi bernilai Rp200.000.

#### **Retention Rate**



> Mengukur berapa user yang kembali menggunakan produk.

Rumus

$$
Retention\ Rate =
\frac{\text{User Aktif Kembali}}
{\text{Total User Awal}}
$$

Contoh

- Awalnya ada:  
    1000 user.  
- Sebulan kemudian, 700 user masih aktif.  
- Maka:  
    
    $
    \text{Retention\ Rate} =
    \frac{700}
    {1000} = 0.7 = 70\%
    $

Interpretasi
- 70% user tetap menggunakan produk.

#### **Churn Rate**



> Mengukur persentase user yang berhenti menggunakan produk.

Rumus  

$$\text{Churn\ Rate} =
\frac{\text{User Hilang}}
{\text{Total User}}$$  

Contoh

- Total user awal adalah 1000
- 300 user berhenti menggunakan produk.  
- Maka:  
  
  $\text{Churn\ Rate} =
    \frac{300}
    {1000} = 0.3 = 30\%$

#### **Lift**  



> Mengukur peningkatan treatment dibanding control.

Rumus

$$
Lift =
\frac{Metric_B - Metric_A}
{Metric_A}
\times 100%
$$

Contoh

- Control:

    $
    CR_A = 5%
    $

- Treatment:

    $
    CR_B = 6%
    $

- Maka:

    $
    Lift =
    \frac{6 - 5}{5} \times 100%
    = 20%
    $

    Artinya treatment meningkatkan conversion sebesar 20%.

## **Descriptive Analysis**

**Descriptive Analysis**

Fokus:

> memahami performa dan perubahan metric.

Biasanya:

- belum melakukan uji statistik,
- hanya melihat hasil secara deskriptif.
- Contoh Pertanyaan
- Conversion naik atau turun?
- Revenue bertambah?
- Bounce rate membaik?
- CTR meningkat?

Metrik yang umum dipakai:  

| Metric          | Fungsi                    |
|------------ |---------------------- |
| Conversion Rate | persentase conversion     |
| CTR             | persentase click          |
| Bounce Rate     | persentase keluar         |
| Revenue         | pendapatan                |
| AOV             | rata-rata nilai transaksi |

Kelemahan:
> Belum bisa memastikan "apakah peningkatan benar-benar nyata atau cuma kebetulan/random".

## **Statistical Analysis**

### **Introduction**

**Statistical Analysis**

Fokus:

> memastikan apakah perbedaan benar-benar signifikan secara statistik.

Pertanyaan Utama

> “Apakah kenaikan ini cukup besar dibanding noise data?”

Yang Biasanya Dipakai

| Statistik           | Fungsi                |
|---------------- |------------------ |
| P-value             | mengukur signifikansi |
| Confidence Interval | range estimasi        |
| Effect Size         | besar dampak          |
| Hypothesis Testing  | pengujian statistik   |


### **Example**

#### **Situation 1 - New Checkout Button**

Permintaan Atasan

> “Tim product membuat tombol checkout baru.
> Coba cek apakah conversion meningkat atau tidak.”


Tujuan Bisnis

> Meningkatkan checkout conversion.


Metric Utama - Conversion Rate

$$
CR =
\frac{Conversion}{Total\ User}
$$

Jalankan Experiment

| Group | Deskripsi   |
|-- |-------- |
| A     | Tombol lama |
| B     | Tombol baru |


Hasil Data

| Group | User | Conversion |
|-- |- |------- |
| A     | 1000 | 50         |
| B     | 1000 | 70         |


Hitung Conversion Rate

- Group A

    $
    CR_A =\frac{50}{1000}=5\%
    $


- Group B

    $
    CR_B =\frac{70}{1000}=7%
    $


Hitung Lift

$$
Lift =\frac{7\%-5\%}{5\%}\times 100\%=40%
$$


Interpretasi
> Versi B meningkatkan conversion sebesar 40%.


Confidence Interval

Misalnya:

| Group | CI          |
|-- |-------- |
| A     | 3.6% – 6.3% |
| B     | 5.4% – 8.5% |



Kemungkinan conversion rate asli:

> Group B memang lebih tinggi.


Uji Statistik

- Pakai:

    > Z-Test atau Chi-Square.


- Hasil

    $$
    p = 0.03
    $$


- Interpretasi

    Karena:
    
    $$
    p < 0.05
    $$
    
    maka:
    
    * hasil signifikan,
    * kemungkinan besar perubahan memang berhasil.


Kemudian dihitung effect size-nya, misalnya:

$$
d = 0.45
$$

Efek termasuk:

> sedang/moderat.

Decision

> Deploy tombol baru.

Karena:

* conversion naik,
* signifikan,
* dan efeknya cukup meaningful.

#### **Situation 2 - New Landing Page**


Permintaan Atasan

> “Marketing membuat landing page baru.
> Apakah user jadi lebih sering klik register?”


Tujuan Bisnis
> Meningkatkan CTR (Click Through Rate).


Metric - CTR

$$
CTR =
\frac{
Clicks
}{
Impressions
}
$$


Data

| Group | Impression | Click |
|-- |------- |-- |
| A     | 5000       | 300   |
| B     | 5000       | 420   |


Hitung CTR

- Group A

    $
    CTR_A =\frac{300}{5000}=6\%
    $


- Group B

    $CTR_B =\frac{420}{5000}=8.4\%
    $


Lift

$$
\frac{8.4-6}{6}\times100%=40%
$$


P-value

Misalnya:
$$
p = 0.01
$$


Interpretasi
> CTR meningkat signifikan.


Decision

✅ Gunakan landing page baru.

#### **Situation 3 - New Discount**

Permintaan Atasan

> “Apakah diskon baru meningkatkan revenue?”

Tujuan Bisnis

> Meningkatkan average revenue per user.


Metric - Revenue Per User

$$
RPU =
\frac{
Revenue
}{
Total\ User
}
$$


Data

| Group | Revenue | User |
|-- |---- |- |
| A     | 50 juta | 1000 |
| B     | 62 juta | 1000 |


Hitung RPU

- Group A

    $$
    RPU_A =\frac{50jt}{1000}=50rb
    $$


- Group B

    $$
    RPU_B =
    62rb
    $$


Statistical Test

Karena:

* data numerik,
* membandingkan rata-rata,

pakai:

* T-Test.


Hasil

$$
p = 0.04
$$


Effect Size

$$
d = 0.3
$$


Interpretasi

* signifikan,
* efek kecil-menengah.


Decision

✅ Diskon baru cukup efektif.
