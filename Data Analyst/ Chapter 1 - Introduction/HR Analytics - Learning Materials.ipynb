# **HR Analytics**

## **Introduction**

**Apa Itu HR Analytics?**

HR Analytics adalah:

> analisis data karyawan untuk membantu pengambilan keputusan dalam Human Resources (HR).


**Latar Belakang**

Perusahaan sering melihat:

* jumlah karyawan
* daftar absensi
* hasil penilaian

Tapi…

> Sulit menjawab: “Apakah tim kita sehat dan produktif?”

Masalah umum:

* Turnover tinggi tanpa tahu penyebab
* Karyawan terlihat hadir, tapi tidak produktif
* Performa tidak merata antar tim

Masalah utama:  
**Kurangnya visibility terhadap kondisi dan dinamika karyawan**

**Contoh Pertanyaan Bisnis**

* Kenapa banyak karyawan resign?
* Divisi mana turnover tertinggi?
* Apakah training meningkatkan performa?
* Bagaimana produktivitas karyawan?
* Faktor apa memengaruhi retention?

**Tujuan**

- Tujuan Utama:

    > Menciptakan tenaga kerja yang stabil, produktif, dan puas

- Tujuan Spesifik:

    - Memantau performa karyawan
    - Mengidentifikasi risiko turnover
    - Meningkatkan produktivitas
    - Mendukung keputusan HR (promosi, pelatihan, dll)

**Kenapa Perlu Metrik?**

Hal seperti:

* “karyawan puas”
* “karyawan perform bagus”

Itu subjektif

Maka:

> Kita ubah jadi angka → agar bisa dibandingkan, dipantau, dan dianalisis

Kenapa HR Analytics Penting?

Karena:

* kehilangan karyawan mahal,
* hiring membutuhkan biaya besar,
* produktivitas memengaruhi bisnis.

Jenis HR Analytics

| Jenis                       | Fokus                 |
|------------------------ |------------------ |
| Employee Analysis           | data karyawan         |
| Attrition/Turnover Analysis | resign/churn karyawan |
| Performance Analysis        | performa kerja        |
| Recruitment Analysis        | hiring                |
| Compensation Analysis       | gaji                  |
| Workforce Analysis          | struktur tenaga kerja |

## **Metrics**

### **Introduction**

Metrik


- Employee Turnover Rate

    **Definisi:**    
    Persentase karyawan yang keluar dalam periode tertentu

    Konsep:

    $$Turnover\ Rate = \frac{\text{Employees Leaving}}{\text{Total Employees}} \times 100\% $$


    **Intuisi:**

    > Seberapa “stabil” tim kamu

    * Tinggi → banyak yang keluar ❌
    * Rendah → tim stabil ✔

    **Makna penting:**  

    * Turnover tinggi = biaya tinggi (rekrut, training)
    * Bisa indikasi masalah internal

    **Kontribusi ke tujuan:**  

    ✔ Mengontrol stabilitas organisasi  

    $$ \ $$

- Employee Satisfaction
 
    **Definisi:**  
    Tingkat kepuasan karyawan (biasanya dari survey)

    **Intuisi:**  

    > “Apakah mereka senang bekerja di sini?”

    * Tinggi → engagement tinggi ✔
    * Rendah → risiko churn ❌

    Ini sering jadi **leading indicator** turnover

    **Kontribusi ke tujuan:**  

    ✔ Mendeteksi masalah sebelum terjadi

    $$ \ $$

- Absenteeism Rate

    **Definisi:**  
    Tingkat ketidakhadiran karyawan

    Konsep:

    $$Absenteeism\ Rate = \frac{\text{Days Absent}}{\text{Total Work Days}} \times 100\% $$


    **Intuisi:**

    > Seberapa sering karyawan “tidak hadir”

    * Tinggi → potensi masalah
    * Rendah → stabil

    Bisa jadi sinyal:

    * burnout
    * disengagement
    * masalah kesehatan

    **Kontribusi ke tujuan:**

    ✔ Monitoring kedisiplinan & kesejahteraan

    $$ \ $$

- Performance Score

    **Definisi:**  
    Penilaian performa kerja karyawan

    **Intuisi:**

    > Seberapa besar kontribusi karyawan

    Biasanya:

    * berdasarkan KPI
    * atau penilaian manajer

    Penting:

    > bukan hanya rata-rata, tapi distribusi

    **Kontribusi ke tujuan:**

    ✔ Evaluasi produktivitas

### **Addition**

**Hubungan Antar Metrik**


- Satisfaction ↓ → Turnover ↑

    **Artinya:**

    > karyawan tidak puas → keluar


- Absenteeism ↑ → Performance ↓

    **Artinya:**

    > sering tidak hadir → output turun


- Performance Tinggi tapi Turnover Tinggi

    **Artinya:**

    > karyawan bagus → justru keluar


    Intuisi:

    > Perusahaan kehilangan “talent terbaik”

    Ini sangat berbahaya


- Turnover Tinggi di Satu Divisi

    Ini bukan masalah individu, tapi sistem


**Dari Metrik → Insight Nyata**

Contoh kamu:

> Turnover tinggi di divisi tertentu


Insight:

> Ada masalah spesifik di divisi itu

Kemungkinan:

* manajemen buruk
* beban kerja berlebihan
* budaya kerja tidak sehat

**Dari Insight → Keputusan**

- Action yang bisa diambil:

    - Evaluasi manajer/divisi
    - Review workload
    - Lakukan survey lebih dalam
    - Perbaiki lingkungan kerja


- Contoh lain:

    * Satisfaction rendah
    * Absenteeism tinggi


    Keputusan:

    * program employee engagement
    * fleksibilitas kerja
    * peningkatan kesejahteraan


- Intuisi

    > Data HR bukan hanya tentang karyawan, tapi tentang “kesehatan organisasi”

## **Attrition / Turnover Analysis**



Fokus:

> memahami dan mengurangi resign karyawan.

Karena turnover tinggi:

- mahal,
- mengganggu operasional,
- kehilangan knowledge perusahaan.

Attrition / Turnover

Mengukur:

> berapa banyak karyawan keluar/resign.

Contoh
|Employee	|Leaving|
|-----------|-------|
|1000	    |80|

Hitung
$$\frac{80}{1000}×100%=8%$$

Pertanyaan Umum
- Divisi mana turnover tertinggi?
- Karyawan senior atau junior lebih sering resign?
- Apakah salary memengaruhi resign?
- Apakah workload terlalu tinggi?

Analisis yang Sering Dipakai
- correlation analysis,
- logistic regression,
- churn prediction,
- survival analysis.

### **Statistics Analysis**

Statistik yang dipakai untuk HR Analytics

| Statistik   | Fungsi                       |
| ----------- | ---------------------------- |
| Mean        | rata-rata salary/performance |
| Correlation | hubungan antar variabel      |
| T-Test      | membandingkan rata-rata      |
| Chi-Square  | hubungan kategori            |
| Regression  | mencari faktor pengaruh      |
| Effect Size | melihat besar efek           |

Contoh

| Pertanyaan                                                  | Statistik / Metode               |
| ----------------------------------------------------------- | -------------------------------- |
| Apakah work-from-home meningkatkan retention?               | chi-square test, proportion test |
| Apakah rata-rata salary divisi A dan B berbeda?             | independent t-test               |
| Apakah training meningkatkan performance score?             | paired t-test                    |
| Apakah overtime berhubungan dengan resign?                  | correlation, chi-square test     |
| Apakah salary memengaruhi attrition?                        | regression                       |
| Apakah usia memengaruhi performance?                        | correlation, regression          |
| Apakah terdapat hubungan antara gender dan promosi?         | chi-square test                  |
| Apakah employee senior memiliki performa lebih tinggi?      | independent t-test               |
| Faktor apa yang paling memengaruhi resign?                  | regression                       |
| Apakah employee engagement berhubungan dengan productivity? | correlation                      |
| Apakah work-life balance memengaruhi turnover?              | regression, correlation          |
| Apakah rata-rata performa setelah training meningkat?       | paired t-test                    |
| Apakah divisi tertentu memiliki turnover lebih tinggi?      | chi-square test                  |
| Apakah bonus memengaruhi performance?                       | correlation, regression          |
| Seberapa besar pengaruh training terhadap performance?      | effect size                      |


Contoh Statistical Analysis

- Pertanyaan

    > “Apakah work-from-home meningkatkan retention?”

- Hipotesis

    - H0

        Tidak ada pengaruh.

    - H1

        Ada pengaruh.

- Hasil

    $$
    p < 0.05
    $$

- Interpretasi

    > WFH berpengaruh signifikan terhadap retention.

Confidence Interval

- Digunakan untuk:

    > estimasi range metric.

Effect Size

- Digunakan untuk:

    > melihat besar dampak.
