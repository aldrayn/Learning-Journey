# **Sales Analysis**

## **Introduction**

**Apa Itu Sales Analysis?**

Sales Analysis adalah:

> analisis data penjualan untuk memahami performa penjualan bisnis.


**Latar Belakang**  

Perusahaan punya banyak data penjualan:

- ribuan transaksi
- berbagai produk
- banyak channel (online/offline)

Masalah utamanya:

> Data itu “ada”, tapi tidak “bermakna” tanpa struktur.

Akibatnya:

- Manajemen tidak tahu apakah bisnis sedang sehat atau tidak
- Sulit menjawab pertanyaan cepat:
    - “Apakah bulan ini lebih baik dari bulan lalu?”
    - “Produk mana yang sebenarnya menopang bisnis?”
    - “Apakah peningkatan revenue karena harga naik atau jumlah penjualan naik?”

Jadi masalah inti:  
Kurangnya visibility (keterlihatan) dan insight dari data penjualan

**Tujuan**  

Dari masalah tadi, kita turunkan tujuan:

- Tujuan Utama:

    > Membuat sistem yang bisa menjawab kondisi bisnis dengan cepat dan akurat

- Tujuan Spesifik:
    - Mengetahui kondisi saat ini (baik / buruk)
    - Mengetahui arah tren (naik / turun)
    - Mengetahui penyebab perubahan
    - Mendukung keputusan cepat dan tepat

**Contoh Pertanyaan Bisnis**

* Produk mana paling banyak terjual?
* Sales bulan ini turun kenapa?
* Region mana performanya buruk?
* Customer mana paling banyak membeli?
* Target sales tercapai?

**Kenapa Perlu Metrik?**

Di sinilah inti yang sering tidak disadari:

> Tujuan itu abstrak, metrik itu konkret

Contoh:

- Tujuan: “mengetahui performa”
- Tapi… performa itu apa?

Harus diterjemahkan ke angka → itulah metrik

Jenis Sales Analysis

| Jenis                   | Fokus              |
|-------------------- |--------------- |
| Revenue Analysis        | pendapatan         |
| Sales Trend Analysis    | trend penjualan    |
| Product Sales Analysis  | performa produk    |
| Customer Sales Analysis | customer           |
| Regional Sales Analysis | wilayah            |
| Sales Forecasting       | prediksi penjualan |

## **Metrics**

### **Introduction**

**Metrik**

- Revenue (Pendapatan)

	- Definisi:  
		Total uang yang masuk dari penjualan

	- Intuisi:

		> Ini adalah “denyut nadi” bisnis

		Kalau revenue:

		- naik → bisnis tumbuh
		- turun → ada masalah

	- Masalah yang bisa dijawab:

		> Apakah bisnis menghasilkan uang cukup?

	- Kontribusi ke tujuan:

		✔ Mengetahui kondisi saat ini
		✔ Monitoring performa

	$$ \ $$

- Sales Growth (%)

	- Definisi:  
		Perubahan revenue dibanding periode sebelumnya

	- Intuisi:

		> Revenue itu “foto”, growth itu “video”

		- Revenue tinggi belum tentu bagus (bisa stagnan)
		- Growth menunjukkan arah

	- Contoh:

		- Revenue 1M tapi growth 0% → stagnan
		- Revenue 500k tapi growth 50% → berkembang cepat

	- Kontribusi ke tujuan:

		✔ Mengetahui tren (naik/turun)  
		✔ Evaluasi strategi

	$$ \ $$

- Units Sold

	- Definisi:    
		Jumlah produk yang terjual

	- Intuisi:

		> Ini menunjukkan “volume aktivitas”

	- Kenapa penting?

		Revenue bisa naik karena:
		- harga naik ❌
		- jumlah barang terjual naik ✔

		Units Sold membantu membedakan ini

	- Kontribusi ke tujuan:

		✔ Mengetahui penyebab perubahan revenue  
		✔ Analisis demand  

	$$ \ $$

- Average Order Value (AOV)

	- Definisi:  
		Rata-rata nilai transaksi per order

		$\text{AOV}=\frac{\text{Revenue}}{\text{JumlahTransaksi}}$


	- Intuisi:

		Seberapa “mahal” setiap transaksi

		Kalau AOV naik:

		- pelanggan beli lebih banyak
		- atau beli produk lebih mahal

	- Insight penting:

		- Revenue naik karena AOV naik → strategi bundling berhasil
		- Revenue naik tapi AOV turun → mungkin diskon besar

	- Kontribusi ke tujuan:

		✔ Mengetahui perilaku customer  
		✔ Optimasi strategi pricing & bundling   
	
	$$ \ $$

- Top Product

	- Definisi:
		Produk dengan performa terbaik (berdasarkan revenue / unit)

	- Intuisi:

		> “Pilar utama” bisnis

	- Biasanya:

		20% produk menghasilkan 80% revenue (Pareto)

	- Pertanyaan yang dijawab:

		- Produk mana yang harus diprioritaskan?
		- Produk mana yang harus dihentikan?

	- Kontribusi ke tujuan:

		✔ Pengambilan keputusan strategis  
		✔ Optimasi inventory & marketing

### **Addition**

**Hubungan Antar Metrik**

Banyak orang lihat metrik secara terpisah. Padahal kekuatan sebenarnya ada di hubungan antar metrik.

- Kasus 1: Revenue Naik

    Kemungkinan penyebab:

    - Units Sold naik
    - AOV naik
    - Keduanya naik

    Tanpa metrik lain → kita tidak tahu penyebabnya

- Kasus 2: Revenue Turun

    Analisis:

    - Units turun → demand turun
    - AOV turun → harga / strategi bermasalah

    Insight jadi actionable

**Dari Metrik ke Keputusan**

- Contoh Dashboard:
    - Revenue: +20%
    - Growth: meningkat
    - Units Sold: naik
    - AOV: stabil
    - Top Product: Produk A
- Insight (bukan sekadar angka):
    - Kenaikan revenue bukan karena harga, tapi karena volume
    - Produk A adalah driver utama
- Keputusan:
    - Tambah stok Produk A
    - Fokus marketing ke Produk A
    - Replikasi strategi Produk A ke produk lain

**Intuisi**

> Dashboard itu bukan soal visualisasi, tapi soal “menerjemahkan realita bisnis menjadi angka yang bisa dipahami dan ditindaklanjuti”

Alur Berpikir
- Masalah:   
    Data banyak, sulit dipahami
- Tujuan:  
    Memahami kondisi & tren bisnis
- Solusi:  
    Gunakan metrik
- Peran Metrik:   
    - Mengubah abstrak → konkret
    - Menjawab pertanyaan bisnis
    - Menjadi dasar keputusan
- Output Akhir:  
    Insight → Action → Impact

## **Revenue Analysis**

Analisis untuk:

> mengukur pendapatan dari penjualan.

Biasanya fokus pada:

* total revenue,
* growth revenue,
* profitabilitas bisnis.


Revenue menjadi indikator utama:

> performa bisnis menghasilkan uang.

Rumus Revenue

$$Revenue=Price\times Quantity\ Sold$$

Insight

* revenue naik → bisnis berkembang.
* revenue turun → sales atau demand menurun.

## **Sales Trend Analysis**

Analisis untuk:

> melihat pola penjualan dari waktu ke waktu.


Digunakan untuk:

* melihat trend,
* seasonality,
* pertumbuhan sales.

Rumus Growth Rate

$$Growth\ Rate=\frac{Current\ Sales-Previous\ Sales}{Previous\ Sales}\times100%
$$

Insight

* trend naik → demand meningkat.
* trend turun → perlu evaluasi strategi.

## **Product Sales Analysis**

Analisis untuk:

> mengevaluasi performa tiap produk.


Membantu bisnis mengetahui:

* produk paling laku,
* produk slow moving,
* kontribusi produk terhadap revenue.

Metric Umum

| Metric              | Fungsi                |
| ------------------- | --------------------- |
| Units Sold          | jumlah produk terjual |
| Revenue per Product | kontribusi revenue    |
| Product Growth      | pertumbuhan produk    |

Insight

* produk tertentu menjadi best seller.
* produk lain mungkin overstock atau kurang diminati.

## **Customer Sales Analysis**

Analisis untuk:

> memahami perilaku pembelian customer.


Fokus pada:

* customer loyal,
* high spender,
* customer segment.

Rumus Average Order Value

$$AOV=\frac{Total\ Revenue}{Total\ Orders}$$

Insight

* customer tertentu memberi revenue besar.
* segment tertentu lebih profitable.

## **Regional Sales Analysis**

Analisis untuk:

> membandingkan performa sales antar wilayah.


Digunakan untuk:

* melihat market potensial,
* menentukan strategi regional.

Metric Umum

| Metric              | Fungsi              |
| ------------------- | ------------------- |
| Revenue by Region   | revenue wilayah     |
| Sales Growth        | pertumbuhan wilayah |
| Market Contribution | kontribusi wilayah  |

Insight

* wilayah tertentu mendominasi sales.
* wilayah lain membutuhkan strategi marketing tambahan.

## **Sales Forecasting**

Analisis untuk:

> memprediksi penjualan di masa depan.


Forecasting membantu:

* perencanaan stok,
* target sales,
* budgeting.

Metode Umum

| Metode         | Fungsi                  |
| -------------- | ----------------------- |
| Moving Average | smoothing trend         |
| ARIMA          | forecasting time series |
| Regression     | prediksi sales          |

Rumus Moving Average

$$Moving\ Average=\frac{\sum Sales}{n}$$

Insight

* prediksi sales membantu planning bisnis.
* forecasting buruk bisa menyebabkan stockout atau overstock.

### **Statistics Analysis**

Statistik dalam Sales Analysis

Sales analysis biasanya digunakan untuk:

* memahami performa penjualan,
* mengevaluasi strategi sales,
* dan memprediksi sales di masa depan.

Statistik yang Sering Digunakan

| Statistik   | Fungsi                  |
| ----------- | ----------------------- |
| Mean        | rata-rata sales         |
| Growth Rate | pertumbuhan sales       |
| Correlation | hubungan variabel       |
| T-Test      | compare campaign        |
| Regression  | prediksi sales          |
| Effect Size | besar pengaruh strategi |





Contoh Statistical Analysis

| Pertanyaan                                                | Statistik / Metode                                |
| --------------------------------------------------------- | ------------------------------------------------- |
| Apakah diskon meningkatkan sales?                         | A/B testing, t-test, p-value, confidence interval |
| Apakah campaign tertentu meningkatkan revenue?            | t-test, regression                                |
| Apakah harga memengaruhi jumlah penjualan?                | correlation, regression                           |
| Faktor apa yang paling memengaruhi sales?                 | regression                                        |
| Apakah sales bulan ini meningkat dibanding bulan lalu?    | growth rate, trend analysis                       |
| Apakah terdapat perbedaan sales antar region?             | ANOVA                                             |
| Apakah promosi meningkatkan conversion rate?              | chi-square test, A/B testing                      |
| Apakah customer loyal menghasilkan sales lebih tinggi?    | t-test                                            |
| Apakah seasonal event meningkatkan sales?                 | time series analysis                              |
| Apakah ads spending berhubungan dengan sales?             | correlation, regression                           |
| Apakah sales team A lebih baik dibanding team B?          | independent t-test                                |
| Apakah jumlah kunjungan customer memengaruhi sales?       | regression                                        |
| Seberapa besar pengaruh diskon terhadap sales?            | effect size                                       |
| Apakah bundling product meningkatkan average order value? | t-test                                            |
| Apakah perubahan harga memengaruhi conversion?            | regression, correlation                           |




Statistik yang Sering Dipakai Junior Sales Analyst

| Statistik           | Fungsi                       |
| ------------------- | ---------------------------- |
| Mean                | rata-rata sales              |
| Growth Rate         | melihat pertumbuhan          |
| Correlation         | hubungan antar variabel      |
| T-Test              | membandingkan performa       |
| Regression          | prediksi & faktor pengaruh   |
| A/B Testing         | membandingkan strategi sales |
| P-Value             | melihat signifikansi         |
| Confidence Interval | estimasi rentang hasil       |

Intuisi

Sales analysis membantu bisnis memahami:

* faktor yang meningkatkan penjualan,
* strategi sales yang paling efektif,
* dan pola pertumbuhan revenue dari waktu ke waktu.
