# **Market Campaign Analysis**

## **Introduction**

**Apa Itu Marketing Campaign Analysis?**

Marketing Campaign Analysis adalah:

> analisis performa campaign marketing untuk mengetahui apakah campaign efektif atau tidak.


**Latar Belakang**

Perusahaan menghabiskan budget besar:

* Ads (Meta, Google)
* Influencer
* Email campaign

Tapi sering terjadi:

* Traffic tinggi, tapi sales rendah
* Banyak klik, tapi tidak jadi pelanggan
* Biaya besar, tapi ROI negatif

Masalah inti:

> **Tidak tahu di mana “kebocoran” dalam funnel marketing**

**Tujuan**

- Tujuan Utama:

    > Memaksimalkan profit dari setiap rupiah yang dikeluarkan untuk marketing

- Tujuan Spesifik:

    - Mengukur efektivitas tiap campaign
    - Menemukan titik lemah dalam funnel
    - Mengoptimalkan alokasi budget
    - Meningkatkan conversion dan ROI

**Konsep - Marketing = Funnel**  

- Ini inti yang harus kamu pegang:

    > Marketing bukan satu langkah, tapi **proses bertahap**

- Funnel Sederhana:

    - Impression (lihat iklan)
    - Klik → (CTR)
    - Masuk website
    - Beli → (Conversion Rate)


- Insight penting:

    > Setiap tahap punya metrik sendiri, dan bisa “bocor”

Contoh Pertanyaan Bisnis

* Campaign mana paling efektif?
* Channel mana conversion tertinggi?
* ROI campaign bagus?
* CTR iklan rendah?
* Customer acquisition terlalu mahal?

Jenis Marketing Campaign Analysis

| Jenis                | Fokus             |
|----------------- |-------------- |
| Campaign Performance | performa campaign |
| Conversion Analysis  | conversion        |
| Engagement Analysis  | interaksi user    |
| Attribution Analysis | sumber conversion |
| ROI Analysis         | profit campaign   |
| Funnel Analysis      | perjalanan user   |

## **Metrics**

### **Introduction**

**Metrik**
- Click-Through Rate (CTR)

    $$CTR = \frac{\text{Clicks}}{\text{Impressions}} \times 100\% $$


    **Intuisi:**

    > Seberapa kuat iklan menarik perhatian

    * Tinggi → hook bagus ✔
    * Rendah → iklan tidak menarik ❌


- Conversion Rate

    $$Conversion\ Rate = \frac{\text{Conversions}}{\text{Visitors}} \times 100\% $$

    **Intuisi:**

    > Setelah klik, apakah mereka yakin untuk membeli?

    * Tinggi → produk & UX bagus ✔
    * Rendah → ada masalah di landing page ❌

- Cost Per Acquisition (CPA)

    $$CPA = \frac{\text{Total Cost}}{\text{Number of Customers Acquired}} $$



    **Intuisi:**

    > Berapa “harga” satu customer

    * Rendah → efisien ✔
    * Tinggi → boros ❌


- Return on Investment (ROI)

    $$ROI = \frac{\text{Revenue} - \text{Cost}}{\text{Cost}} \times 100\% $$
    
    
    **Intuisi:**
    
    > Ini keputusan akhir: untung atau tidak

### **Addition**

**Hubungan Antar Metrik (Funnel Diagnosis)**

Ini bagian paling penting—karena di sinilah insight muncul.


- Kasus Utama:

    > CTR Tinggi, Conversion Rendah


- Analisis Step-by-Step:

    - CTR tinggi  
       Iklan berhasil menarik perhatian

    - Conversion rendah  
       Setelah masuk, user tidak jadi beli


- Intuisi:

    > Ada “janji” di iklan yang tidak terpenuhi di landing page


- mungkinan masalah:

    * Landing page tidak relevan
    * UX buruk (lemot, ribet)
    * Penawaran tidak menarik
    * Trust rendah (tidak ada review, dll)


Dari Insight → Keputusan

Action yang tepat:

- **Perbaiki landing page**

   * lebih cepat
   * lebih jelas
   * lebih fokus ke value

- **Sinkronisasi iklan & halaman**

   * pesan harus konsisten

- **Optimasi penawaran**

   * diskon
   * bonus
   * urgency (limited offer)

**Pola**



- CTR Rendah, Conversion Tinggi

    **Artinya:**

    * sedikit yang klik
    * tapi yang klik pasti beli

    Insight:

    > targeting bagus, tapi iklan kurang menarik



- CTR Tinggi, CPA Tinggi

    **Artinya:**

    * banyak klik
    * tapi mahal

    Insight:

    > traffic banyak, tapi tidak berkualitas



- Conversion Tinggi, ROI Negatif

    **Artinya:**

    * banyak yang beli
    * tapi biaya terlalu besar

    Insight:

    > pricing atau cost bermasalah



- CPA Rendah, Tapi CLV Rendah

    (ini advanced)   

    **Artinya:**    

    * murah dapat customer  
    * tapi mereka tidak bertahan    

    Insight:

    > kualitas customer rendah




Intuisi

> Marketing bukan tentang “traffic”, tapi tentang  
> **traffic yang menghasilkan profit**

**Alur Berpikir Lengkap**

- **Masalah:**  
   Budget besar, hasil tidak jelas

- **Tujuan:**  
   Efisiensi & profit

- **Solusi:**  
   Gunakan metrik funnel

- **Insight:**  
   Temukan titik kebocoran

- **Action:**  
   Optimasi di titik tersebut

**Insight Lanjutan (Level Advanced)**

- Hubungkan dengan CLV

    Rule penting:

    > **CPA < CLV → sehat**
    > **CPA > CLV → rugi**


- Gunakan A/B Testing

    * test iklan
    * test landing page
    * test penawaran


- Attribution Analysis

  * channel mana paling efektif?







**Ringkasan**

| Metrik     | Peran             |
| ---------- | ----------------- |
| CTR        | Menarik perhatian |
| Conversion | Meyakinkan        |
| CPA        | Efisiensi biaya   |
| ROI        | Profit            |


Output akhirnya:

> **Menentukan campaign mana yang layak di-scale, diperbaiki, atau dihentikan**

## **Campaign Performance Analysis**


Analisis untuk:

> mengukur performa keseluruhan campaign marketing.

Biasanya fokus pada:

* reach,
* clicks,
* sales,
* conversion,
* revenue.

Membandingkan:

* hasil campaign,
* target,
* atau campaign lain.

Metric Umum

| Metric      | Fungsi          |
| ----------- | --------------- |
| Impressions | jumlah tayangan |
| Clicks      | jumlah klik     |
| Conversion  | jumlah konversi |
| Revenue     | pendapatan      |

Rumus CTR

$$CTR=\frac{Clicks}{Impressions}\times100\%$$

Insight

* CTR tinggi → campaign menarik.
* CTR rendah → ads kurang menarik.

## **Conversion Analysis**



Analisis untuk:

> mengukur berapa banyak user melakukan aksi yang diinginkan.

Contoh:

* membeli,
* signup,
* subscribe.

Konsep

> Tidak semua user yang melihat campaign akan convert.

Fokus utamanya:

> efisiensi mengubah user menjadi customer.

Rumus Conversion Rate

$$Conversion\ Rate=\frac{Conversions}{Visitors}\times100\%$$

Insight

* conversion tinggi → campaign efektif.
* conversion rendah → funnel bermasalah.

## **Engagement Analysis**

Analisis untuk:

> mengukur interaksi user terhadap content atau campaign.


Semakin tinggi engagement:

* biasanya user lebih tertarik,
* awareness lebih baik.

Metric Umum

| Metric     | Fungsi               |
| ---------- | -------------------- |
| Likes      | ketertarikan         |
| Shares     | penyebaran content   |
| Comments   | interaksi user       |
| Watch Time | ketertarikan content |

Rumus Engagement Rate

$$Engagement\ Rate=\frac{Total\ Engagement}{Total\ Audience}\times100\%$$

Insight

* engagement tinggi → content relevan.
* engagement rendah → content kurang menarik.

## **Attribution Analysis**

Analisis untuk:

> mencari sumber conversion berasal dari mana.


Customer sering melewati banyak channel:

* Instagram,
* Google Ads,
* Email,
* TikTok.

Attribution mencoba menentukan:

> channel mana yang paling berkontribusi.

Contoh Model

| Model       | Fokus            |
| ----------- | ---------------- |
| First Click | channel pertama  |
| Last Click  | channel terakhir |
| Linear      | semua channel    |

Rumus Sederhana Attribution

$$Attribution\ Contribution=\frac{Channel\ Conversion}{Total\ Conversion}\times100\%$$

Insight

* channel tertentu paling efektif.
* budget marketing bisa dioptimalkan.


## **ROI Analysis**

Analisis untuk:

> mengukur apakah campaign menghasilkan profit.


Campaign bagus bukan hanya banyak klik,
tetapi:

> menghasilkan keuntungan.

Rumus ROI

$$ROI=\frac{Revenue-Cost}{Cost}\times100\%$$

Insight

* ROI positif → campaign profitable.
* ROI negatif → campaign merugi.


## **Funnel Analysis**

Analisis untuk:

> melihat perjalanan user dari awal sampai conversion.

User biasanya melewati beberapa tahap:
> 
> ```text
> View → Click → Add to Cart → Purchase
> ```

Fokus Funnel

> Mencari di tahap mana user banyak keluar.

Rumus Funnel Conversion

$$Funnel\ Conversion=\frac{Next\ Stage}{Previous\ Stage}\times100\%$$

Contoh

| Tahap    | User |
| -------- | ---- |
| View     | 1000 |
| Click    | 300  |
| Purchase | 30   |

Insight

* banyak drop setelah click,
* kemungkinan landing page buruk,
* atau checkout terlalu sulit.

## **Addition**

### **Statistics Analysis**

Statistik dalam Marketing Campaign Analysis

Marketing campaign analysis biasanya digunakan untuk:

* mengukur efektivitas campaign,
* meningkatkan conversion,
* dan memahami faktor yang memengaruhi customer response.

Statistik yang Sering Digunakan

| Statistik   | Fungsi                  |
| ----------- | ----------------------- |
| Mean        | rata-rata conversion    |
| Correlation | hubungan metric         |
| T-Test      | compare campaign        |
| Chi-Square  | compare conversion      |
| Regression  | faktor conversion       |
| Effect Size | besar pengaruh campaign |





Contoh Statistical Analysis

| Pertanyaan                                                    | Statistik / Metode                                     |
| ------------------------------------------------------------- | ------------------------------------------------------ |
| Apakah landing page baru meningkatkan conversion?             | A/B testing, p-value, confidence interval, effect size |
| Apakah email campaign meningkatkan click-through rate?        | t-test, chi-square                                     |
| Apakah diskon besar meningkatkan sales?                       | regression, correlation                                |
| Apakah campaign A lebih efektif dibanding campaign B?         | t-test, A/B testing                                    |
| Faktor apa yang paling memengaruhi conversion?                | regression                                             |
| Apakah segment tertentu lebih responsif terhadap campaign?    | chi-square test                                        |
| Apakah ads spending berhubungan dengan revenue?               | correlation, regression                                |
| Apakah campaign baru meningkatkan retention customer?         | t-test, effect size                                    |
| Apakah terdapat perbedaan conversion antar channel marketing? | ANOVA                                                  |
| Apakah push notification meningkatkan engagement?             | A/B testing                                            |
| Apakah waktu pengiriman email memengaruhi open rate?          | ANOVA, regression                                      |
| Apakah campaign tertentu menghasilkan CLV lebih tinggi?       | regression, t-test                                     |
| Apakah social media engagement berhubungan dengan conversion? | correlation                                            |
| Seberapa besar pengaruh campaign terhadap conversion rate?    | effect size                                            |
| Apakah conversion rate setelah redesign meningkat signifikan? | p-value, confidence interval                           |




Statistik yang Sering Dipakai Junior Analyst

| Statistik           | Fungsi                             |
| ------------------- | ---------------------------------- |
| Mean                | rata-rata conversion/revenue       |
| Correlation         | hubungan antar metric              |
| T-Test              | membandingkan campaign             |
| Chi-Square          | membandingkan conversion kategori  |
| Regression          | faktor yang memengaruhi conversion |
| A/B Testing         | membandingkan dua versi campaign   |
| P-Value             | melihat signifikansi               |
| Confidence Interval | estimasi rentang hasil             |

Intuisi

Marketing campaign analysis membantu bisnis memahami:

* campaign mana yang paling efektif,
* faktor yang meningkatkan conversion,
* dan apakah perubahan campaign benar-benar memberi dampak signifikan.
