# **Cohort Analysis**

## **Introduction**

### **Concept**

**Latar Belakang**  

Kebanyakan perusahaan melihat data seperti ini:

- Total user: 10.000
- Total revenue: naik
- Total retention: 40%

Kelihatannya oke… tapi ada masalah besar:

> Semua user “dicampur jadi satu”

Padahal:

- User yang datang bulan Januari bisa sangat loyal
- User yang datang bulan Februari bisa cepat churn

Masalah utama:  
Kita kehilangan “cerita perjalanan” tiap kelompok user

**Tujuan**  

- Tujuan Utama:

    > Memahami perilaku user berdasarkan waktu mereka masuk

- Tujuan Spesifik:
    - Melihat retensi dari waktu ke waktu
    - Membandingkan kualitas user antar periode
    - Mengetahui kapan user mulai churn
    - Menghubungkan perubahan performa dengan event (campaign, fitur, dll)
    - Mengetahuin cohort terbaik

Masalah yang Ingin Dijawab


- Contoh 1

    > “Kenapa user cepat hilang setelah daftar?”


- Contoh 2

    > “Apakah onboarding baru meningkatkan retention?”


- Contoh 3

    > “Cohort bulan mana paling loyal?”

**Konsep**  

- Cohort
  
  > Cohort = kelompok user berdasarkan waktu atau karakteristik tertentu
  
  Contoh paling umum:

  | Cohort          | Contoh                               |
  |------------ |--------------------------------- |
  | Join Cohort     | user yang daftar di bulan yang sama  |
  | Purchase Cohort | user yang pertama beli di waktu sama |
  | Campaign Cohort | user dari campaign tertentu          |  

  $$ \ $$
- Cohort Analysis  

    > analisis perilaku kelompok user berdasarkan waktu atau karakteristik tertentu.

    Biasanya digunakan untuk melihat:

    * retention,
    * churn,
    * engagement,
    * dan lifecycle customer.

  $$ \ $$
- Intuisi Penting:

    > Cohort analysis itu seperti “melacak generasi user”

    Bukan sekadar:  
    > “berapa user sekarang?”

    Tapi:
    > “bagaimana nasib user yang datang di bulan tertentu?”

Kenapa Cohort Analysis Penting?

- Karena total metric kadang menipu.

- Misal 
    > retention total terlihat stabil.

- Tetapi:

    > ternyata user baru cepat churn, hanya user lama yang bertahan.

    Nah cohort analysis membantu melihat detail seperti itu.

Jenis Cohort Analysis

| Jenis             | Fokus              |
|-------------- |--------------- |
| Retention Cohort  | user bertahan      |
| Revenue Cohort    | revenue per cohort |
| Engagement Cohort | aktivitas user     |
| Product Cohort    | penggunaan feature |

Hubungan Antar Cohort Analysis

Sebenarnya ketiga jenis ini saling berhubungan. Alur Umumnya

> Engagement tinggi → Retention tinggi → Revenue tinggi

Karena:

- user aktif lebih sering,
- user bertahan lebih lama,
- user lebih banyak bertransaksi.

Flow Cohort Analysis

> ```text id="fjlwmh"
> Business Problem
>       ↓

>       ↓
> Kelompokkan User
>       ↓

Contoh Business Problem

| Problem           | Fokus             |
|-------------- |-------------- |
| User cepat hilang | retention         |
| Revenue turun     | revenue cohort    |

### **Additional**

#### **Statistics Analysis**

Statistical Analysis dalam Cohort

Cohort analysis kadang dikombinasikan dengan:

- hypothesis testing,
- confidence interval,
- A/B testing,

untuk memahami:

> apakah perubahan strategi benar-benar memengaruhi retention atau behavior customer.

Contoh Statistical Analysis dalam Cohort

| Pertanyaan                                                         | Statistik / Metode                        |
| ------------------------------------------------------------------ | ----------------------------------------- |
| Apakah onboarding baru meningkatkan Month-1 retention?             | p-value, confidence interval, effect size |
| Apakah retention cohort tahun ini lebih baik dibanding tahun lalu? | t-test, confidence interval               |
| Apakah campaign tertentu meningkatkan repeat purchase?             | A/B testing, p-value                      |
| Apakah cohort premium memiliki churn lebih rendah?                 | chi-square test, t-test                   |
| Apakah user yang onboarding lewat mobile lebih aktif?              | t-test, regression                        |
| Apakah terdapat perbedaan retention antar cohort bulanan?          | ANOVA                                     |
| Apakah feature baru meningkatkan engagement cohort baru?           | A/B testing, effect size                  |
| Apakah cohort tertentu memiliki CLV lebih tinggi?                  | segmentation analysis, ANOVA              |
| Apakah retention berubah signifikan setelah perubahan pricing?     | hypothesis testing, p-value               |
| Apakah user lama dan baru memiliki behavior berbeda?               | t-test                                    |
| Seberapa besar pengaruh onboarding terhadap retention?             | effect size, regression                   |
| Apakah cohort tertentu lebih cepat churn?                          | survival analysis, regression             |
| Apakah email campaign meningkatkan activation rate?                | A/B testing, confidence interval          |
| Apakah seasonality memengaruhi retention cohort?                   | time series analysis                      |
| Apakah perubahan UI meningkatkan conversion cohort baru?           | A/B testing, p-value                      |

## **Metric**

### **Retention Rate per Cohort**

Definisi:  
> Persentase user yang kembali di periode berikutnya

Konsep:

$$\text{Retention}= \frac{\text{Active Users in Period}}{\text{Total Users in Cohort}}×100\% $$

Intuisi:

> Seberapa “tahan lama” user dari cohort tertentu

Contoh:

- Day 1: 100%
- Day 7: 50%
- Day 30: 20%

> Ini seperti “survival curve”

Kontribusi ke tujuan:

✔ Mengukur kualitas user  
✔ Melihat penurunan engagement  

### **Churn per Cohort**

Definisi:
> Kapan user mulai berhenti

Intuisi:

> Kita ingin tahu “titik kematian” user

Contoh:

> Banyak user hilang di minggu pertama
> Stabil setelah itu

Insight:   
> masalah ada di awal (onboarding)

Kontribusi ke tujuan:

✔ Identifikasi titik kritis churn

### **Lifetime per Cohort**



Definisi:
> Rata-rata durasi user tetap aktif

Intuisi:

 > Berapa lama user “hidup” di sistem

- Lifetime panjang → user loyal
- Lifetime pendek → user cepat bosan

Kontribusi ke tujuan:

✔ Evaluasi kualitas jangka panjang  
    

### **Revenue per Cohort**



Definisi:
> Total revenue yang dihasilkan oleh cohort tertentu

Intuisi:

> Tidak semua user punya nilai yang sama

Contoh:

- Cohort A: kecil tapi menghasilkan banyak uang
- Cohort B: besar tapi tidak profitable

Kontribusi ke tujuan:

✔ Menilai kualitas dari sisi bisnis (bukan hanya aktivitas)

### **Additional**

**Bentuk Cohort**  

- Biasanya ditampilkan seperti tabel:

    | Cohort | Month 0 | Month 1 | Month 2 | Month 3 |
    | ------ | ------- | ------- | ------- | ------- |
    | Jan    | 100%    | 60%     | 40%     | 30%     |
    | Feb    | 100%    | 30%     | 20%     | 10%     |

    Intuisi :
    - Baris = “generasi user”
    - Kolom = “waktu sejak mereka join”

- Fokus:  
    Bandingkan antar baris (antar cohort)

**Dari Data → Insight Nyata**  

- Dari data:
    - Januari: 60% retention
    - Februari: 30% retention

- Insight:

    > User Februari jauh lebih cepat churn

- Pertanyaan lanjutan:

    > Apa yang berubah di Februari?

- Kemungkinan:

    - Campaign berbeda
    - Target user berbeda
    - Produk berubah
    - UX berubah

- Cohort analysis bukan hanya “lihat angka”, tapi:

    > “menghubungkan perubahan dengan penyebab”

**Pola**

- Semua Cohort Turun Cepat di Awal  
    Artinya:  
    > Masalah onboarding

- Cohort Baru Lebih Buruk dari Lama  
    Artinya:
    - Kualitas acquisition menurun  
    - Campaign jelek
- Cohort Baru Lebih Baik

    Artinya:  
    > Ada improvement (fitur / UX / marketing)

- Retention Stabil Setelah Turun

    Artinya:  
    > Ada “core loyal users”

**Dari Insight → Keputusan**

Kasus kamu:
> Februari retention lebih rendah
 
Keputusan:
- Evaluasi campaign Februari
- Cek apakah target user berubah
- Analisis landing page / onboarding
- Bandingkan behavior user Januari vs Februari

Intuisi

> Tanpa cohort analysis, kamu hanya melihat “jumlah”  
> Dengan cohort analysis, kamu melihat “perjalanan”

## **Retention Cohort Analysis**

Jenis paling umum dan paling penting.

Fokus utama:

> Berapa banyak user yang tetap aktif setelah beberapa periode waktu.

Biasanya digunakan untuk:

- aplikasi,
- SaaS,
- e-commerce,
- game,
- subscription business.

Misalkan cohort dibentuk berdasarkan bulan registrasi user.

| Join Month | Month 0 | Month 1 | Month 2 |
| ---------- | ------- | ------- | ------- |
| Januari    | 100%    | 60%     | 40%     |
| Februari   | 100%    | 70%     | 55%     |
| Maret      | 100%    | 75%     | 60%     |

Interpretasi
- Month 0   
  - Periode saat user pertama kali masuk.
  - Selalu 100% karena semua user cohort masih dihitung aktif pada awal.

- Month 1
  - Persentase user yang masih aktif setelah 1 bulan.

  - Contoh:  
    Januari: hanya 60% user yang kembali aktif.

    Artinya:

    > 40% user churn/tidak kembali.

- Month 2

    - Persentase user yang masih aktif setelah 2 bulan.
    - Biasanya retention akan terus menurun seiring waktu.



Interpretasi Insight  
> Cohort Maret lebih baik karena:  

| Cohort  | Month 1 | Month 2 |
| ------- | ------- | ------- |
| Januari | 60%     | 40%     |
| Maret   | 75%     | 60%     |

Artinya:

- onboarding mungkin lebih baik,
- fitur baru berhasil,
- kualitas acquisition lebih bagus,
- marketing lebih tepat sasaran.

Insight yang Sering Dicari

| Kondisi                       | Insight                  |
| ----------------------------- | ------------------------ |
| Retention turun sangat cepat  | onboarding buruk         |
| Cohort baru lebih baik        | improvement berhasil     |
| Semua cohort turun            | product issue            |
| Retention stabil tinggi       | product-market fit bagus |
| Retention naik setelah update | fitur baru efektif       |


**Bentuk Visual Cohort**

Biasanya dibuat dalam bentuk:

- heatmap,
- pivot table,
- matrix retention.

Semakin gelap warna:

> semakin tinggi retention.





**Kenapa Retention Sangat Penting?**

> Retention sering dianggap lebih penting dibanding acquisition.

Karena:

- mendapatkan user baru mahal,
- mempertahankan user lebih murah,
- retention tinggi biasanya meningkatkan:
    - revenue,
    - engagement,
    - lifetime value.

## **Revenue Cohort Analysis**

> Revenue Cohort Analysis melihat bagaimana pendapatan dari setiap cohort berkembang seiring waktu.

Fokus:

> cohort mana yang menghasilkan revenue terbesar dan paling stabil.

Contoh Revenue Cohort

| Cohort   | Month 1 Revenue | Month 2 Revenue |
| -------- | --------------- | --------------- |
| Januari  | 10 juta         | 7 juta          |
| Februari | 15 juta         | 12 juta         |
| Maret    | 18 juta         | 16 juta         |

Interpretasi
> Cohort Maret Lebih Baik

Karena:

- revenue awal lebih tinggi,
- penurunan revenue lebih kecil,
- kemungkinan customer lebih loyal.

Insight Revenue Cohort

| Kondisi                       | Insight                  |
| ----------------------------- | ------------------------ |
| Revenue turun cepat           | customer tidak loyal     |
| Revenue stabil                | recurring customer bagus |
| Cohort baru lebih profitable  | strategi berhasil        |
| Revenue naik setelah campaign | marketing efektif        |

Hubungan dengan CLV

> Revenue cohort sangat berkaitan dengan:
> - Customer Lifetime Value (CLV),
> - monetization,
> - repeat purchase.

## **Engagement Cohort Analysis**




Digunakan untuk melihat tingkat keterlibatan user dari waktu ke waktu.

Fokus:

- login frequency,
- session duration,
- watch time,
- jumlah aktivitas,
- jumlah transaksi.

Contoh

| Cohort   | Avg Session |
| -------- | ----------- |
| Januari  | 20 menit    |
| Februari | 35 menit    |
| Maret    | 40 menit    |


Jika engagement meningkat:

- user semakin nyaman,
- fitur lebih menarik,
- UX lebih baik,
- kemungkinan retention ikut naik.

Insight  

| Kondisi            | Insight                     |
| ------------------ | --------------------------- |
| Session pendek     | user kurang tertarik        |
| Watch time naik    | content menarik             |
| Login makin sering | engagement meningkat        |
| Engagement turun   | kemungkinan churn meningkat |
