## **Introduction**

**Definisi**  

> Analisis Bivariat (Bivariate Analysis) adalah proses mengeksplorasi hubungan antara dua variabel — bisa numerik, kategorikal, atau kombinasi keduanya.

Kalau dalam univariate analysis kamu hanya fokus pada satu variabel (misal distribusi, sebaran, pusat data, dll.), maka pada bivariate kamu melihat interaksi dua variabel sekaligus.

$ $

---
**Tujuan Utama**

- Mengetahui apakah ada hubungan antara dua variabel.
- Memahami arah dan kekuatan hubungan (positif, negatif, lemah, kuat, dll).
- Menilai bentuk hubungan (linear, non-linear, proporsional, asimetris, dsb).
- Menjadi dasar untuk model prediktif, misalnya menentukan feature penting dalam regresi.

💡 Analogi Intuitif

Bayangkan kamu seorang analis pemasaran.
Kamu ingin tahu: "Apakah semakin banyak iklan meningkatkan penjualan?"
Itu berarti kamu sedang melihat hubungan antara dua variabel: jumlah iklan (X) dan penjualan (Y).
Jika ternyata saat X naik, Y juga naik → kamu menemukan korelasi positif.

Itulah inti dari analisis bivariat.

## **Correlation & Association**

### **Numerik-Numerik**

#### **Pearson**

---
**Konsep**  

Bayangkan kamu punya dua hal yang bisa berubah bersama, misalnya:

- ❄️ Suhu udara
- 🍦 Jumlah es krim terjual

Kamu mungkin sadar:

> kalau suhu naik, biasanya penjualan es krim juga naik.

Nah, hubungan “kalau satu naik, yang lain juga naik” — itulah yang disebut korelasi positif.
Kalau satu naik tapi yang lain turun, itu korelasi negatif.

👉 Pearson Correlation adalah angka yang bilang seberapa kuat dua hal berubah bersama.
$$ \ $$

---
**Intuisi**

Coba bayangkan kamu memegang dua tali:

- Tali pertama mewakili X (misal suhu)
- Tali kedua mewakili Y (misal penjualan es krim)

Sekarang kamu tarik bersamaan:

- Kalau kedua tali bergerak searah, artinya hubungan mereka kuat dan positif
- Kalau satu tali maju dan satu lagi mundur, hubungan mereka negatif
- Kalau gerakannya acak, berarti tidak ada hubungan

Pearson correlation memberi angka antara -1 sampai +1:
$$ \ $$

| Nilai | Arti                                                           |
| ----- | -------------------------------------------------------------- |
| +1    | Hubungan sempurna searah (satu naik, yang lain juga naik)      |
| 0     | Tidak ada hubungan linear                                      |
| -1    | Hubungan sempurna berlawanan arah (satu naik, yang lain turun) |

$$ \ $$

---
**Rumus**

Rumusnya terlihat seperti ini:

$$ \ $$
$$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}
$$
$$ \ $$

Keterangan :   
- Bagian 1 $\to (x_i - \bar{x})$  
  Ini artinya “seberapa jauh nilai X saat ini dari rata-rata X”.

- Bagian 2 $\to (y_i - \bar{y})$  
  Sama, tapi untuk Y (misal penjualan es krim).
  - Kalau hasilnya positif → lebih tinggi dari rata-rata.
  - Kalau negatif → lebih rendah.

- Bagian 3 $\to (x_i - \bar{x})(y_i - \bar{y})$  
  Ini disebut **“produk deviasi”**, artinya kita mengalikan penyimpangan X dan Y.

  Maknanya:

  * Kalau **dua-duanya naik bersama (positif × positif)** → hasil positif
  * Kalau **satu naik dan satu turun (positif × negatif)** → hasil negatif
  * Jadi tanda dari hasil ini menunjukkan **arah hubungan**.

- Bagian 4 $\to \sum (x_i - \bar{x})(y_i - \bar{y})$  
  Ini menjumlahkan semua pasangan deviasi tadi.  
  
  Semakin besar hasilnya, berarti X dan Y sering **naik-turun bersama**.

- Bagian 5 $\to \sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}$

  Ini bagian “penyeimbang” (normalisasi). Kita bagi dengan seberapa besar “penyebaran” data X dan Y.

  Tujuannya supaya nilai $r$ **selalu antara -1 dan +1**, biar mudah dibandingkan.


Kesimpulan intuitif:

$ $

$$r = \frac{\text{seberapa sering X dan Y bergerak bersama}}{\text{seberapa besar X dan Y bisa bergerak secara umum}}
$$

$$ \ $$

---
**Kapan digunakan**

Gunakan **Pearson** kalau:

* Kedua variabel **numerik kontinu** (misal suhu, umur, pendapatan, skor ujian).
* Kamu **menduga hubungan linier**, misalnya naik-turun proporsional.
* Data tidak terlalu memiliki **outlier ekstrem**.
* Distribusi mendekati **normal (symmetrical)**.

$$ \ $$

---
**Kesalahan Umum dan Miskonsepsi**

| Kesalahan                                        | Penjelasan                                                                                         |
| ------------------------------------------------ | -------------------------------------------------------------------------------------------------- |
| ❌ Menganggap korelasi berarti sebab-akibat       | Korelasi **tidak berarti** X menyebabkan Y. Bisa jadi ada faktor ketiga yang memengaruhi keduanya. |
| ❌ Menggunakan data non-linear                    | Pearson hanya menangkap **hubungan garis lurus (linear)**, bukan pola melengkung.                  |
| ❌ Tidak menstandardisasi data yang berbeda skala | Pearson aman terhadap skala, tapi tetap butuh **data numerik** dan **tidak terlalu outlier**.      |
| ❌ Mengabaikan outlier                            | Satu titik ekstrem bisa **mengubah nilai r secara drastis**.                                       |
| ❌ Salah tanda interpretasi                       | r negatif bukan berarti “buruk”, tapi artinya **berlawanan arah**.                                 |

$$ \ $$

---
**Contoh**

Misal data mini berikut:

| Hari | X (Suhu) | Y (Penjualan Es Krim) |
| ---- | -------- | --------------------- |
| 1    | 28       | 100                   |
| 2    | 30       | 130                   |
| 3    | 32       | 150                   |

Langkah:

- Hitung rata-rata:

   * $\bar{X} = 30$
   * $\bar{Y} = 126.7$

- Hitung penyimpangan dari rata-rata dan kalikan:

  | Hari |$x_i$-$\bar{x}$|$y_i$-$\bar{y}$| Produk |
  | ---- | --------------- | --------------- | ------ |
  | 1    | -2              | -26.7           | 53.4   |
  | 2    | 0               | 3.3             | 0      |
  | 3    | 2               | 23.3            | 46.6   |

- Jumlahkan produk = **100**
   Jumlah kuadrat X = 8
   Jumlah kuadrat Y = 1250

- Maka:
   $$ \ $$
   $$
   r = \frac{100}{\sqrt{8 \times 1250}} = \frac{100}{100} = 1
   $$

Hasilnya **1**, artinya hubungan **sempurna positif** — setiap kali suhu naik, penjualan juga naik secara proporsional.


---
**Implementasi**

````python
import pandas as pd
from scipy.stats import pearsonr

df = pd.DataFrame({
    'tinggi': [150, 155, 160, 165, 170, 175, 180],
    'berat': [45, 48, 52, 55, 60, 65, 70]
})

r, p = pearsonr(df['tinggi'], df['berat'])
print(f"Pearson r = {r:.3f}, p-value = {p:.10f}")
````

#### **Spearman Rank**

**Konsep**  

- Bayangkan kamu ingin tahu:

  > “Apakah dua hal berubah **bersamaan arah**, meski tidak **segaris lurus**?”

  Misalnya:

  * X = waktu belajar
  * Y = nilai ujian

  Kalau semakin banyak belajar **selalu** bikin nilai **semakin tinggi**, itu jelas hubungan **searah**.
  Tapi hubungan itu **tidak harus lurus sempurna**, bisa melengkung juga.

  Nah, **Spearman correlation** mengukur seberapa kuat **urutan (ranking)** antara dua hal bergerak bersama.

- Analogi

  Bayangkan kamu punya dua daftar ranking:

  | Siswa | Ranking Belajar | Ranking Nilai |
  | ----- | --------------- | ------------- |
  | A     | 1               | 2             |
  | B     | 2               | 1             |
  | C     | 3               | 3             |

  Kalau ranking mereka **naik-turun bersama**, berarti korelasinya **tinggi**.
  Kalau satu ranking naik, tapi yang lain turun jauh, berarti korelasinya **rendah**.

  Jadi Spearman tidak peduli angka aslinya (70, 80, 90)— yang penting **urutan siapa lebih besar dari siapa**.


$$ \ $$

---
**Intuisi**

Pearson melihat garis lurus (linear).  
Spearman melihat urutan (monotonic).

| Hubungan                   | Pearson       | Spearman      |
| -------------------------- | ------------- | ------------- |
| Garis lurus                | ✅ Bagus       | ✅ Bagus       |
| Melengkung tapi naik terus | ❌ Tidak cocok | ✅ Masih bagus |
| Naik-turun acak            | ❌ Buruk       | ❌ Buruk       |

$$ \ $$
Jadi Spearman cocok kalau:

- Data punya skala ordinal (misal ranking, rating bintang)
- Hubungan tidak linear tapi arahnya tetap sama


$$ \ $$

---
**Rumus**

Rumus sederhananya:

$$ \ $$
$$
r_s = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
$$
$$ \ $$

Keterangan:

  * $d_i^2$ = seberapa besar beda posisi ranking

  * $\sum d_i^2$ = seberapa sering peringkat mereka “tidak searah”
  * $n(n^2 - 1)/6$ = ukuran maksimal ketidaksamaan ranking
  * Jadi kita pakai 1 dikurangi (perbandingan itu) →
  hasilnya antara -1 (berlawanan total) dan +1 (searah total)

  $$ \ $$
Intuisi Rumus

  | Nilai $r_s$ | Makna                                      |
  | ------------- | ------------------------------------------ |
  | +1            | Ranking X dan Y sama persis (naik bersama) |
  | 0             | Tidak ada pola urutan yang jelas           |
  | -1            | Ranking X kebalikan sempurna dari Y        |


$$ \ $$

---
**Kapan digunakan**

Gunakan **Spearman** kalau:

* Hubungan **tidak linear tapi tetap “monoton”** (naik terus atau turun terus).
* Data punya **outlier** yang ekstrem.
* Data mungkin ordinal (ranking, skor, tingkat kepuasan 1–5).
* Kamu ingin tahu apakah **urutan** antar dua variabel konsisten.

$$ \ $$

---
**Contoh**  

| Siswa | X (belajar) | Rank X | Y (nilai) | Rank Y | d | d² |
|-------|-------------|--------|-----------|--------|----|----|
| A | 2 | 1 | 50 | 1 | 0 | 0 |
| B | 4 | 2 | 60 | 2 | 0 | 0 |
| C | 6 | 3 | 65 | 3 | 0 | 0 |
| D | 8 | 4 | 80 | 4 | 0 | 0 |

$$ \ $$

$$
r_s = 1 - \frac{6(0)}{4(16 - 1)} = 1
$$
$$ \ $$
Korelasi sempurna positif.

Kalau rankingnya terbalik semua, misal nilai makin kecil saat jam belajar naik:

$$ \ $$

$$
r_s = -1
$$
$$ \ $$
(karena semua $d_i$ besar dan arah berlawanan)

---

**Jika Ada “Ties” (Nilai Sama)**

Kalau dua nilai sama (misal dua siswa punya nilai 80),
kita kasih **rata-rata rankingnya**.

Contoh:  

Nilai = [60, 70, 80, 80, 90]  

Rank-nya:

* 60 → 1
* 70 → 2
* 80 → (3 + 4)/2 = 3.5
* 80 → 3.5
* 90 → 5

Dan setelah dapat ranking rata-rata, baru lanjut ke rumus yang sama.

$$ \ $$

---

**Kesalahan Umum dan Miskonsepsi**

| Kesalahan                                    | Penjelasan                                                                         |
| -------------------------------------------- | ---------------------------------------------------------------------------------- |
| ❌ Mengira Spearman hanya untuk ordinal       | Bisa juga untuk data numerik kalau kamu curiga **hubungan tidak linear**           |
| ❌ Menggunakan data yang tidak bisa diurutkan | Kalau data kategorikal murni (warna, kota, nama) → tidak bisa diranking            |
| ❌ Lupa ubah ke ranking sebelum hitung manual | Spearman = Pearson pada **ranking**, bukan pada nilai asli                         |
| ❌ Tidak perhatikan ties                      | Harus dirata-ratakan kalau ada nilai sama                                          |
| ❌ Menganggap Spearman lebih “lemah”          | Sebenarnya dia **lebih kuat menghadapi outlier dan non-linear**, bukan lebih lemah |

$$ \ $$

---
Perbandingan Ringkas Pearson vs Spearman

| Aspek                     | **Pearson**                                                                                              | **Spearman**                                  |
| ------------------------- | -------------------------------------------------------------------------------------------------------- | --------------------------------------------- |
| Tipe hubungan             | Linear                                                                                                   | Monotonic (naik/turun terus)                  |
| Data yang dipakai         | Nilai asli                                                                                               | Ranking                                       |
| Sensitif terhadap outlier | Ya                                                                                                       | Tidak terlalu                                 |
| Cocok untuk               | Data interval / rasio                                                                                    | Data ordinal atau non-linear                  |
| Rumus                     | $r$ = $\frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$ | $r_s = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}$ |

$$ \ $$

---
**Implementasi**

````python
from scipy.stats import spearmanr

nilai_seleksi = [1, 2, 3, 4, 5, 6]
nilai_kerja   = [2, 1, 3, 5, 4, 6]

rho, p = spearmanr(nilai_seleksi, nilai_kerja)
print(f"Spearman rho = {rho:.3f}, p-value = {p:.3f}")
````

#### **Kendall Tau**

**Konsep**  

Bayangkan kamu punya **pasangan data (Xᵢ, Yᵢ)** untuk banyak orang.

Contoh sederhana:

| Orang | X (Tinggi) | Y (Berat) |
| ----- | ---------- | --------- |
| A     | 160        | 50        |
| B     | 165        | 55        |
| C     | 170        | 70        |

Nah, Kendall tidak tertarik pada nilai-nilai angka mentah (160, 165, 170).
Dia hanya peduli **urutan (ranking)**: siapa lebih besar, siapa lebih kecil.

Dia bilang:

> “Kalau dua orang urutannya sama di X dan Y, itu pasangan yang *kompak* (concordant).
> Kalau urutannya berlawanan, itu *tidak kompak* (discordant).”


$$ \ $$

---
**Intuisi**

- Ambil dua observasi, misalnya orang **A dan B**.

  Kita lihat dua hal:

  * Apakah urutan X sama dengan urutan Y?
  * Atau berlawanan?

  Kondisi

  - Kasus searah (concordant)

    Jika
    $X_A < X_B$ dan $Y_A < Y_B$  

    atau

    $X_A > X_B$ dan $Y_A > Y_B$  

    Maka pasangan ini **concordant (+1)**

  - Kasus berlawanan (discordant)

    Jika
    $X_A < X_B$ tapi $Y_A > Y_B$  

    atau  

    $X_A > X_B$ tapi $Y_A < Y_B$  

    Maka pasangan ini **discordant (–1)**

  Kalau kebetulan sama $X_A = X_B$ atau $Y_A = Y_B$ → disebut **tie**, nilainya 0 (tidak ikut dihitung pada versi dasar).
  $$ \ $$

- Kendall itu seperti “cek berapa banyak pasangan data yang setuju urutannya”.

  > Kalau dua variabel cenderung menempatkan pasangan (i, j) dalam urutan yang sama (keduanya naik/turun), maka τ positif besar.

$$ \ $$

---
**Rumus**


$$ \ $$


---
**Kapan digunakan**

Gunakan **Kendall** kalau:

* Data **ordinal** (berupa ranking, bukan nilai mutlak).
* Banyak **ties** (nilai yang sama, misal banyak orang dapat ranking sama).
* Dataset kecil, tapi kamu ingin korelasi yang **lebih stabil dan interpretable**.

### **Kategorikal-Kategorikal**

#### **Chi-Square Test of Independence**

**Konsep**  

ChChi-Square Test of Independence digunakan untuk menjawab pertanyaan:

> “Apakah dua variabel kategorikal punya hubungan, atau mereka berdiri sendiri (independen)?”

Contoh Pertanyaan

> Apakah jenis kelamin memengaruhi preferensi minuman?

- Variabel 1: Jenis kelamin → (Pria, Wanita)
- Variabel 2: Minuman → (Kopi, Teh)

Kalau ternyata proporsi orang yang suka kopi/teh sama aja di pria dan wanita, berarti dua variabel ini independen (nggak saling pengaruh).
Kalau proporsinya beda jauh, berarti ada hubungan.

$$ \ $$

---
**Intuisi**

Bayangin kamu bikin tabel kayak gini:

|           | Kopi | Teh | Total |
| --------- | ---- | --- | ----- |
| Pria      | 30   | 20  | 50    |
| Wanita    | 10   | 40  | 50    |
| **Total** | 40   | 60  | 100   |

$$ \ $$

Sekarang, kalau jenis kelamin dan minuman tidak berhubungan,
kita bisa mengharapkan jumlahnya mengikuti proporsi keseluruhan.

👉 Misal, 40% orang suka kopi (40 dari 100), jadi di antara 50 pria, kita harap-harap 40% × 50 = 20 pria suka kopi.
Tapi ternyata yang suka kopi 30 orang pria! Wah, beda jauh.

Jadi Chi-Square test itu intinya:

> "Seberapa jauh data asli berbeda dari yang seharusnya (ekspektasi) jika tidak ada hubungan?"

$$ \ $$

---
**Rumus**

- Expected Frequency (E)

  > Expected frequency (E) adalah berapa kali seharusnya sesuatu terjadi
  kalau dua variabelnya tidak punya hubungan sama sekali (independen).

  Dengan kata lain, E adalah frekuensi “harapan” yang dihitung berdasarkan proporsi umum dari tiap kategori.

  $$ \ $$
  $$
  E = \frac{(\text{total baris}) \times (\text{total kolom})}{\text{grand total}}
  $$
  $$ \ $$


- Chi Square  
  Rumus ini berasal dari ide:

  > “Seberapa jauh distribusi data nyata (O) berbeda dari distribusi yang diharapkan (E)?”

  Kita pakai konsep dari **Goodness of Fit**, mirip dengan menghitung “total error” dalam statistik.  

  Rumus dasarnya adalah:
  $$ \ $$
  $$
  \chi^2 = \sum \frac{(O - E)^2}{E}
  $$
  $$ \ $$
  Mari kita pecah satu-satu:

  | Simbol                  | Arti                     | Bahasa Bayi                                                                                                     |
  | ----------------------- | ------------------------ | --------------------------------------------------------------------------------------------------------------- |
  | $O$                   | *Observed frequency*     | Angka yang benar-benar terjadi di data (misalnya: 30 pria suka kopi)                                            |
  | $E$                   | *Expected frequency*     | Angka yang seharusnya terjadi kalau nggak ada hubungan (misalnya: 20 pria suka kopi, dari perhitungan proporsi) |
  | $O$ - $E$               | Selisih                  | Seberapa “aneh” atau “beda” data nyata dari harapan                                                             |
  | $(O$ - $E)^2$           | Kuadrat selisih          | Biar selisih positif/negatif jadi sama-sama dianggap besar                                                      |
  | $\frac{(O - E)^2}{E}$ | Proporsi selisih relatif | Mengukur “kebesaran selisih” dengan menyesuaikan skala E (biar adil kalau E kecil/besar)                        |

  Lalu semua ini dijumlahkan untuk semua sel tabel → hasil akhirnya disebut **nilai Chi-Square (χ²)**.



  $$ \ $$

---
**Asal-Usul Rumus & Penurunan**



Langkah asal matematiknya kira-kira begini:

- Kita punya dua distribusi:

   * **Distribusi nyata:** proporsi berdasarkan data
   * **Distribusi harapan:** proporsi berdasarkan asumsi independen
- Untuk menghitung “jarak” antara dua distribusi, dipakai bentuk mirip **kuadrat selisih**, karena kuadrat menjaga tanda positif:
   $$ \ $$
   $$
   D = \sum (O - E)^2
   $$
   $$ \ $$

- Tapi... jumlah “orang” di setiap kategori bisa beda besar.
   Jadi kalau E-nya besar, selisih 10 orang *tidak terlalu berarti*, tapi kalau E-nya kecil (misal cuma 5), selisih 10 itu *wow banget*.

   Jadi kita perlu **menormalkan** dengan (E):
   $$ \ $$
   $$
   \chi^2 = \sum \frac{(O - E)^2}{E}
   $$
   $$ \ $$

- Lalu, secara teori statistik, nilai χ² ini kalau variabelnya benar-benar independen, akan mengikuti **distribusi chi-square** dengan derajat bebas tertentu:
   $$ \ $$
   $$
   df = (r - 1)(c - 1)
   $$
   $$ \ $$
   (r = jumlah baris, c = jumlah kolom)

$$ \ $$

---
**Kesalahan Umum & Miskonsepsi**

| Kesalahan                                                                 | Penjelasan                                                                                                                                                        |
| ------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| ❌ Menggunakan Chi-Square untuk data numerik                               | Harus dua variabel **kategorikal**, bukan angka kontinu                                                                                                           |
| ❌ Frekuensi terlalu kecil (E < 5)                                         | Rumus jadi tidak valid, sebaiknya gunakan **Fisher’s Exact Test**                                                                                                 |
| ❌ Menafsirkan besar kecilnya χ² langsung sebagai “kuat lemahnya hubungan” | χ² besar ≠ hubungan kuat. Itu hanya tanda bahwa *ada hubungan signifikan*, bukan ukurannya                                                                        |
| ❌ Lupa hitung *expected frequency*                                        | Banyak orang langsung pakai data mentah tanpa menghitung nilai harapan                                                                                            |
| ❌ Tidak tahu arah hubungan                                                | Chi-square hanya bilang “ada hubungan”, tapi tidak bilang “hubungannya bagaimana” (positif/negatif, |
|| siapa suka apa). Untuk itu perlu lihat **proporsi tabelnya**.|


$$ \ $$

---
**Contoh**  

Gunakan tabel tadi:

|           | Kopi | Teh | Total |
| --------- | ---- | --- | ----- |
| Pria      | 30   | 20  | 50    |
| Wanita    | 10   | 40  | 50    |
| **Total** | 40   | 60  | 100   |

$$ \ $$

- Langkah 1: Hitung E (expected frequency)


  |        | Kopi             | Teh              |
  | ------ | ---------------- | ---------------- |
  | Pria   | (50×40)/100 = 20 | (50×60)/100 = 30 |
  | Wanita | (50×40)/100 = 20 | (50×60)/100 = 30 |

  $$ \ $$

- Langkah 2: Hitung (O−E)² / E

  |             | O  | E  | (O−E)²/E         |
  | ----------- | -- | -- | ---------------- |
  | Pria-Kopi   | 30 | 20 | (10)²/20 = 5     |
  | Pria-Teh    | 20 | 30 | (−10)²/30 ≈ 3.33 |
  | Wanita-Kopi | 10 | 20 | (−10)²/20 = 5    |
  | Wanita-Teh  | 40 | 30 | (10)²/30 ≈ 3.33  |

  Total:
  $$ \ $$
  $$
  \chi^2 = 5 + 3.33 + 5 + 3.33 = 16.66
  $$
  $$ \ $$

- Langkah 3: Bandingkan dengan nilai kritis dengan derajat kebebasan
  
   $$ \ $$
   $$
   df = (r - 1)(c - 1)
   $$
   $$ \ $$
   (r = jumlah baris, c = jumlah kolom)
   $$ \ $$
   maka:  
  $$df = (2−1)(2−1) = 1$$
  $$ \ $$
  Dari tabel chi-square (α=0.05), nilai kritis ≈ 3.84. Karena 16.66 > 3.84 → **tolak H₀**, berarti ada hubungan antara jenis kelamin dan preferensi minuman.


---
**Implementasi**

````python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Buat tabel kontingensi
data = pd.DataFrame({
    "JenisKelamin": ["Pria"] * 40 + ["Wanita"] * 40,
    "SukaKopi":     [1]*30 + [0]*10 + [1]*10 + [0]*30
})

# Buat crosstab
ct = pd.crosstab(data["JenisKelamin"], data["SukaKopi"])

# Hitung Chi-Square
chi2, p, dof, expected = chi2_contingency(ct)
chi2
````

#### **Phi Coefficient**

**Konsep**  

  > Phi Coefficient (φ) adalah ukuran kekuatan hubungan antara dua variabel kategorikal biner (misal: pria/wanita, suka/tidak suka).

  Kalau Chi-Square bilang “ada hubungan atau tidak”, maka Phi menjawab “seberapa kuat hubungannya”.

$$ \ $$

---
**Intuisi**

Bayangin Chi-Square kayak “detektor hubungan” (nyala atau mati),
sementara Phi kayak “termometer hubungan” (seberapa panas hubungan itu).

Contoh

|            | Suka Kopi | Tidak Suka | Total |
| ---------- | --------- | ---------- | ----- |
| **Pria**   | 30        | 10         | 40    |
| **Wanita** | 10        | 30         | 40    |
| **Total**  | 40        | 40         | 80    |

$$ \ $$

Kita lihat, sepertinya pria lebih suka kopi, wanita lebih tidak suka —
nah, Phi membantu kita **mengukur kekuatan** hubungan ini.

$$ \ $$

---
**Rumus**  

$$ \ $$
$$
\phi = \sqrt{\frac{\chi^2}{n}}
$$
$$ \ $$

Atau bisa juga dari langsung tabel kontingensi 2×2:

|     | Positif (B=1) | Negatif (B=0) |
| --- | ------------- | ------------- |
| A=1 | a             | b             |
| A=0 | c             | d             |

$$ \ $$
Rumusnya:

$$ \ $$
$$
\phi = \frac{ad - bc}{\sqrt{(a+b)(a+c)(b+d)(c+d)}}
$$
$$ \ $$
Intuisi di Balik Rumus

- Bagian 1: Numerator → “arah hubungan”

  $$ \ $$
  $$
  ad - bc
  $$
  $$ \ $$

  * $a$ dan $d$ : “sepakat” (dua variabel cocok → sama-sama tinggi atau sama-sama rendah)
  * $b$ dan $c$ : “tidak sepakat” (berlawanan arah)

  Jadi $ad - bc$ itu mengukur **berapa banyak kesepakatan dibanding ketidaksepakatan**.
  Kalau nilainya besar → dua variabel “jalan bareng” (berhubungan).

- Bagian 2: Denominator → “penyeimbang skala”

  $$ \ $$
  $$
  \sqrt{(a+b)(a+c)(b+d)(c+d)}
  $$
  $$ \ $$

  Ini memastikan hasil Phi selalu antara -1 dan 1,
  sehingga bisa dibandingkan seperti **Pearson correlation**.

  * Nilai +1 → hubungan sempurna positif (selalu bareng)
  * Nilai -1 → hubungan sempurna negatif (selalu berlawanan)
  * Nilai 0 → tidak ada hubungan (independen)

$$ \ $$

---
**Intuisi Probabilistiknya**

Kalau kita tulis dalam bentuk probabilitas (bukan frekuensi), dengan $P_{ij} = \frac{\text{frekuensi}}{n}$, maka:

$$ \ $$
$$
\phi = \frac{P_{11}P_{22} - P_{12}P_{21}}{\sqrt{P_{1.}P_{2.}P_{.1}P_{.2}}}
$$
$$ \ $$
* $P_{1.}$ = probabilitas baris 1 (misalnya pria)
* $P_{.1}$ = probabilitas kolom 1 (misalnya suka kopi)
* dst.

Kalau variabel independen (tidak berhubungan), maka:
$$ \ $$
$$
P_{11}P_{22} = P_{12}P_{21}
$$
$$ \ $$⟹ φ = 0

Kalau ada hubungan kuat (semua pria suka kopi, semua wanita tidak suka), maka:

$$ \ $$

$$
φ → 1
$$

$$ \ $$

---
**Hubungan dengan Chi-Square**

Phi sebenarnya versi “ternormalisasi” dari Chi-Square.

Karena:
$$ \ $$
$$
\chi^2 = n\phi^2
$$
$$ \ $$


Artinya kalau kamu udah punya hasil Chi-Square dan jumlah total data (n),
kamu bisa langsung dapetin Phi dengan akar hasil bagi:
$$ \ $$
$$
\phi = \sqrt{\frac{\chi^2}{n}}
$$
$$ \ $$

---
**Contoh**

Kita pakai data:

|            | Suka Kopi | Tidak Suka | Total |
| ---------- | --------- | ---------- | ----- |
| **Pria**   | a = 30    | b = 10     | 40    |
| **Wanita** | c = 10    | d = 30     | 40    |
| **Total**  | 40        | 40         | 80    |

Masukkan ke rumus:

$$ \ $$
$$
\phi = \frac{(30×30) - (10×10)}{\sqrt{(30+10)(30+10)(10+30)(10+30)}} = \frac{900 - 100}{\sqrt{(40)(40)(40)(40)}} = \frac{800}{1600} = 0.5
$$
$$ \ $$

Jadi:
$$ \ $$
$$
\phi = 0.5
$$
$$ \ $$

Artinya ada hubungan **cukup kuat** antara jenis kelamin dan kesukaan kopi.

$$ \ $$

---
**Interpretasi**  

|        | Nilai                     | Makna |
| ------ | ------------------------- | ----- |
| 🔹 0   | Tidak ada hubungan        |       |
| 🔹 0.1 | Lemah                     |       |
| 🔹 0.3 | Sedang                    |       |
| 🔹 0.5 | Kuat                      |       |
| 🔹 1   | Sempurna (hubungan total) |       |

Tapi ini **bukan batas baku** — hanya panduan umum seperti interpretasi korelasi Pearson.

$$ \ $$

---

**Kesalahan Umum dan Miskonsepsi**

| Kesalahan                           | Penjelasan                                                                                                    |
| ----------------------------------- | ------------------------------------------------------------------------------------------------------------- |
| ❌ Menggunakan Phi untuk tabel > 2×2 | Phi hanya cocok untuk tabel 2×2, kalau lebih besar pakai **Cramer’s V**                                       |
| ❌ Mengira Phi = Signifikansi        | Phi mengukur **kekuatan**, bukan apakah hubungan itu signifikan (untuk signifikan tetap pakai uji Chi-Square) |
| ❌ Mengabaikan arah hubungan         | Nilai Phi tidak menunjukkan *arah*, hanya besar kecilnya hubungan                                             |
| ❌ Menggunakan data numerik          | Phi hanya buat variabel **kategorikal biner**                                                                 |
| ❌ Salah menghitung E atau n         | Karena Phi bergantung langsung pada hasil Chi-Square dan total data                                           |

$$ \ $$

---
**Implementasi**

````python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Buat tabel kontingensi
data = pd.DataFrame({
    "JenisKelamin": ["Pria"] * 40 + ["Wanita"] * 40,
    "SukaKopi":     [1]*30 + [0]*10 + [1]*10 + [0]*30
})

# Buat crosstab
ct = pd.crosstab(data["JenisKelamin"], data["SukaKopi"])

# Hitung Chi-Square
chi2, p, dof, expected = chi2_contingency(ct)

# Hitung Phi Coefficient
n = ct.sum().sum()
phi = np.sqrt(chi2 / n)
````

````python
print(f"Chi-Square: {chi2:.2f}, Phi: {phi:.2f}, p-value: {p:.4f}")
````

#### **Cramér’s V**

**Konsep**  

> **Cramér’s V** adalah ukuran **kekuatan hubungan** antara dua **variabel kategorikal**
> — sama seperti **Phi**, tapi bisa dipakai untuk **tabel lebih besar dari 2×2** (misalnya 3×3, 4×2, dst).

Analogi sederhana

* **Chi-Square** ➜ bilang “ada hubungan atau tidak”
* **Phi** ➜ mengukur kekuatan hubungan kalau tabelnya **2×2**
* **Cramér’s V** ➜ mengukur kekuatan hubungan kalau tabelnya **lebih besar dari 2×2**

Jadi:

> Cramér’s V = Phi yang lebih fleksibel ✨

$$ \ $$

---
**Intuisi**

Chi-Square (χ²) memberi tahu *berapa besar selisih antara data nyata dan yang diharapkan*,
tapi **tidak punya batas atas** → jadi nggak bisa bilang “hubungannya kuat atau lemah” secara relatif.

Cramér’s V memperbaikinya dengan **menormalisasi nilai Chi-Square** supaya hasilnya **selalu antara 0 dan 1**:

* 0 = tidak ada hubungan
* 1 = hubungan sempurna

$$ \ $$

---
**Rumus**

$$ \ $$
$$
V = \sqrt{\frac{\chi^2}{n(k - 1)}}
$$
$$ \ $$

Keterangan:

| Simbol     | Arti                            | Bahasa Bayi                                   |
| ---------- | ------------------------------- | --------------------------------------------- |
| $\chi^2$ | Nilai Chi-Square                | “Seberapa jauh data nyata dari harapan”       |
| $n$      | Total data                      | Banyaknya seluruh pengamatan                  |
| $k$      | Min(jumlah baris, jumlah kolom) | Ukuran tabel terkecil (biar normalisasi adil) |

$$ \ $$

Intuisi:  
Kalau kita lihat bentuk rumus:

$$ \ $$
$$
V = \sqrt{\frac{\chi^2}{n(k - 1)}}
$$
$$ \ $$

* Bagian atas $\chi^2$ : “besar selisih antara dunia nyata dan dunia ideal”
* Bagian bawah $n(k-1)$ : “penyeimbang” → menyesuaikan dengan ukuran data dan kompleksitas tabel
  (semakin besar tabel, semakin banyak kemungkinan kombinasi, jadi perlu pembagian biar adil)
* Akar kuadrat: untuk mengembalikan hasil ke skala 0–1, biar mudah dibaca kayak “kekuatan hubungan”.

$$ \ $$

---
**Contoh**

Misal kita punya data tentang **Jenis Kelamin** dan **Jenis Minuman Favorit**:

| Jenis Kelamin | Kopi | Teh | Jus | Total |
| ------------- | ---- | --- | --- | ----- |
| Pria          | 40   | 20  | 40  | 100   |
| Wanita        | 10   | 50  | 40  | 100   |
| **Total**     | 50   | 70  | 80  | 200   |

$$ \ $$
- Hitung nilai Chi-Square

  Pakai rumus biasa atau Python:
  $$ \ $$
  $$
  \chi^2 = \sum \frac{(O - E)^2}{E}
  $$
  $$ \ $$

  Kita langsung aja hasil hitungan (biar nggak panjang):
  $$ \ $$
  $$
  \chi^2 = 25.71
  $$
  $$ \ $$

  Jumlah total data $n = 200$

  Jumlah baris = 2
  Jumlah kolom = 3
  → $k = \min(2, 3) = 2$

- Masukkan ke rumus Cramér’s V

  $$ \ $$
  $$
  V = \sqrt{\frac{25.71}{200(2 - 1)}} = \sqrt{\frac{25.71}{200}} = \sqrt{0.12855} = 0.358
  $$
  $$ \ $$

- Interpretasi

  $$ \ $$
  $$
  V = 0.36
  $$
  $$ \ $$
  Artinya ada **hubungan sedang** antara jenis kelamin dan jenis minuman favorit.
  $$ \ $$

---

**Interpretasi Nilai Cramér’s V**

| Nilai V   | Interpretasi Umum           |
| --------- | --------------------------- |
| 0.0 – 0.1 | Tidak ada atau sangat lemah |
| 0.1 – 0.3 | Lemah                       |
| 0.3 – 0.5 | Sedang                      |
| > 0.5     | Kuat                        |

$$ \ $$
> Catatan: ini bukan batas matematis mutlak, tapi **aturan jempol** yang sering dipakai oleh data analyst.

$$ \ $$

---
**Kesalahan Umum dan Miskonsepsi**

| Kesalahan                                 | Penjelasan                                                                                                                     |
| ----------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| ❌ Pakai Phi untuk tabel besar             | Phi hanya valid untuk 2×2. Kalau lebih besar, hasilnya bisa >1 (tidak valid).                                                  |
| ❌ Anggap Cramér’s V = signifikansi        | Cramér’s V hanya ukuran **kekuatan**, bukan “ada hubungan atau tidak”. Untuk signifikansi tetap lihat p-value dari Chi-Square. |
| ❌ Tidak tahu arah hubungan                | Cramér’s V tidak bilang “siapa suka apa”, cuma seberapa kuat hubungannya.                                                      |
| ❌ Nilai V besar = hubungan sebab-akibat   | Salah! Ini hanya asosiasi/statistik, bukan hubungan sebab-akibat.                                                              |
| ❌ Frekuensi terlalu kecil di beberapa sel | Uji Chi-Square (dan otomatis Cramér’s V) bisa jadi tidak valid jika banyak nilai harapan < 5.                                  |

**Implementasi**

````python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Data
data = pd.DataFrame({
    "JenisKelamin": ["Pria"]*100 + ["Wanita"]*100,
    "Minuman": ["Kopi"]*40 + ["Teh"]*20 + ["Jus"]*40 +
                ["Kopi"]*10 + ["Teh"]*50 + ["Jus"]*40
})

# Tabel kontingensi
ct = pd.crosstab(data["JenisKelamin"], data["Minuman"])

# Chi-Square
chi2, p, dof, expected = chi2_contingency(ct)

# Hitung Cramer's V
n = ct.sum().sum()
k = min(ct.shape)
cramers_v = np.sqrt(chi2 / (n * (k - 1)))
````

````python
print(f"Chi-Square: {chi2:.2f}, p-value: {p:.4f}, Cramer's V: {cramers_v:.3f}")
````

### **Kategorikal–Numerik**

#### **Point-Biserial Correlation**

**Konsep**  

Point-Biserial Correlation $r_pb$ itu:

Cara mengukur hubungan antara 1 variabel dikotomi (cuma 2 nilai, misal: “lulus/gagal”) dan 1 variabel numerik (misal: nilai ujian).

> Artinya:
Apakah perbedaan rata-rata nilai antara dua kelompok itu bermakna dan konsisten, bukan cuma kebetulan.

Contoh:  
| Siswa | Status (Lulus=1, Gagal=0) | Nilai Ujian |
| ----- | ------------------------- | ----------- |
| A     | 1                         | 90          |
| B     | 1                         | 85          |
| C     | 0                         | 60          |
| D     | 0                         | 55          |

$$ \ $$
Kita mau tahu:
“Apakah status lulus/gagal berhubungan dengan nilai ujian?”

> Kemungkinan:
> - Kalau yang lulus rata-ratanya jauh lebih tinggi dari yang gagal, berarti ada hubungan positif.  
> - Kalau malah yang gagal punya nilai lebih tinggi, berarti hubungan negatif.  
> - Kalau sama saja, berarti tidak ada hubungan.

---
**Intuisi**

Bayangkan kamu punya dua tumpukan nilai:

- Tumpukan 1 (Y₁): nilai orang yang lulus
- Tumpukan 0 (Y₀): nilai orang yang gagal

Kalau kedua tumpukan ini jauh beda rata-ratanya, artinya nilai memang berhubungan dengan status lulus/gagal.

Kalau rata-ratanya mirip, berarti nggak ada hubungan.

Jadi, si Point-Biserial ini cuma bilang:

> “Seberapa jauh dua dunia (nilai lulus vs gagal) itu berbeda?”

$$ \ $$

---
**Rumus**

Rumusnya begini:

$$ \ $$
$$
r_{pb} = \frac{\bar{Y_1} - \bar{Y_0}}{s_Y} \sqrt{p q}
$$
$$ \ $$
Keterangan:  
$$ \ $$

| Simbol        | Artinya                                                       | Intuisi                                            |
| ------------- | ------------------------------------------------------------- | -------------------------------------------------- |
| $\bar{Y_1}$ | Rata-rata nilai untuk kelompok yang nilainya 1 (misal: lulus) | Seberapa tinggi rata-rata di kelompok 1            |
| $\bar{Y_0}$ | Rata-rata nilai untuk kelompok 0 (misal: gagal)               | Seberapa tinggi rata-rata di kelompok 0            |
| $s_Y$       | Simpangan baku dari semua nilai (gabungan dua kelompok)       | Ukuran seberapa “menyebar” nilai-nilai secara umum |
| $p$         | Proporsi anggota kelompok 1 (misal: persentase yang lulus)    | Berapa banyak yang 1                               |
| $q$         | Proporsi anggota kelompok 0 (yaitu 1 - p)                     | Berapa banyak yang 0                               |

$$ \ $$
Intuisi:

- $(\bar{Y_1} - \bar{Y_0})$ =  “Jarak” rata-rata dua kelompok $\to$ Semakin besar bedanya, semakin kuat hubungannya.  

- $\frac{(\bar{Y_1} - \bar{Y_0})}{s_Y}\to$ Telah dinormalisasi terhadap sebaran data (supaya adil).

- $\sqrt{pq}$ = Koreksi karena ukuran kelompok bisa nggak seimbang $\to$ Kalau satu kelompok terlalu kecil, kontribusinya menurun.

$$ \ $$

---
**Contoh**

| Siswa | Status (X) | Nilai (Y) |
| ----- | ---------- | --------- |
| A     | 1          | 90        |
| B     | 1          | 85        |
| C     | 0          | 60        |
| D     | 0          | 55        |

$$ \ $$
Langkah-langkah:

* $\bar{Y_1} = (90 + 85)/2 = 87.5$
* $\bar{Y_0} = (60 + 55)/2 = 57.5$
* $s_Y =$ simpangan baku seluruh nilai = 15.0
* $p = 0.5, \ \ q = 0.5$
  $$ \ $$

Perhitungan
$$ \ $$
$$
r_{pb} = \frac{87.5 - 57.5}{15.0} \times \sqrt{0.5 \times 0.5}
$$
$$ \ $$
$$
r_{pb} = \frac{30}{15} \times 0.5 = 1.0 \times 0.5 = 0.5
$$

Ada **korelasi positif sedang (0.5)**
Artinya: makin besar nilainya, makin besar kemungkinan lulus.

$$ \ $$

---
**Kesalahan Umum dan Miskonsepsi**

| Kesalahan                                | Penjelasan                                                        |
| ---------------------------------------- | ----------------------------------------------------------------- |
| ❌ Menggunakan untuk dua variabel nominal | Point-biserial cuma buat **1 numerik + 1 biner**                  |
| ❌ Menganggap “hubungan sebab-akibat”     | Korelasi ≠ kausalitas                                             |
| ❌ Lupa menyesuaikan nilai 0/1            | Urutan (0=fail, 1=pass) penting, kebalik bisa ubah tanda korelasi |
| ❌ Tidak cek distribusi Y                 | Jika Y sangat tidak normal, bisa bias hasilnya                    |
| ❌ Mengabaikan ukuran kelompok            | Kalau p atau q sangat kecil (<0.1), hasil kurang stabil           |

$$ \ $$

#### **Eta Coefficient (η)**

**Konsep**

Eta coefficient (η) adalah ukuran kekuatan hubungan antara variabel kategorikal (X) dan variabel numerik (Y).

Jadi kalau kamu punya:

- X = kategori (misal: jenis kelamin → laki/perempuan, atau jurusan → IPA/IPS)
- Y = angka (misal: nilai ujian, tinggi badan)

Maka Eta ingin menjawab:

> “Seberapa besar perbedaan nilai Y yang bisa dijelaskan oleh perbedaan kategori X?”

Dengan kata lain:

- Kalau nilai rata-rata Y di setiap kategori X berbeda jauh, berarti ada hubungan kuat → η besar.
- Kalau rata-rata Y di tiap kategori X mirip-mirip, berarti hubungan lemah → η kecil.

$$ \ $$

---
**Intuisi**

Bayangkan kamu guru.
Kamu ingin tahu apakah jurusan siswa (IPA/IPS) berhubungan dengan nilai matematika.

Kamu pisahkan murid berdasarkan jurusan, lalu kamu hitung:

- Rata-rata nilai matematika anak IPA
- Rata-rata nilai matematika anak IPS

Sekarang kamu lihat:

- Kalau beda rata-rata besar → jurusan mungkin berpengaruh (hubungan kuat)
- Kalau beda kecil → jurusan mungkin tidak berpengaruh (hubungan lemah)

Eta (η) mengubah intuisi ini jadi angka dari 0 sampai 1:

- η = 0 → tidak ada hubungan (rata-rata antar kategori sama semua)
- η = 1 → hubungan sempurna (kategori menentukan nilai dengan pasti)

$$ \ $$

---
**Rumus**

Rumus umumnya:
$$ \ $$
$$
η = \sqrt{\frac{SS_{between}}{SS_{total}}}
$$
$$ \ $$

Keterangan:  

* $SS_{total}$ :
  Total variasi seluruh data Y (seberapa lebar penyebaran semua nilai Y dari rata-rata keseluruhan).

  Secara rumus:
  $$ \ $$
  $$
  SS_{total} = \sum (Y_i - \bar{Y})^2
  $$
  $$ \ $$
  Ini mirip total “energi penyimpangan” dari rata-rata.

* $SS_{between}$ :
  Variasi **antara kelompok** (seberapa jauh rata-rata tiap kategori dari rata-rata total).

  $$ \ $$
  $$
  SS_{between} = \sum n_j (\bar{Y_j} - \bar{Y})^2
  $$
  $$ \ $$

  Di sini:

  * $n_j$ : jumlah data dalam kategori ke-j
  * $\bar{Y_j}$ : rata-rata Y dalam kategori j
  * $\bar{Y}$ : rata-rata Y semua kategori

  Jadi ini mengukur *berapa besar jarak antar kelompok*.

Makna intuitif:

* Pembilang $(SS_{between})$ = “variasi karena kategori”
* Penyebut $(SS_{total})$ = “total variasi semua data”

Maka perbandingan:
$$ \ $$
$$
\frac{SS_{between}}{SS_{total}}
$$
$$ \ $$
= proporsi variasi Y yang bisa dijelaskan oleh perbedaan kategori X.

Lalu akar kuadratnya $(\sqrt{})$ dipakai agar hasilnya dalam skala **0–1**, mirip korelasi biasa.


$$ \ $$

---
**Contoh**

Misal kamu punya data:

| Jurusan | Nilai Matematika |
| ------- | ---------------- |
| IPA     | 80, 85, 90       |
| IPS     | 60, 65, 70       |

Langkah-langkah:

- Rata-rata keseluruhan:
   $$ \ $$
   $$
   \bar{Y} = (80 + 85 + 90 + 60 + 65 + 70)/6 = 75
   $$
   $$ \ $$

- Rata-rata tiap jurusan:

   * IPA → $(80+85+90)/3 = 85$
   * IPS → $(60+65+70)/3 = 65$

- Hitung $SS_{between}$:
   $$ \ $$
   $
   SS_{between} = 3(85 - 75)^2 + 3(65 - 75)^2 = 3(100) + 3(100) = 600
   $
   $$ \ $$

- Hitung $SS_{total}$:
   $$ \ $$
   $
   SS_{total} = \sum (Y_i - 75)^2 = (80−75)²+(85−75)²+(90−75)²+(60−75)²+(65−75)²+(70−75)²
   = 25 + 100 + 225 + 225 + 100 + 25 = 700
   $
   $$ \ $$
- Maka:
   $$ \ $$
   $$
   η = \sqrt{600 / 700} = \sqrt{0.857} ≈ 0.926
   $$
   $$ \ $$

  **Interpretasi:**  
  Hubungan antara jurusan dan nilai matematika sangat kuat (η ≈ 0.93).

  $$ \ $$

---
Kesalahan Umum dan Miskonsepsi

- **Menganggap η menunjukkan arah hubungan.**  
   Salah. Eta hanya menunjukkan **kuat atau lemahnya** hubungan, **tanpa arah** (karena X adalah kategori, bukan angka).

- **Menyamakan η dengan korelasi Pearson.**  
   Pearson mengukur hubungan linear antara dua variabel numerik.
   Eta mengukur hubungan antara kategori dan angka (nonlinear pun bisa).

- **Mengira η bisa negatif.**  
   Tidak mungkin. Karena hasilnya dari akar kuadrat, nilainya **0 ≤ η ≤ 1**.

- **Mengabaikan ukuran sampel per kategori.**  
   Kalau satu kategori punya sedikit data, rata-rata kategori bisa menyesatkan — η jadi tidak stabil.

### **Chi Square (Significance)**

**Konsep**  

Kita pakai Chi-Square bukan cuma buat lihat ada hubungan atau tidak,
tapi buat jawab:

> “Hubungan ini beneran pola… atau cuma kebetulan doang?”

Kalau hasilnya bilang “kayaknya ini bukan kebetulan”,
nah hubungan ini bisa masuk daftar kandidat kausal untuk dicek lebih serius.

Bukan berarti sudah pasti sebab-akibat ya — cuma calon aja.

$$ \ $$

---
**Intuisi**

- Kita lihat tabel data (misalnya laki/perempuan × suka kopi/teh).
- Kita hitung:  
  > Kalau tidak ada hubungan, harusnya angkanya jadi berapa (harapan).
- Kita bandingkan dengan angka asli.

- Mesin Chi-Square ngitung:

  > “Perbedaan ini gede banget nggak sih sampai keliatan mencurigakan?”

- Kalau hasilnya p-value kecil →

  > “Wah, ini jarang terjadi kalau murni kebetulan. Ada sesuatu nih.”

$$ \ $$

---

**Setelah dapet hasil signifikan, ngapain?**

Kalau hasilnya "Wah ini bukan kebetulan":

- Lihat proporsinya
→ siapa lebih besar, siapa lebih kecil?

- Lihat ukuran efeknya (mis. Cramér’s V / Phi / OR / RR) untuk menilai practical magnitude → bedanya besar atau cuma sedikit?

- Cek kemungkinan faktor lain (confounder)
→ jangan sampai salah curiga.

- Kalau penting, lanjut ke tes kausal beneran
kayak regresi yang dikontrol, skor kecenderungan, A/B test, dll.


$$ \ $$

---
**Perbedaan dengan Chi Square Test of Independence**  

Chi-Square biasa:

> “Ada hubungan nggak sih? Titik.”

Chi-Square versi signifikansi (buat kandidat kausalitas):

> “Hubungan ini kuat dan konsisten, atau cuma kebetulan?
Kalau bukan kebetulan, ini patut dicurigai jadi calon sebab-akibat.”
